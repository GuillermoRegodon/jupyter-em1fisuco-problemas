# Problema B2.S5.1
 
Tenemos **una esfera a un potencial conocido** de radio $R$ que sigue la expresión $V = K \cos(3\theta)$, donde $(r, \theta, \varphi)$ son las coordenadas **esféricas** e implícitamente, se establece *la simetría en la variable $\varphi$*.

```{note}
Si pidiera el potencial en un punto del eje $z$ o en algún otro punto con una simetría clara, la integración directa sería posible.

Dado que se pide el potencial en cualquier punto externo e interno, la estrategia de los polinomios de Legendre es la única factible.
```

Se trata entonces de **resolver la ecuación de Poisson** con las condiciones de contorno apropiadas.

$$
\nabla^2 \Phi = 0
$$

Para *el volumen interno a la esfera*, la única condición de contorno es el propio potencial $\Phi(r {=} R) = K \cos(3\theta)$ de la esfera. Para *el volumen exterior a la esfera*, además de $\Phi(r {=} R) = K \cos(3\theta)$, usamos el potencial en el infinito, $\Phi(r {\rightarrow} \infty) = 0$.

Este es precisamente el problema que tiene por solución los Polinomios de Legendre:

$$
\Phi = \sum_{l=0}^{\infty} \left( A_l r^l + \frac{B_l}{r^{l+1}} \right) P_l(\cos\theta)
$$

```{note}
El problema trata de poner la expresión $V = K \cos(3\theta)$, que sólo conocemos en la superficie de la esfera ($r = R$), en función de los polinomios de Legendre, de tal forma que la solución es válida en todo el espacio, y para $r=R$ coincide con la del enunciado.

**Por eso, para conseguir la igualdad en $r=R$, tenemos que sustituir la coordenada radial $r$ por $R$.**

Lo podemos imaginar como un problema en dos pasos. El primer paso es el siguiente:
* **Descomponemos la expresión buscada en función de los polinomios de Legendre**, ignorando la coordenada radial.

$$
\Phi = \sum_{l=0}^{\infty} C_l P_l(\cos\theta) = K \cos (3\theta)
$$

Obtenemos por tanto expresiones para $C_l$ que dependen de los datos del problema. El segundo paso sería el siguiente:
* **En cada región, igualamos los coeficientes $C_l$ con las expresiones con $A_l$ y $B_l$ apropiadas, sustituyendo $r=R$**.

$$
C_l = A_l R^l + \frac{B_l}{R^{l+1}}
$$

De tal forma que obtenemos expresiones para $A_l$ y $B_l$ que dependen de los datos del problema.
```

Existen metodologías para obtener estos coeficientes, *pero en nuestros problemas lo intentaremos hacer por inspección*. En general el número de términos necesarios es infinito, pero en nuestro caso *es suficiente con tan solo unos pocos términos*. La condición de contorno que nos dan *es una función de $\cos(3 \theta)$*, por lo que **seguro que necesitamos el término de $l = 3$**. Efectivamente:

$$
P_3(\cos\theta) = \frac{1}{2} \left( 5 \cos^3\theta - 3 \cos\theta \right) = \frac{5}{8} \cos(3 \theta) + \frac{3}{8} \cos\theta 
$$

Vemos que este polinomio de Legendre incluye un término en $\cos(\theta)$, por lo que de por sí solo, podemos conseguir el término $\cos(3 \theta)$, pero nos sobra el de $\cos\theta$. **Por esto tenemos que añadir el término del polinomio de Legendre con $l=1$**.

```{note}
En general, si nos vemos obligados a añadir **un polinomio de orden $l$**, tendremos que añadir también todos los polinomios **de grado inferior $l'$ del mismo tipo par o impar**. (Es decir, si $l$ es par, pondremos todos los términos pares inferiores o iguales a $l$, y si $l$ es impar, todos los términos impares inferiores o iguales a $l$).

Por ejemplo, si la condición de contorno fuera $V = K_1 \cos(9 l) + K_2 \cos(4 l)$, por el primer término tendríamos que poner $l \in \{1, 3, 5, 7, 9\}$ mientras que por el segundo término, tendríamos que poner $l \in \{0, 2, 4\}$ (Ojalá nunca nos encontremos con un problema así).

Obsérvese que los polinomios de Lengendre incluyen términos $\cos^l(\theta)$, que siempre se van a poder poner en forma de **una suma de términos que contiene $\cos(l \theta)$**.
```

Ahora, los coeficientes $C_l$ generales que resuelven la descomposición se pueden obtener fácilmente...

$$
\Phi(r {=} R) = C_1 P_1(\cos(\theta)) + C_3 P_3(\cos(\theta)) = K \cos(3\theta) \\
$$

...nos da el siguiente resultado:

$$
C_1 = -\frac{3 K}{5}; \quad
C_3 = \frac{8 K}{5}
$$

Con esto, tenemos la descomposición en $r=R$, pero aún no en el resto del espacio. Ahora introducimos la dependencia en $r$, de tal forma que la expresión $A_l R^l + \frac{B_l}{R^{l+1}}$ se iguala al valor obtenido en $r=R$.

$$
C_l = A_l R^l + \frac{B_l}{R^{l+1}}
$$

Sólo tenemos una ecuación para dos variables, $A_l$ y $B_l$, por lo que tenemos que utilizar condiciones adicionales. También es importante observar que los coeficientes $C_l$ son únicos, pero los coeficientes $A_l$ y $B_l$ son distintos en el interior y en el exterior de la esfera:

* **Para el volumen interior**, denominamos el potencial $\Phi_d$ y usamos la condición de que todos las magnitudes físicas deben ser finitas y no pueden diverger en el origen $r{\rightarrow} 0$, y por tanto, $B_l = 0$.

* **Para el volumen exterior**, denominamos el potencial $\Phi_f$ y usamos la condición de que el potencial en el infinito se anula, y por tanto, $A_l = 0$.

$$
\Phi_d = \sum_{l=0}^{\infty} A_l r^l P_l(\cos(\theta)) \\
\Phi_f = \sum_{l=0}^{\infty} \frac{B_l}{r^{l+1}} P_l(\cos(\theta))
$$

```{note}
Podemos **resolver los dos problemas por separado o en paralelo**, ya que son independientes. Están separados por la condición de contorno $V$ que descompone el espacio en dos volumenes separados.
```

Si realizamos la igualación de los términos con la condición de contorno...

$$
A_1 = -\frac{3 K}{5 R}; \quad
A_3 = \frac{8 K}{5 R^3}; \quad
B_1 = -\frac{3}{5} K R^2; \quad
B_3 = \frac{8}{5} K R^4
$$

Por supuesto, la expresión de $\Phi_d$ es válida dentro de la esfera y la de $\Phi_f$ es válida fuera de la esfera.

```{note}
En un volumen con simetría esférica nos vamos a encontrar tres posibilidades:

* **Incluye el origen**: La expresión del potencial en general no puede diverger en el origen, así que, salvo que haya alguna carga puntual, dipolo, o algo que lo justifique, $B_l = $, para todo $l$.

* **Incluye el infinito**: La expresión del potencial en general no diverge en el infinito, así que, salvo que la condición de contorno en el infinito lo imponga, $A_l = 0$, para todo $l$.

* **No incluye ni el origen ni el infinito**: Es una esfera con un hueco esférico, por lo que tiene dos condiciones de contorno, la interior y la exterior, con las que podemos obtener $A_l$ y $B_l$, para todo $l$
```

A partir del potencial, se puede calcular el campo eléctrico $\vec{E_d}$ y $\vec{E_f}$, dentro y fuera, y utilizando la discontinuidad en el campo eléctrico radial en $r=R$, podemos obtener la densidad superficial de carga $\sigma$ en la superficie de la esfera.

$$
\sigma = \varepsilon_0 (E_{f,r} - E_{d,r})
$$
