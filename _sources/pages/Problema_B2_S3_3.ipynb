# Problema B2.S3.3
 
Podemos suponer que la imagen de un **dipolo** $\vec{p}$ en un sistema con una esfera de **CEP (Conductor Eléctrico Perfecto)**, por similitud con la imagen de una carga puntual que conocemos, será algo parecido a otro dipolo *pero no podemos darlo por hecho*.

```{figure} ../images/B2_S3_3_1.png
:alt: q_puntual
:width: 350px
Imagen de un dipolo frente a una esfera de CEP.
```

```{note}
Tenemos que demostrarlo matemáticamente. Por lo tanto, tenemos que utilizar la relación matemática entre el dipolo, del que desconocemos la imagen, y la carga puntual, **de la que sí conocemos la imagen**.
```

Empezamos por lo que sí sabemos seguro: La imagen de una carga puntual $q$ a una distancia $d$ del centro de una esfera de CEP de radio $a$ ($d>a$) es otra carga $q' = -q\frac{a}{d}$ a una distancia $d'=\frac{a^2}{d}$.

```{figure} ../images/B2_S3_3_2.png
:alt: q_puntual
:width: 350px
Imagen de una carga puntual frente a una esfera de CEP.
```

Ahora descomponemos el dipolo en *dos cargas separadas una distancia $\Delta \ll d$*, tal que $q \Delta = p$:

```{figure} ../images/B2_S3_3_3.png
:alt: dipolo_eq
:width: 260px
Equivalencia entre dipolo y cargas puntuales.
```

Veamos cual sería la imagen de las dos cargas puntuales equivalentes al dipolo.

```{figure} ../images/B2_S3_3_4.png
:alt: dipolo_imagen
:width: 350px
Imagen de las cargas equivalentes al dipolo.
```

De forma que *podemos escribir los valores* de $q_1$, $q_2$, $d_1$ y $d_2$:

$$
q_1 &= -(-q)\frac{a}{d}\\
d_1 &= \frac{a^2}{d}\\
q_2 &= -q\frac{a}{d+\Delta}\\
d_2 &= \frac{a^2}{d+\Delta}
$$

Estas dos cargas parecen un dipolo, pero si nos fijamos en detalle, las dos cargas no son exactamente iguales, por lo que no es solamente un dipolo. Podemos utilizar **la configuración con estas dos cargas y realizar el análisis multipolar**, para comprobar **cuál es el valor del dipolo imagen**.

El término monopolar imagen, llamémoslo $Q'_T$, será la suma de las cargas:

$$
Q'_T = q_1+q_2 = q \frac{a}{d} - q \frac{a}{d+\Delta} = q \frac{a}{d} - q \frac{a}{d^2}(d-\Delta) = q\frac{a\Delta}{d^2} = p \frac{a}{d^2}
$$

Vemos que **el término monopolar no es despreciable**, sino que es del orden de $q$, $a$ y $d$. Respecto al término dipolar $\vec{p'}$ de la configuración de las imágenes utilizamos el valor de las cargas aproximado para $\Delta \ll d$, es decir, $q_1 = -q_2 = q\frac{a}{d}$ (*Vamos a poner el eje $z$ en la linea recta que une las cargas y el centro de la esfera*):

$$
\vec{p'} = q_1 \vec{r_1} + q_2 \vec{r_2} = q_1 d_1 \vec{u_z} + q_2 d_2 \vec{u_z} = \left( q \frac{a^3}{d^2} - q \frac{a^3}{d(d+\Delta)} \right) \vec{u_z} = \vec{p}\frac{a^3}{d^3}
$$

Por lo tanto, **el potencial en cualquier punto del espacio** fuera de la esfera de CEP, es la suma de 3 términos:
- El potencial $\Phi_p$ creado por el dipolo original $\vec{p}$ en la posición $d$.
- El potencial $\Phi_{p'}$ creado por el dipolo imagen $\vec{p}' = \vec{p} \frac{a^3}{d^3}$ en la posición $d'=\frac{a^2}{d}$.
- El potencial $\Phi_{Q'_T}$ creado por la carga imagen $Q'_T = q_1+q_2$ en la posición $d'=\frac{a^2}{d}$.

$$
\Phi(r{\ge}a) = \Phi_p + \Phi_{p'} + \Phi_{Q'_T}
$$


```{figure} ../images/B2_S3_3_5.png
:alt: dipolo_imagen
:width: 350px
Imagen del dipolo.
```

Podemos verificar que, efectivamente, el potencial $\Phi(r{=}a)$ es nulo.

$$
\Phi(r{=}a) &= \Phi_p(r{=}a) + \Phi_{p'}(r{=}a) + \Phi_{Q'_T}(r{=}a) = \\
&= \frac{p\cos\alpha}{4\pi\varepsilon_0\rho^2} + \frac{p'\cos\alpha'}{4\pi\varepsilon_0{\rho'}^2} + \frac{Q'_T}{4\pi\varepsilon_0\rho'}
$$

```{figure} ../images/B2_S3_3_6.png
:alt: dipolo_imagen
:width: 350px
Imagen del dipolo.
```

Con:

$$
\cos\alpha &= -\cos\Omega = -\frac{d-a\cos\theta}{\rho} \\
\cos\alpha' &= \frac{a\cos\theta-\frac{a^2}{d}}{\rho'} \\
\rho^2 &= d^2 + a^2 -2da\cos\theta \\
{\rho'}^2 &= a^2 + (\frac{a^2}{d})^2 -2\frac{a^3}{d}\cos\theta
$$

Sustituyendo, podemos verificar $\Phi(r{=}a) = 0$.

```{note}
Uno se puede preguntar por qué en el termino dipolar de la imagen usamos $q_2= - q\frac{a}{d}$ en lugar de $q_2 = -q\frac{a}{d+\Delta}$:

* Por una parte, veamos que **el desarrollo multipolar se realiza sobre variaciones de $\frac{1}{|\vec{r} - \vec{r'}|}$**, mientras que las cargas se consideran dadas. Por esto, tiene sentido que la carga se aproxime a su valor final antes de realizar la integral/sumatoria del momento dipolar.

* Por otra parte, hemos comprobado que usando $q_2= - q\frac{a}{d}$ **obtenemos el potencial correcto en la superficie del CEP**.
```
