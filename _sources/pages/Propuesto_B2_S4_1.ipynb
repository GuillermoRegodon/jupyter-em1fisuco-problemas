# Problema Propuesto B2.S4.1

Este problema se resuelve usando una metodología similar a la teoría que desarrollamos para obtener la imagen de una carga puntual. Se trata de la imagen de un hilo infinito de densidad lineal de carga $\lambda$ cuando se pone paralelo a un cilindro infinito de conductor eléctrico perfecto (*CEP*) conectado a un potencial $\Phi_{CEP} = V$. Igual que en el caso esférico, vamos a suponer que el CEP está conectado a tierra, y $\Phi_{CEP} = 0$, y a posteriori añadiremos un hilo de carga en el eje del CEP para tener en cuenta la posible diferencia.

```{figure} ../images/B2_S4_P1_1.png
:alt: cilindro_hilo
:width: 300px
```

La metodología resumida es:

* Primero, planteamos una imagen para el hilo, que posiblemente será otro hilo (*y de hecho es así*).

* Buscamos la posición $d`$ y la densidad lineal de carga $\lambda'$ de la imagen para que el potencial sea 0, o lo que es lo mismo, para que el campo eléctrico sea perpendicular a la superficie del CEP.

* Escribimos las magnitudes físicas potencial $\Phi$ y campo eléctrico $\vec{E}$ fuera del CEP como la superposición de las magnitudes físicas generadas por la el hilo de carga original y por su imagen.

Este problema tiene algunas peculiaridades derivadas del hecho de que la configuración de carga se extiende hasta el infinito. Veamos antes el campo eléctrico y el potencial que generan un hilo de carga infinito.

````{note}
Un hilo infinito de densidad lineal de carga $\lambda$ genera un campo eléctrico $\vec{E}$ radial simetría cilíndrica. Lo resolvemos por Gauss. Para una superficie de Gauss que sea un cilindro de altura $H$ y radio $s$:

```{figure} ../images/B2_S4_P1_2.png
:alt: cilindro_hilo
:width: 200px
```

$$
E2\pi sH = \frac{\lambda H}{\varepsilon_0} \quad \Rightarrow \quad \vec{E} = \frac{\lambda}{2\pi s} \vec{u_s}
$$

El potencial $\Phi$ se puede obtener por integración, pero nos encontramos con el problema de que la referencia del potencial no se puede poner en el infinito, porque la expresión diverge:

$$
\Phi = \frac{\lambda}{2\pi\varepsilon_0}\log{s} + C
$$
````

Para resolver el problema, podemos intentar plantear la condición sobre el potencial. Sin embargo, dado que la referencia del potencial $\Phi$ no está bien definida, al ser un hilo infinito, nos da problemas. Sin embargo, puedo imponer la condición sobre el campo eléctrico $\vec{E}$. Para obtener el campo eléctrico de un hilo desplazado, tenemos que modificar el vector unitario y la coordenada radial:

$$
\vec{E} = \frac{\lambda}{2\pi s} \vec{u_s} = \frac{\lambda}{2\pi s^2} \vec{s} \quad \Rightarrow \quad \vec{E}_{\lambda,R} = \frac{\lambda}{2\pi (s\vec{u_s} - R\vec{u_x})^2} (s\vec{u_s} - R\vec{u_x})
$$

De la misma forma, para el hilo $\lambda'$ en $d'$:

$$
\vec{E}_{\lambda',d'} = \frac{\lambda'}{2\pi (s\vec{u_s} - d'\vec{u_x})^2} (s\vec{u_s} - d'\vec{u_x})
$$

La suma de los dos campos eléctricos $\vec{E}_{\lambda,R}$ y $\vec{E}_{\lambda',d'}$ da un vector con componente radial y componente $x$. La condición de CEP nos impone que el campo eléctrico es perpendicular en la superficie del CEP, por lo que si sustituyo $s$ por $b$, el radio del CEP, la componente $x$ debe desaparecer:

$$
\frac{-R\lambda}{2\pi (b\vec{u_s} - R\vec{u_x})^2} + \frac{-d'\lambda'}{2\pi (b\vec{u_s} - d'\vec{u_x})^2} = 0
$$

Podemos intentar obtener los elementos $(s\vec{u_s} - R\vec{u_x})^2$ y $(s\vec{u_s} - d'\vec{u_x})^2$ aplicando trigonometría, pero podemos hacerlo directamente realizando el producto escalar. Así vemos:

$$
(b\vec{u_s} - R\vec{u_x})^2 = b^2 + R^2 -2bR\vec{u_s}\cdot\vec{u_x} \\
(b\vec{u_s} - d'\vec{u_x})^2 = b^2 + d'^2 -2bd'\vec{u_s}\cdot\vec{u_x} \\
$$

Vemos que una solución posible es que si $d' = R$ y $lambda' = -lambda$, el campo eléctrico en dirección $x$, se anula, pero esta solución es un hilo que cancela al otro, por lo que es una solución trivial que no nos interesa. Pensemos en la otra solución posible: Si $d'$ fuera proporcional a $b$ y $b$ a su vez fuera proporcional a $R$ en el mismo factor $k$:

$$
\frac{d'}{b} = \frac{b}{R} = k
$$

Entonces, los denominadores son proporcionales:

$$
b^2 + d'^2 -2bd'\vec{u_s}\cdot\vec{u_x} = k^2(b^2 + R^2 -2bR\vec{u_s}\cdot\vec{u_x})
$$

Mientras que en los numeradores tenemos que $d'=kb = k^2R$, por lo que, probando esta condición, vemos que la condición de que la componente $x$ del campo eléctrico $\vec{E}$ se anule, se cumple si además $\lambda' = -\lambda$. Además, de la relación de proporcionalidad $d'=\frac{b^2}{R}$. Aunque esta solución se ha obtenido por inspección, es la única (*junto con la trivial*) para la que el campo eléctrico es perpendicular a la superficie del CEP. El resto de los apartados:

* El potencial total $\Phi$ se puede obtener como la superposición del potencial de los dos hilos, real e imagen, $\Phi_{\lambda,R}$ y $\Phi_{\lambda',d'}$. Observemos que tenemos una constante arbitraria relacionada con la imposibilidad de establecer una referencia de potencial para una configuración cilíndrica infinita. En este sentido, da igual que el hilo tenga un potencial u otro, o que consideremos un hilo central para tener en cuenta el potencial $V$, simplemente, la constante arbitraria es inevitable.

* La densidad superficial de carga inducida se puede obtener a partir de la condición de contorno para el campo eléctrico en la superficie del CEP, en la que tiene una discontinuidad.

* Finalmente, la fuerza por unidad de longitud es la fuerza que sufre un diferencial de hilo:

$$
f = \lambda E_{\lambda'\rightarrow\lambda} \quad \left( {\rm \frac{N}{m}} \right)
$$
