# Problema B1.S2.1

Este problema tiene tres apartados independientes. El primero es un cálculo que no es de dipolos, que hay que resolver con expresiones integrales del campo eléctrico, al no tener ninguna simetría conveniente. El segundo y el tercero son dos aplicaciones de dipolos distintas. Vamos a comprobar cada apartado por separado.

## Apartado $a)$

Se pide el potencial $\Phi$ y el campo eléctrico $\vec{E}$ de una configuración lineal de carga $\lambda = \lambda_0 \cos \phi'$ semicircular en el plano $xy$ en los puntos del eje z. Dado que no tenemos ninguna simetría conveniente para realizar la integral, no tenemos más alternativa que calcular la integral de linea:

$$
\vec{E} = \frac{1}{4\pi\varepsilon_0}\int\frac{\lambda(\vec{r'}) dl'}{R^2} \vec{u_R}
$$

Con $\vec{R} = \vec{r}-\vec{r'}$. Como no nos indica la orientación exacta del semicírculo, la podemos elegir nosotros. Elegimos por tanto que el semicírculo esté en el semiplano $y>0$.

```{figure} ../images/B1_S2_1_1.png
:alt: semicirculo de carga.png
:width: 350px
```

Aunque la simetría no nos permite calcular la integral, si que **podemos utilizar la simetría de reflexión sobre el plano $yz$ para observar que algunas componentes deben ser nulas en el eje $z$** (*Pensar cuáles durante unos instantes*).

Vemos que por cada trozo de carga positiva en el 1º cuadrante (azul en la figura) que genera un campo eléctrico $\vec{E}$ en la dirección $z$ positiva y en la dirección $y$ negativa, tenemos un trozo de carga negativa en el segundo cuadrante (rojo en la figura) que genera un campo eléctrico de igual magnitud pero en la dirección de $z$ negativa y en la de $y$ positiva. Por lo tanto, el campo eléctrico total, al ser la superposición de ambos, debe cancelarse en las direcciones $z$ e $y$. La única componente que sobrevive es la $x$, ya que todos los trozos de carga generan campo en la dirección $x$ positiva.

```{figure} ../images/B1_S2_1_2.png
:alt: semicirculo de carga.png
:width: 350px
```

Tenemos que interpretar todos los elementos que aparecen en la integral de línea.

$$
\lambda(\vec{r'}) &= \lambda_0 \cos \phi' \\
\vec{r} &= (0,0,z) \\
\vec{r'} &= (a\cos\phi',a\sin\phi',0) \\
\vec{R} &= (-a\cos\phi',-a\sin\phi',z) \\
R &= \sqrt{a^2 + z^2} \\
\vec{u_R} &= (-\frac{a\cos\phi'}{\sqrt{a^2 + z^2}},-\frac{a\sin\phi'}{\sqrt{a^2 + z^2}},\frac{z}{\sqrt{a^2 + z^2}}) \\
dl' &= a d\phi' \quad \quad \phi' \in [0,\pi]
$$

Como hemos argumentado, nos quedamos tan sólo con la componente $x$ del campo eléctrico, y obtenemos la siguiente expresión:

$$
E_x = \frac{1}{4\pi\varepsilon_0} \int_{\phi'=0}^{\phi'=\pi} \frac{\lambda_0 \cos \phi' a d\phi'}{(a^2 + z^2)^{\frac{3}{2}}} \left(-a\cos\phi'\right) = -\frac{\lambda_0 a^2}{8 \varepsilon_0 (a^2 + z^2)^{\frac{3}{2}}}
$$

En este punto, es fácil plantear las integrales para las componentes $z$ e $y$ y verificar que son cero (*¡En el eje $z$!*). Del mismo modo, si planteamos la integral del potencial $\Phi$, también se anula (*¡También en el eje $z$!*).

$$
\Phi = \frac{1}{4\pi\varepsilon_0}\int\frac{\lambda(\vec{r'}) dl'}{R} = \frac{1}{4\pi\varepsilon_0} \int_{\phi'=0}^{\phi'=\pi} \frac{\lambda_0 \cos \phi' a d\phi'}{\sqrt{a^2 + z^2}} = 0
$$

```{note}
Observar que las simetrías se pueden aprovechar sólo en el eje $z$ (*más bien en el plano $x=0$*). En muchos problemas, calcular el campo eléctrico o el potencial en cualquier punto del espacio es impracticable o directamente imposible. Por eso, en muchos casos, se pide el campo eléctrico o el potencial en algún punto en el que se sabe que se puede calcular (*En este caso, en los puntos en el eje $z$*).
```


## Apartado $b)$

En este apartado se pide comprobar el efecto mecánico sobre un dipolo. Observamos que el dipolo es muy pequeño en comparación con la configuración de carga, y gracias a esto, en lugar de considerar las dos cargas $q_0$ de distinto signo que conforman el dipolo separadas una cierta distancia $d_0$, podemos usar las expresiones simplificadas:

$$
\vec{F} &= (\vec{p_0} \cdot \nabla) \vec{E} \\
\vec{M} &= \vec{p_0} \times \vec{E}
$$

```{figure} ../images/B1_S2_1_3.png
:alt: semicirculo de carga.png
:width: 350px
```

Es decir, la fuerza $\vec{F}$ y el momento de la fuerza $\vec{M}$ sobre las dos cargas del dipolo junto con la ligadura que las mantiene unidas se puede obtener utilizando esta expresión, debido a que el tamaño del dipolo (*la separación de las cargas $d_0$*) es muy pequeño en comparación con las distancias características en las que varía el campo eléctrico. Dado que la configuración de carga tiene una escala característica del orden del radio $a$, la separación entre las cargas $d_0$ debe $d_0 \ll a$.

* Fuerza $\vec{F}$

La expresión general incluye el operador $\vec{p_0} \cdot \nabla = p_{0x} \frac{\partial}{\partial x} + p_{0y} \frac{\partial}{\partial y} + p_{0z} \frac{\partial}{\partial z}$  actuando sobre las tres componentes del campo eléctrico $\vec{E}$. La única componente del campo eléctrico $\vec{E}$ es la componente $x$, mientras que la única componente del dipolo $\vec{p_0}$ es la componente $z$:

$$
F_x = p_0 \frac{\partial E_x}{\partial z} = \frac{p_0 \lambda_0 a^2}{8 \varepsilon_0} \frac{3 z}{(a^2 + z^2)^{\frac{5}{2}}}
$$

```{note}
La carga negativa $-q_0$ del dipolo está sometida a un campo eléctrico $E_x$ un poco más intenso en el sentido de $x$ negativo que la carga positiva $+q_0$, por lo que la fuerza neta sobre el dipolo es hacia el sentido de $x$ positivo.
```

```{note}
Nos preguntamos en este punto, **¿qué pasaría si el dipolo estuviera orientado en la dirección $x$ o en la dirección $y$?** (*¡Pensar durante unos instantes!*)

Es importante observar una cosa. Dado que el campo eléctrico $\vec{E}$ solo se conoce en el eje $z$, *no se puede calcular la derivada parcial del campo eléctrico $\vec{E}$ en las direcciones $x$ e $y$*. Es un error común, dado que la expresión no tiene $x$ o $y$, calcular que la derivada parcial con respecto a ambas variables es cero, pero lo cierto es que solo conocemos el campo eléctrico en el eje $z$, por lo que no sabemos como depende de las otras coordenadas $x$ e $y$. Lo que si sabemos es que, *si fuera posible obtener una expresión de la variación en $x$ e $y$, al sustituir en $x=y=0$, obtendríamos la expresión calculada en el apartado $a)$*.

Es una suerte (*intencionadamente*) que el problema pida que el dipolo $\vec{p_0}$ este orientado en la dirección $z$.
```

* Momento de la fuerza $\vec{M}$

La expresión del momento de la fuerza es tan solo un producto vectorial.

$$
\vec{M} = \vec{p_0} \times \vec{E} = p_0 E_x \vec{u_z} \times \vec{u_x} = p_0 E_x \vec{u_y}
$$


## Apartado $c)$

Este apartado pregunta por el dipolo equivalente de la configuración a gran distancia del medio anillo. Observamos que a grandes distancias, la configuración es muy pequeña con respecto a la escala característica $a$ de la configuración de carga, es decir, $r \gg a$. Desde estas distancias, la configuración se percibe como un dipolo $\vec{p}$ que se puede calcular utilizando la expresión apropiada, aplicada a una configuración lineal de carga:

$$
\vec{p} = \int \vec{r'} \lambda(\vec{r'}) dl'
$$

Realizando la sustitución de los elementos de la integral obtenidos en el apartado $a)$.

$$
\vec{p} = \int_{\phi'=0}^{\phi'=\pi} (a\cos\phi',a\sin\phi',0) \lambda_0 \cos \phi' a d\phi'
$$

Esta integral es en realidad tres integrales, una para cada componente.

* La componente $x$, es una integral $\cos^2\phi'$, ya resuelta anteriormente

$$
p_x = \int_{\phi'=0}^{\phi'=\pi} \lambda_0 a^2 \cos^2\phi' d\phi' = \lambda_0 a^2 \frac{\pi}{2}
$$

* La componente $y$, es una integral que se anula (*¡Comprobar!*).

* La componente $z$ es directamente 0.

Por lo tanto, la configuración, **a grandes distancias**, es equivalente a un dipolo orientado en la dirección x.

```{note}
Observar como los apartados $b)$ y $c)$ utilizan el concepto de dipolo pero en escalas diferentes. Sin embargo, en ambos casos el requisito para poder hablar de dipolo es que el tamaño del dipolo sea muy pequeño en comparación con las distancias en las que se desarrolla el campo eléctrico $\vec{E}$.
```

Queda comprobar que la configuración en los puntos en los que conocemos el campo eléctrico $\vec{E}$, es decir en el eje $z$, tiene la misma expresión que la del dipolo equivalente $\vec{p}$. La expresión general es (*¿por qué no podemos usar la expresión en coordenadas esféricas? ¿Podríamos usarla de alguna manera?*):

$$
\vec{E} = \frac{1}{4\pi\varepsilon_0}\left[ \frac{(3\vec{p}\cdot\vec{r})\vec{r}}{r^5} - \frac{\vec{p}}{r^3}\right]
$$

Sustituyendo los elementos obtenidos en el apartado $a)$ y en este apartado, $\vec{p}\cdot\vec{r} = 0$, por lo que la expresión es:

$$
E_x = -\frac{\lambda_0 a^2}{8\varepsilon_0 z^3}
$$

Comprobamos que es el mismo resultado que si usamos la expresión exacta, valida en el eje $z$, y obtenemos la expresión cuando $r$ es muy grande, $r \uparrow$ (*o lo que es lo mismo en el eje $z$, la expresión cuando $z \gg a$*). Así mismo, si calculamos el potencial en el eje $z$:

$$
\Phi = -\frac{1}{4\pi\varepsilon_0} \frac{\vec{u_r}\cdot\vec{p}}{r^2} = 0
$$
