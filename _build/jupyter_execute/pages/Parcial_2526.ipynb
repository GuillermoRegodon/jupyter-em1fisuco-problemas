```{note}
### Ejercicio 1
1.- Una esfera de radio $a$ centrada en el origen tiene una polarización permanente $\vec{P}$:

$$
\vec{P} = (br^2 + c\cos\theta)\vec{u_r} -c\sin\theta\vec{u_\theta}
$$

* A) Calcule las densidades de carga de polarización de la esfera (2 puntos)
* B) Calcule el potencial $\Phi(\vec{r})$ en todo el espacio exterior e interior a la esfera (3 puntos)
```

La esfera tiene una polarización permanente, es decir, que **la polarización no depende del campo eléctrico**, por lo que el concepto de permitividad eléctrica $\varepsilon$ o vector desplazamiento $\vec{D}$ no tienen aplicación aquí. Dado que la polarización ya viene dada, será equivalente a unas densidades de carga de polarización que serán a su vez las fuentes del campo eléctrico, y a partir de las cuales podremos calcular el potencial.

#### A) Calcule las densidades de carga de polarización de la esfera (2 puntos)
Las densidades de carga de polarización se pueden obtener a partir de su definición. Empezamos por la densidad de carga superficial de polarización:

$$
\sigma_P = \vec{P}\cdot\vec{n}
$$

Necesitamos el vector unitario normal a la superficie. Como se trata de una esfera centrada, $\vec{n} = \vec{u_r}$, es decir, la densidad superficial de carga de polarización es directamente la componente radial $P_r$ de la polarización. Si queremos verlo por pasos detalladamente:

$$
\sigma_P = \vec{P}\cdot\vec{n} = \left((br^2 + c\cos\theta)\vec{u_r} -c\sin\theta\vec{u_\theta}\right)\cdot\vec{u_r} = (br^2 + c\cos\theta)\vec{u_r}\cdot\vec{u_r} -c\sin\theta\vec{u_\theta}\cdot\vec{u_r} = br^2 + c\cos\theta
$$

Por último, **debemos observar que $r$ es una coordenada que en la superficie de la esfera vale $r=a$** (*la coordenada $\theta$ si varia en la superficie de la esfera*). Por esto:

$$
\sigma_P = ba^2 + c\cos\theta
$$

En segundo lugar, calculamos la densidad volumétrica de carga de polarización:

$$
\rho_P = -\nabla\cdot\vec{P}
$$

Debemos poner la divergencia en esféricas, que son las coordenadas en las que se ha definido la polarización $\vec{P}$.

$$
\rho_P = -\left( \frac{1}{r^2}\frac{\partial}{\partial r} r^2 P_r + \frac{1}{r\sin\theta}\frac{\partial}{\partial\theta} \sin\theta P_\theta \right)
$$

Por una parte, he ignorado el sumando que contiene la componente $\phi$ porque no aparece en la definición.

```{note}
La definición que aparece en el formulario es:

$$
\rho_P(\vec{r`}) = -\nabla'\cdot\vec{P}
$$

No debemos confundirnos: el concepto que tenemos que tener claro es que la densidad volumétrica de carga de polarización aparece en la región en la que hay una polarización. Dicha densidad volumétrica de carga de polarización se convierte en una fuente, y a las fuentes, en el formulario, las denotamos con la prima. Es decir, estamos hablando de $\rho_P(\vec{r'})$

En este enunciado, se ha usado $\vec{r}$, por lo no nos debe extrañar que haya quitado la prima en la solución.
```

Observamos que $P_r = br^2 + c\cos\theta$ y. que $P_\theta = -c\sin\theta$:

$$
\rho_P = -\left( \frac{1}{r^2}\frac{\partial}{\partial r} (br^4 + cr^2\cos\theta) + \frac{1}{r\sin\theta}\frac{\partial}{\partial\theta} (-cr\sin^2\theta) \right)
$$

Nótese que los productos $r^2P_r$ y $\sin\theta P_\theta$ deben hacerse **antes de derivar, es decir, influyen en la derivada**. De esta forma:

$$
\rho_P = -\left( \frac{1}{r^2}(4br^3 + 2cr\cos\theta) + \frac{1}{r\sin\theta} (-2c\sin\theta\cos\theta) \right) = -4br
$$

#### B) Calcule el potencial $\Phi(\vec{r})$ en todo el espacio exterior e interior a la esfera (3 puntos)

Las densidades de carga de polarización son las fuentes de campo electrostático y del potencial. Como tenemos dos tipos de fuentes, podemos calcular el potencial de **cada una de las dos por separado y aplicar superposición**:

$$
\Phi_T = \Phi_\rho + \Phi_\sigma
$$

```{note}
Notemos que **ambos tipos de carga generan potencial en todo el espacio**. En un ejemplo concreto, el de un Conductor Eléctrico Perfecto, se impone que el campo eléctrico $\vec{E}$ es nulo y el potencial $\Phi$ es constante o nulo en el interior, pero eso no ocurre en general (*salvo coincidencia*), en ninguna otra configuración. No ocurre en esferas de carga, ni en materiales polarizados ni en materiales con permitividad.
```

Por lo tanto, vamos a calcular por separado:

* **El potencial $\Phi_\rho$, está generado por una densidad volumétrica de carga con simetría esférica, $\rho_P = -4br$**. Debido a la simetría esférica podemos aplicar el método de Gauss para calcular el campo eléctrico y después integrar en la coordenada radial para obtener el potencial $\Phi_\rho$. Alternativamente, podríamos usar la ecuación de Poisson con simetría esférica, ($\nabla^2 = \frac{1}{r^2}\frac{\partial}{\partial r} r^2 \frac{\partial}{\partial r}$), distinguiendo la ecuación dentro y fuera (*dentro $\rho_P = -4br$; fuera $\rho_P = 0$*), siendo el resultado el mismo. Usemos Gauss.

**Primero, el campo eléctrico dentro**:

$$
\oint_S\vec{E}\cdot\vec{dS} = \frac{1}{\varepsilon_0} \int_V \rho_P dV
$$

Elegimos una esfera de radio $r<=a$ arbitrario. La expresión de la izquierda queda simplemente:

$$
\oint_S\vec{E}\cdot\vec{dS} = 4\pi r^2E_{d,r}
$$

Para calcular la expresión de la derecha, vamos a usar la coordenada $r$ arbitraria, que va a ser un límite de integración. Por lo tanto, para no confundir, voy a usar $dV'$ en lugar de $dV$, y $\rho_P(\vec{r'}) = -4br'$

$$
\int_{V'} \rho_P(\vec{r'}) dV' = \int_{r'{=}0}^{r'{=}r}\int_{\theta'{=}0}^{\theta'{=}\pi}\int_{\varphi'{=}0}^{\varphi'{=}2\pi} (-4br')r'^2\sin\theta'dr'd\theta'd\varphi' = -4\pi b r^4
$$

Por lo que el campo eléctrico radial dentro:

$$
4\pi r^2E_{d,r} = -4\pi b r^4 \quad \Rightarrow \quad E_{d,r} = -\frac{br^2}{\varepsilon_0}
$$

**Ahora, el campo eléctrico fuera**, de nuevo por Gauss:

$$
\oint_S\vec{E}\cdot\vec{dS} = \frac{1}{\varepsilon_0} \int_V \rho_P dV
$$

Elegimos una esfera de radio $r>a$ arbitrario. La expresión de la izquierda queda igual que antes:

$$
\oint_S\vec{E}\cdot\vec{dS} = 4\pi r^2E_{f,r}
$$

La expresión de la derecha incluirá la esfera de carga entera, por lo que el límite de integración es hasta $r'=a$

$$
\int_{V'} \rho_P(\vec{r'}) dV' = \int_{r'{=}0}^{r'{=}a}\int_{\theta'{=}0}^{\theta'{=}\pi}\int_{\varphi'{=}0}^{\varphi'{=}2\pi} (-4br')r'^2\sin\theta'dr'd\theta'd\varphi' = -4\pi b a^4
$$

Por lo que el campo eléctrico radial fuera:

$$
E_{f,r} = -\frac{ba^4}{\varepsilon_0r^2}
$$

**El potencial fuera se puede obtener por integración**, utilizando como referencia el potencial en el infinito, $\Phi_\rho(r{\rightarrow}\infty) = 0$.

$$
\Phi_\rho = -\int_C\vec{E}\cdot\vec{l}
$$

Usamos como camino un radio desde el infinito. De nuevo, ya que vamos a usar la coordenada $r$ para el punto final de la integral, integramos en $r'$:

$$
\Phi_\rho = -\int_C\vec{E}(\vec{r'})\cdot\vec{l'} = -\int_{r'{=}\infty}^{r'{=}r}E_r(r') dr'
$$

Para el potencial fuera, solo utilizamos la expresión del campo fuera, ya que toda la integral ocurre en el exterior de la esfera.

$$
\Phi_{\rho,f} = \int_{r'{=}\infty}^{r'{=}r} \frac{ba^4}{\varepsilon_0r'^2} dr' = - \frac{ba^4}{\varepsilon_0r'}|_{r'{=}\infty}^{r'{=}r} = - \frac{ba^4}{\varepsilon_0r}
$$

Por supuesto, el potencial justo en el borde de la esfera:

$$
\Phi_{\rho,f}(r{=}a) = - \frac{ba^3}{\varepsilon_0}
$$

**El potencial dentro también se puede obtener por integración**, utilizando como referencia el potencial en el infinito, $\Phi_P(r{\rightarrow}\infty) = 0$.

$$
\Phi_\rho = -\int_C\vec{E}\cdot\vec{l}
$$

En este caso, hay que distinguir el tramo que ocurre fuera del que ocurre dentro:

$$
\Phi_{\rho,d} = -\int_C\vec{E}(\vec{r'})\cdot\vec{l'} = -\int_{r'{=}\infty}^{r'{=}r}E_r(r') dr' = -\int_{r'{=}\infty}^{r'{=}a}E_r(r') dr' -\int_{r'{=}a}^{r'{=}r}E_r(r') dr' = - \frac{ba^3}{\varepsilon_0} + \int_{r'{=}a}^{r'{=}r}\frac{br'^2}{\varepsilon_0} dr'
$$

Por lo tanto, finalmente el potencial dentro:

$$
\Phi_{\rho,d} = - \frac{ba^3}{\varepsilon_0} + \frac{br'^3}{\varepsilon_0}|_{r'{=}a}^{r'{=}r} = - \frac{ba^3}{\varepsilon_0} + \frac{b(r^3-a^3)}{3\varepsilon_0}
$$

* **El potencial $\Phi_\sigma$ generado por la densidad superficial de carga de polarización $\sigma_P = ba^2 + c\cos\theta$** tiene una condición de contorno en una esfera que tiene simetría en la coordenada $\varphi$, y además nos piden el potencial en todo el espacio. Es un problema para el que se puede usar el método de separación de variables y la expansión en polinomios de Legendre. Por inspección, vemos que la condición de contorno está en función de una constante $ba^2$ y un término multiplicado por $\cos\theta$. Por lo tanto solo vamos a tener los dos primeros términos de la expansión en polinomios de Legendre. Además, el potencial en el infinito tiende a 0, $\Phi_\sigma(r{\rightarrow}\infty) = 0$, y dentro de la esfera el potencial no puede divergir, por lo que podemos utilizar que los términos $A_l$ se anulan fuera, y los términos $B_l$ se anulan dentro.

$$
\Phi_{\sigma,f} &= \frac{B_0}{r} + \frac{B_1}{r^2}\cos\theta \\
\Phi_{\sigma,d} &= A_0 + A_1r\cos\theta
$$

La condición de continuidad del potencial en $r=a$:

$$
\frac{B_0}{a} + \frac{B_1}{a^2}\cos\theta = A_0 + A_1r\cos\theta
$$

O lo que es lo mismo, igualando los términos constantes y los términos en $\cos\theta$:

$$
A_0 &= \frac{B_0}{a} \\
A_1 a &= \frac{B_1}{a^2}
$$

La otra condición es la del campo eléctrico radial, $E_r = -\frac{\partial\Phi}{\partial r}$. Primero calculamos el campo eléctrico radial dentro y fuera:

$$
E_{f,r} &= \frac{B_0}{r^2} + 2\frac{B_1}{r^3}\cos\theta \\
E_{d,r} &= -A_1\cos\theta
$$

La condición de contorno del campo eléctrico en $r=a$ la da la densidad superficial de carga:

$$
\varepsilon_0(E_{f,r} - E_{d,r}) = \sigma_P
$$

Particularizando:

$$
\varepsilon_0( \frac{B_0}{a^2} + 2\frac{B_1}{a^3}\cos\theta + A_1\cos\theta ) = ba^2 + c\cos\theta
$$

Ya tenemos expresiones suficientes para despejar todas las constantes:

$$
\varepsilon_0\frac{B_0}{a^2} = ba^2 \\
\varepsilon_0(2\frac{B_1}{a^3} + A_1) = 3\varepsilon_0\frac{B_1}{a^3} = c
$$

Por lo tanto:

$$
B_0 = \frac{ba^4}{\varepsilon_0}; \quad
B_1 = \frac{ca^3}{3\varepsilon_0}; \quad
A_1 = \frac{c}{3\varepsilon_0}; \quad
A_0 = \frac{ba^3}{\varepsilon_0}
$$

Y finalmente:

$$
\Phi_{\sigma,f} &= \frac{ba^4}{\varepsilon_0r} + \frac{ca^3}{3\varepsilon_0r^2}\cos\theta \\
\Phi_{\sigma,d} &= \frac{ba^3}{\varepsilon_0} + \frac{cr}{3\varepsilon_0}\cos\theta
$$

* **El potencial total es la suma de los dos**

$$
\Phi_{T,f} &= \Phi_{\rho,f} + \Phi_{\sigma,f} = \frac{ca^3}{3\varepsilon_0r^2}\cos\theta \\
\Phi_{T,d} &= \Phi_{\rho,d} + \Phi_{\sigma,d} = \frac{b(r^3-a^3)}{3\varepsilon_0} + \frac{cr}{3\varepsilon_0}\cos\theta
$$

````{note}
### Ejercicio 2
2.- Un hilo de densidad de carga lineal $\lambda$ constante de longitud $b-a$ se coloca perpendicularmente a una distancia $a$ con respecto a un plano de Conductor Eléctrico Perfecto (CEP) conectado a potencial nulo.

* A) Calcule la densidad de carga superficial inducida en el CEP (3 puntos).
* B) Integre la densidad de carga superficial obtenida anteriormente e interprete el resultado (2 puntos).

```{figure} ../images/e2a2526_1.png
:alt: e2a252_1.png
:width: 320px
```

````

La imagen debe ser un hilo de densidad de carga lineal $\lambda' = -\lambda$ de longitud $b-a$ a una distancia $a$ dentro del CEP, para que todos los trozos de carga lineal tengan un trozo de carga lineal opuesta a la distancia apropiada para garantizar que el potencial en el conductor eléctrico perfecto sea 0. Visto detalladamente, si la imagen de una carga puntual $q$ a distancia $l$ es conocida:

$$
q' &= -q \\
l' &= l
$$

Cada trozo diferencial de hilo $dq = \lambda dl$ va a tener una imagen $dq' = \lambda' dl'$, con $dq' = -dq$. El tamaño es el mismo, $dl' = dl$, por lo que $\lambda' = -\lambda$. Por supuesto, el trozo más cercano a distancia $a$ tiene su imagen a la misma distancia, y lo mismo para el más alejado, a distancia $b$.

#### A) Calcule la densidad de carga superficial inducida en el CEP (3 puntos).

Para calcular la densidad superficial de carga inducida, debemos utilizar inevitablemente la condición de contorno del campo eléctrico normal, sabiendo que el campo eléctrico en el interior es 0.

$$
\sigma = \varepsilon_0 E_n
$$

El campo eléctrico será la contribución de los dos hilos, el real y el imagen.

$$
\vec{E} = \vec{E_\lambda} + \vec{E_{-\lambda}}
$$

Calculemos el primero. Ya que no tiene ninguna simetría obvia, usamos integración directa.

$$
\vec{E_\lambda} = \frac{1}{4\pi\varepsilon_0} \int_C \frac{\lambda dl' (\vec{r} - \vec{r'})}{|\vec{r} - \vec{r'}|^3}
$$

```{note}
Se puede integrar para obtener el potencial y después hacer el gradiente, pero para hacer el gradiente necesitamos el potencial, no solo en un punto, sino en los puntos vecinos. Dicho de otro modo, no podemos hacer la integral en el plano, sino en cualquier punto del espacio. Sin embargo, si utilizamos directamente la expresión integral del campo eléctrico, la integral es parecida y no es necesario realizar ninguna operación extra.

Tenemos una tendecia a usar la expresión $\sigma = -\varepsilon_0\frac{\partial \Phi}{\partial \vec{n}}$ para la densidad superficial de carga inducida en un CEP, pero casi nunca es la expresión más sencilla para resolver una densidad superficial de carga inducida. La expresión $\sigma = \varepsilon_0 \vec{E}\cdot\vec{n}$ es más práctica.
```

Vamos a definir los elementos de la integral. El hilo lo ponemos por ejemplo en la coordenada $x$, mientras que la coordenada de la fuente es $\vec{r'}$. Tenemos que elegir el origen de coordenadas. Lo ponemos por comodidad en el CEP en el punto más cercano al hilo, con lo que $\vec{r'} = (x',0,0)$. El punto en el que queremos conocer el campo eléctrico es un punto arbitrario del CEP, y su coordenada es $\vec{r}$, por lo que un punto arbitrario del plano es $\vec{r} = (0,y,z)$. El diferencial de linea sería $dl' = dx'$. En conjunto:

$$
\vec{r} &= (0,y,z) \\
\vec{r'} &= (x',0,0) \\
\vec{r}-\vec{r'} &= (-x', y, z) \\
\vec{r}-\vec{r'} &= \sqrt{x'^2 + y^2 + z^2} \\
dl' &= dx' \quad x' \in [a,b]
$$

La integral se puede escribir:

$$
\vec{E_\lambda} = \frac{1}{4\pi\varepsilon_0} \int_{x'{=}a}^{x'{=}b} \frac{ \lambda dx' (-x', y, z)}{(x'^2 + y^2 + z^2)^{3/2}}
$$

Como solo nos interesa la coordenada $x$ (*Las otras se van a cancelar igualmente*):

$$
E_{\lambda,x} = \frac{1}{4\pi\varepsilon_0} \int_{x'{=}a}^{x'{=}b} \frac{ -\lambda x' dx'}{(x'^2 + y^2 + z^2)^{3/2}} = \frac{1}{4\pi\varepsilon_0} \frac{\lambda}{(x'^2 + y^2 + z^2)^{1/2}} |_{x'{=}a}^{x'{=}b} = \frac{\lambda}{4\pi\varepsilon_0} \left(\frac{1}{(b^2 + y^2 + z^2)^{1/2}} - \frac{1}{(a^2 + y^2 + z^2)^{1/2}}\right)
$$

Observar que es negativo, ya que $a<b$ y por lo tanto el término negativo es mayor que el positivo. También, no es constante, sino que depende de la posición en el plano: los puntos más cercanos al hilo tienen una carga inducida de mayor valor absoluto.

El campo eléctrico en dirección $x$ inducido por el hilo $-\lambda$ es igual, por geometría. Matemáticamente, la integral es la misma, cambiando:

$$
\lambda &\rightarrow -\lambda \\
a &\rightarrow -b \\
b &\rightarrow -a
$$

Por lo tanto, el campo eléctrico total $\vec{E}$ en un punto $(0,y,z)$ de la superficie del CEP es:

$$
\vec{E} = \frac{\lambda}{2\pi\varepsilon_0} \left(\frac{1}{(b^2 + y^2 + z^2)^{1/2}} - \frac{1}{(a^2 + y^2 + z^2)^{1/2}}\right) \vec{u_x}
$$

Por lo tanto:

$$
\sigma = \frac{\lambda}{2\pi} \left(\frac{1}{(b^2 + y^2 + z^2)^{1/2}} - \frac{1}{(a^2 + y^2 + z^2)^{1/2}}\right) 
$$


#### B) Integre la densidad de carga superficial obtenida anteriormente e interprete el resultado (2 puntos).

La carga total inducida es la integral de la densidad de carga inducida anteriormente obtenida:

$$
Q_T = \int \sigma dS
$$

Los límites de integración serían todo el plano CEP:

$$
Q_T = \int_{y{=}-\infty}^{y{=}+\infty}\int_{z{=}-\infty}^{z{=}+\infty} \frac{\lambda}{2\pi} \left(\frac{1}{(b^2 + y^2 + z^2)^{1/2}} - \frac{1}{(a^2 + y^2 + z^2)^{1/2}}\right) dy\space dz
$$

La manera más sencilla de plantear esta integral es cambiar a polares en el plano $dS = dy\space dz = \rho d\rho \space d\theta$, con $\rho^2 = y^2 + z^2$.

$$
Q_T = \int_{\rho{=}0}^{\rho{=}+\infty}\int_{\theta{=}0}^{\theta{=}2\pi} \frac{\lambda}{2\pi} \left(\frac{1}{(b^2 + \rho^2)^{1/2}} - \frac{1}{(a^2 + \rho^2)^{1/2}}\right) \rho d\rho \space d\theta
$$

El resultado de la integral en $\theta$ es inmediato, $2\pi$, mientras que en $\rho$ es:

$$
Q_T = \lambda \left( (b^2 + \rho^2)^{1/2} - (a^2 + \rho^2)^{1/2} \right) |_{\rho{=}0}^{\rho{=}+\infty} 
$$

El límite superior es cero. En realidad es una indeterminación de tipo $\infty-\infty$, que se resuelve algebráicamente. Por ejemplo, aproximamos la raiz $(a^2 + \rho^2)^{1/2} \approx \rho$, hacemos lo mismo con $b$ y restamos antes de hacer el límite. El límite inferior da: 

$$
Q_T = -\lambda \left( b-a \right)
$$

Es decir, la carga total inducida es la carga total del hilo imagen, como ya sabíamos.
