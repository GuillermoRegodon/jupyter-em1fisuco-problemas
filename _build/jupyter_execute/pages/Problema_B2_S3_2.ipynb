# Problema B2.S3.2

Cuando el conductor eléctrico perfecto es una esfera, el problema pierde la simetría de reflexión. Sin embargo, el concepto detrás de la imagen es el mismo. Vamos a revisar el problema suponiendo que la esfera estuviera conectada a tierra:

````{note}

* Queremos sustituir el CEP por una configuración de cargas que permita que el potencial en la superficie del CEP sea $\Phi_{CEP} = 0$.


```{figure} ../images/B2_S3_2_1.png
:alt: esfera CEP y carga.png
:width: 350px
```

```{note}
Recordemos que un conductor eléctrico perfecto (*CEP*) tiene las siguiente características, derivadas de la libertad de movimiento de las cargas dentro del CEP.
* El CEP impone un potencial constante en su superficie (*y en general en todo su volumen*).
* El campo eléctrico $\vec{E}_{CEP}$ en la superficie del CEP es perpendicular (*o nulo*) a dicha superficie. Está garantizado si el potencial el constante en la superficie.
* La densidad de carga superficial es la que permite cumplir estas dos condiciones. Puede ser cualquiera, dependiendo de las cargas fuera del conductor. Solamente se puede calcular a posteriori, después de conocer el campo eléctrico $\vec{E}_{CEP}$ en la superficie del CEP.
```

Sabemos que la imagen de una carga $q$ a una distancia $d$ del centro de la esfera de CEP de radio $a$ tiene las siguientes características:

* La carga imagen es una carga puntual en la misma linea que pasa por la carga original y por el centro de la esfera de CEP.
* El valor de la carga imagen y la distancia con respecto al centro de la esfera son:

$$
q' &= -q \frac{a}{d} \\
d' &= \frac{a^2}{d}
$$

Por lo tanto, la imagen es la siguiente:

:alt: esfera CEP y carga.png
:width: 350px
```

````

Teniendo este problema revisado, veamos las dos variaciones más inmediatas que se pueden realizar sobre este problema, que están muy relacionadas entre si:

````{note}

* La esfera de carga tiene un potencial $V \neq 0$ y una carga neta $Q$.

Vamos a ver que los dos valores están muy relacionados entre si, y que la elección de cuál es el dato y cuál se calcula a partir del otro depende del enunciado.

Sabemos que la única imagen que permite que la esfera de carga este a potencial nulo es la imagen $q'$. **¡No existe imagen puntual que permita que la esfera de CEP tenga otro potencial!**. Sin embargo, gracias al **Principio de Superposición** podemos añadir otra configuración de cargas que permita que la esfera de carga esté a potencial $V \neq 0$ constante (*Pensar en qué condición debe cumplir la configuración de cargas imagen para conseguir que $\Phi(r{=}a)=cte$*). Elegimos por simplicidad la más sencilla, una carga puntual en el origen, de valor $Q_c$. 

```{figure} ../images/B2_S3_2_3.png
:alt: esfera CEP y carga.png
:width: 350px
```

Por lo tanto, el valor del potencial en la superficie de la esfera de CEP es la suma del potencial que generan la carga original y su imagen, es decir 0, más el potencial que genera la carga central $Q_c$:

$$
V = \Phi(r{=}a) = \Phi(r{=}a)_q + \Phi(r{=}a)_{-q} + \Phi(r{=}a)_{Q_c} = 0 + \frac{1}{4\pi\varepsilon_0}\frac{Q_c}{a}
$$

* Si el problema nos dice que la esfera de CEP está conectada a una fuente de alimentación de valor $V$, el potencial de la esfera es el dato y la carga central $Q_c$ se puede obtener a partir de dicho valor $V$.

* Si el problema nos dice que la esfera de CEP está aislada con una carga total $Q$ (*Observar que $Q$ podría ser nula, $Q=0$*), calculamos la carga central $Q_c$ utilizando el valor neto $Q$ de la carga dentro de la esfera de CEP, $Q = Q_c - q'$, y obtenemos el potencial $V$ a partir de este valor $Q_c$.

````

Este problema es del segundo tipo, por lo que:

$Q_c = Q - q' = Q + q\frac{a}{d}$

### Fuerza nula sobre la carga

Nos piden el valor de la carga $q$ tal que no sufre fuerza neta cuando se coloca a una distancia $d$ del centro de la esfera, lo que es lo mismo que decir que el campo eléctrico es nulo, $\vec{E} = 0$. Por lo tanto, tan solo tenemos que sumar el campo eléctrico que generan ambas cargas imagen, $q'$ y $Q_c$. Quedándonos sólo con la componente radial que es por simetría la que no es nula (*¿Qué tipo de simetría? No es simetría esférica...*).

$$
E_r(r{=}d) &= E_{r,q'}(r{=}d) + E_{r,Q_c}(r{=}d) = \frac{1}{4\pi\varepsilon_0} \left( \frac{-q\frac{a}{d}}{(d-\frac{a^2}{d})^2} + \frac{Q + q\frac{a}{d}}{d^2} \right) = 0 \\
$$

Con esta expresión, podemos despejar $q$ en función de los datos del problema.

$$
q = \frac{(d^2-a^2)^2 dQ}{a^3(2d^2-a^2)} = \frac{(1-(\frac{a}{d})^2)^2 Q}{(\frac{a}{d})^3(2-(\frac{a}{d})^2)}
$$

### Trabajo

La definición de trabajo es conocida, a partir de la fuerza:

$$
W = \int \vec{F}\cdot\vec{dl}
$$

Es decir, necesitamos el campo eléctrico que se aplica sobre la carga, tenemos que elegir un camino para traer la carga $q$ desde su posición $d$ hasta el infinito, y realizar la integral. En este caso, **¡no es un problema inmediato!** (*¿Por qué?*).

La imagen $q'$, tanto su valor como su posición varían cuando la posición de la carga original $q$ se desplaza. Por ejemplo, cuando la carga $q$ está en el infinito, la imagen es nula y en el origen, y a medida que se acerca, la imagen va modificando su valor y desplazándose. Es decir, la fuerza que afecta a la carga $q$ no es la de la imagen final $q' = -q\frac{a}{d}$ en la posición $d' = \frac{a^2}{d}$, sino que *¡la fuerza que se aplica en cada posición está relacionada con su posición actual!*

Vamos a usar la letra $d$ para la posición inicial, y $l$ como la posición actual. El camino sera a través de la linea infinita que pasa por las dos cargas imagen y la posición final. Por lo tanto, la variable que determina el camino es $l$, su diferencial es $dl$ y el intervalo de integración es $l \in [0, \infty)$. La fuerza $F_r$ debe ser proporcional al campo eléctrico $E_r$ que sufre la partícula, con el signo cambiado. Recopilamos.

$$
W = \int \vec{F}\cdot\vec{dl} = -q\int_d^\infty \vec{E}\cdot\vec{dl} = -q\int_d^\infty \left( E_{r,q'}(r{=}l)\vec{u_r} + E_{r,Q_c}(r{=}l)\vec{u_r} \right) \cdot (dl \vec{u_r})
$$

Antes de sustituir, vamos a pensar en cuál es el signo que debe tener. Suponiendo que $q$ y $Q$ son positivas, el trabajo de acercarlas debe ser positivo, ya que debe suponer un gasto de energía acercarlas. Sin embargo, nos piden alejarlas, por lo que esperamos que el trabajo sea negativo, o en otros términos, que la configuración de cargas devuelva energía eléctrica (*posiblemente en forma de energía cinética de $q$*):

* $\vec{dl}$ es radial hacia afuera, $\vec{dl} = dl\vec{u_r}$ con los límites de integración empezando en $d$ y acabando en el infinito.

* El campo eléctrico es radial hacia afuera, por lo que la fuerza necesaria para superarlo en radial hacia dentro. De aquí surge el signo menos delante de la integral.

* La carga sobre la que se aplica el campo es $q$.

Por lo tanto:

$$
W = -\frac{q}{4\pi\varepsilon_0} \int_d^\infty \left( \frac{-q\frac{a}{l}}{(l-\frac{a^2}{l})^2} + \frac{Q + q\frac{a}{l}}{l^2} \right) dl = \frac{q}{4\pi\varepsilon_0} \left( \frac{aq}{2(d^2-a^2)} - \frac{Q}{d} - \frac{aq}{2d^2} \right)
$$

Por último, sustituyendo el valor de $q$ obtenido anteriormente, obtenemos el trabajo que devuelve la partícula de carga $q$ cuando se desplaza desde la posición de equilibrio (*¿Estable o inestable?*) hasta el infinito.

$$
W = -\frac{(d^2-a^2) Q^2 (3d^2-a^2)}{8\pi\varepsilon_0a^3(2d^2-a^2)}
$$
