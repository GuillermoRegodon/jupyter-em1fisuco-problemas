# Problema B5.S9.3

Este es otro problema de un material con una magnetización constante.

$$
\vec{M} = ks^2\vec{u_\phi}
$$

Dado que la magnetización es fija, no tiene sentido hablar de susceptibilidad magnética $\chi_M$ ni de permeabilidad relativa $\mu_r$. Al tener la magnetización $\vec{M}$, podemos obtener las densidades de corriente de magnetización, que a su vez serán las fuentes del campo magnético $\vec{B}$.

```{note}
De forma similar a lo que nos encontrábamos en los ejercicios de Electrostática de materiales, en los problemas de Magnetostática de materiales, vamos a tener tres posibilidades, una de las cuales es particular de los problemas de magnetostática.

1. **Materiales lineales**: Se caracterizan por que la magnetización $\vec{M}$ es lineal con respecto al vector $\vec{H}$, típicamente tan solo una constante (*aunque podría ser una matriz, o tensor de orden 2*).

$$
\vec{M} = \chi_M \vec{H}
$$

2. **Materiales con magnetización constante**: Se caracterizan por que la magnetización es independiente del vector $\vec{H}$ externo. En este caso, las densidades de corriente de magnetización también son constantes, por lo que podemos considerarlas como fuentes de campo magnético $\vec{B}$ y sustituir el material por dichas corrientes.

$$
\vec{M} = cte \quad \Rightarrow \quad \vec{J_M} = \nabla\times\vec{M}; \space \vec{j_M} = \vec{M}\times\vec{n}
$$

Ambos casos, **1** y **2**, son equivalentes a problemas de Electrostática de materiales. El siguiente caso, **3**, aunque nos lo podríamos encontrar en Electrostática, allí tiene muy poca importancia, mientras que en Magnetostática tiene muchas aplicaciones prácticas.

3. **Materiales no lineales**: Se caracterizan por que la magnetización $\vec{M}$ responde al vector $\vec{H}$ externo de forma no lineal. Un ejemplo importante son los materiales con histéresis. Típicamente, le damos la vuelta a la dependencia para que $\vec{H}$ sea la variable independiente y $\vec{B}$ sea la variable dependiente:

$$
\vec{M} = f(\vec{H}) \quad \Rightarrow \quad \vec{B} = \mu_0(\vec{M} + f(\vec{H})) = g(\vec{H}) \quad \Rightarrow \quad \vec{H} = \vec{H}(\vec{B})
$$
```

Calculamos las densidades de corriente de magnetización. Respecto a la densidad volumétrica de corriente de magnetización $\vec{J_M}$:

$$
\vec{J_M} = \nabla\times\vec{M}
$$

Observamos que la magnetización solo tiene componente $\phi$ que solo depende de la coordenada radial $s$ de cilíndricas, $\vec{M} = M_\phi(s)\vec{u_\phi}$, y el único término del rotacional que sobrevive es (*Importante verificar la expresión: el producto $sM_\phi$ se hace antes de la derivada parcial*):

$$
\vec{J_M} = \nabla\times\vec{M} = \frac{1}{s} \left( \frac{\partial(sM_\phi)}{\partial s} \right) \vec{u_z} = 3ks\vec{u_z}
$$

Mientras que la densidad superficial de corriente de magnetización (*Importante: $s$ es arbitrario, en la superficie $s$ se particulariza a $s=a$*):

$$
\vec{j_M} = \vec{M}\times\vec{n} = \left(ks^2\vec{u_\phi}\times\vec{u_s}\right)_{s{=}a} = -ka^2 \vec{u_z}
$$

El problema resultante tiene simetría cilíndrica, una de las que nos permite utilizar la ley de Ampere. Distinguimos dos casos:

* $s{\le}a$: La corriente encerrada es la fracción de la corriente que atraviesa el circuito.

$$
\oint\vec{B}\cdot\vec{dl} = \mu_0I_{enc} = \mu_0\int \vec{J_M}\cdot\vec{dS}
$$

Para resolver la integral, utilizamos el diferencial de superficie de un plano $z=cte$, es decir, $\vec{dS} = s'ds'd\phi'\vec{u_z}$.

$$
\int \vec{J_M}\cdot\vec{dS} = \int_{s'=0}^{s'=s}\int_{\phi'=0}^{\phi'=2\pi} (3ks'\vec{u_z})\cdot(s'ds'd\phi'\vec{u_z}) = \int_{s'=0}^{s'=s}\int_{\phi'=0}^{\phi'=2\pi} 3 ks'^2 ds'd\phi' = 2\pi ks^3
$$

```{note}
Dado que el radio del circuito de ampere es la variable radial $s$ de coordenadas cilíndricas, **para poder realizar la integral, nos vemos obligados a utilizar una variable radial diferente** para realizar la integral $s`$ que varia $s'\in[0,s]$. Por coherencia, tambien le ponemos la prima a $\phi'$

Otra manera de verlo sería la siguiente: *que la integral en $\phi$ es definida en $\phi \in[0,2\pi]$ mientras que la integral en $s$ es indefinida*.
```

Por lo tanto:

$$
B 2\pi s = \mu_02\pi ks^3 \quad \rightarrow \quad \vec{B}(s) = \mu_0ks^2\vec{u_\phi}
$$

* $s{>}a$: La corriente encerrada es la total, $I$:

$$
\oint\vec{B}\cdot\vec{dl} = \mu_0 I_{enc} = 0
$$

Donde hemos usado el resultado de que la corriente total de magnetización en el material debe ser cero, y dado que el material tiene simetría a lo largo del eje $z$, la corriente total de magnetización que cruza un plano $z=cte$ también debe de ser cero. Podemos verificar por integración que:

$$
\int_S \vec{J_M}\cdot\vec{dS} = - \oint_C j_M dl = 2\pi ka^3
$$

Observemos que hay un salto en el campo magnético $\vec{B}$ tangencial que corresponde con la densidad superficial de corriente de magnetización:

$$
(B_f - B_d)_{s=a} = \mu_0j_M \quad \rightarrow \quad -B_d|_{s=a} = -\mu_0 ka^2
$$
