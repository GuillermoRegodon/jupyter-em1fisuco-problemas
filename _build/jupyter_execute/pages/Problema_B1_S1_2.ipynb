# Problema B1.S1.2

Este problema pide la diferencia de potencial $\Delta \Phi$ entre dos puntos en trono a una configuración de carga superficial constante con forma de semiesfera. Debido a la falta de simetría del problema, es difícil encontrar una expresión general para el potencial $\Phi$ o para el campo eléctrico $\vec{E}$ en cualquier punto del espacio. Por ello, vamos a calcular el potencial $\Phi$ (*con respecto a la referencia  en el infinito*) en los puntos pedidos y realizar la resta.

* **Para calcular el potencial en el primer punto $O$ aplicamos directamente la expresión del potencial $\phi$ con respecto al infinito**.

$$
\Phi(O) = \frac{1}{4\pi\varepsilon_0}\int\frac{\sigma(\vec{r'}) dS'}{R}
$$

Con $\vec{R} = \vec{r}-\vec{r'}$. Tenemos que interpretar todos los elementos que aparecen en la integral de superficie. Para poder aprovechar la simetría esférica de la semiesfera, ponemos el origen en el centro de la semiesfera **de forma que la parametrización natural de la superficie es en coordenadas esféricas**. Llamamos $a$ al radio de la semiesfera.

$$
\sigma(\vec{r'}) &= cte\\
\vec{r} &= (0,0,0) \\
\vec{r'} &= a (\sin \theta' \cos \varphi', \sin \theta' \sin \varphi', \cos \theta') \\
\vec{R} &= - a (\sin \theta' \cos \varphi', \sin \theta' \sin \varphi', \cos \theta') \\
R &= a \\
d S' &= a^2 \sin \theta' d\theta' d \varphi' \\
\theta' &\in [0,\pi/2] \qquad \varphi' \in [0, 2\pi] 
$$

Escribimos la expresión con los elementos sustituidos.

$$
\Phi(O) = \frac{\sigma}{4\pi\varepsilon_0 a}\int dS'
$$

No tenemos que seguir sustituyendo, porque al ser la densidad superficial $\sigma$ y la distancia $R$ ambos constantes, el integrando es 1, y la integral resulta ser directamente la superficie de media esfera, $2 \pi a^2$.

$$
\Phi(O) = \frac{\sigma a}{2 \varepsilon_0} 
$$


* **Para calcular el potencial en el segundo punto $N$ también aplicamos directamente la expresión del potencial $\Phi$ con respecto al infinito**.

$$
\Phi(N) = \frac{1}{4\pi\varepsilon_0}\int\frac{\sigma(\vec{r'}) dS'}{R}
$$

Ponemos el origen del sistema de referencia en el mismo punto por los mismos motivos. Los elementos de la integral son los mismo salvo el punto en el que se calcula la integral.

$$
\vec{r} &= (0,0,a) \\
\vec{R} &= - a (\sin \theta' \cos \varphi', \sin \theta' \sin \varphi', \cos \theta' - 1) \\
R &= a \sqrt{2} \sqrt{1-\cos\theta'}
$$

````{note}
La expresión de $R$ que se puede deducir a partir de la definición, *se puede obtener de otra manera más visual*. La distancia $R$ es la longitud del segmento que une el punto $N$ en la cima de la semiesfera y un punto cualquiera de la semiesfera, que caracterizamos por los angulos $\theta'$ y $\varphi'$.


```{figure} ../images/B1_S1_2_1.png
:alt: semiesfera carga constante.png
:width: 500px
```

Comprobamos que podemos dividir el triangulo $ONr$ en dos triángulos rectángulos de ángulo $\theta/2$, de tal forma que $R = 2a \sin\frac{\theta'}{2}$, y por la expresión del seno del ángulo mitad, obtenemos justo la misma expresión correcta.
````

Con esto, la expresión resulta ser:

$$
\Phi(N) = \frac{\sigma}{4\pi\varepsilon_0}\int_{\theta'=0}^{\theta'=\pi/2} \int_{\varphi'=0}^{\varphi'=2\pi}\frac{a^2 \sin \theta' d\theta' d \varphi'}{2a \sin\frac{\theta'}{2}} = \frac{\sigma a}{\varepsilon_0}\frac{\sqrt 2}{2}
$$

```{note}
La integral se puede realizar cambiando $\theta'/2$ por $\theta'$ o al revés, cambiando $\theta'$ por $\theta'/2$. En ambos casos salen integrales relativamente inmediatas.
```

* Finalmente

$$
\Delta \Phi = \frac{\sigma a}{\varepsilon_0} \left( \frac{\sqrt 2 -1}{2}\right)
$$
