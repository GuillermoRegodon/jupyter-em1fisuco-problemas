# Problema B1.S2.2

Este problema es una aplicación inmediata de las expresiones de los términos monopolar, dipolar y cuadripolar:

$$
Q_T &= \int_{V'} \rho(\vec{r'}) dV' \\
\vec{p} &= \int_{V'} \vec{r'} \rho(\vec{r'}) dV' \\
\overleftrightarrow{Q} &= \int_{V'} \left(3\vec{r'}\vec{r'} - r'^2 \overleftrightarrow{I}\right) \rho(\vec{r'}) dV'
$$

Las dos primeras son aplicación inmediata de expresiones conocidas, solamente es necesario tener cuidado eligiendo los elementos de las integrales. El tercero es menos común, pero no es tan distinto: Es una matriz simétrica, por lo que lo que tenemos que hacer es poner la expresión para los 6 elementos distintos en la base más apropiada, que va a ser la cartesiana, y obtener las seis integrales. En cartesianas, siendo $\vec{r'} = (x',y',z') = (x'_1,x'_2,x'_3)$:

$$
Q_{ij} = \int_{V'} \left(3x'_ix'_j - r'^2 \delta_{ij}\right) \rho(\vec{r'}) dV'
$$

```{note}
El momento cuadripolar es una matriz. Podría elegirse otra base, pero dado que la base varía con respecto al punto en coordonadas esféricas y cilíndricas, los términos de la matríz también, lo que supone una complejidad innecesaria y que dificulta interpretar el resultado.
```

```{note}
Veamos que el momento dipolar $\vec{p}$ es un vector, pero en realidad sólo lo usamos dentro de una expresión multiplicado por el vector de posición. Por ejemplo, para el potencial $\Phi$:

$$
\Phi = \frac{1}{4\pi\varepsilon_0} \frac{\vec{u_r}\cdot\vec{p}}{r^2}
$$

En este sentido, el momento dipolar se puede entender como una función que siempre se aplica sobre un vector, y que devuelve un escalar. El momento cuadripolar es similar, solo que se aplica sobre el vector de posición dos veces.

$$
\Phi = \frac{1}{4\pi\varepsilon_0} \frac{1}{2} \frac{\vec{u_r}\overleftrightarrow{Q}\vec{u_r}}{r^3} = \frac{1}{4\pi\varepsilon_0} \frac{1}{2} \sum_{i,j=1}^3 \frac{x_iQ_{ij}x_j}{r^5}
$$

```

## Apartado a)

Al tratarse de cargas puntuales, utilizamos las expresiones equivalentes en términos de sumatorias (*Es útil tener soltura pasando las expresiones integrales a sumatorias cuando sea necesario*):

$$
Q_T &= \sum_{k=1}^N q_k \\
\vec{p} &= \sum_{k=1}^N \vec{r'_k} q_k \\
\overleftrightarrow{Q} &= \sum_{k=1}^N \left(3\vec{r'_k}\vec{r'_k} - r_k'^2 \overleftrightarrow{I}\right) q_k
$$

Observemos que he preferido utilizar $k$ para la sumatoria, porque tenemos los índices $i$ y $j$ para los índices de las coordenadas cartesianas en el momento cuadripolar. Las tres cargas de este apartado van a ser numeradas como en la siguiente figura:


```{figure} ../images/B1_S2_2_1.png
:alt: cargas puntuales.png
:width: 250px
```

En este caso, ya solo nos queda tener cuidado con los índices y obtener los momentos:

$$
Q_T &= q_1 + q_2 + q_3 = 0 \\
\vec{p} &= q_1 \vec{r'_1} + q_2 \vec{r'_2} + q_3 \vec{r'_3} = q (a \vec{u_z}) -2q (\vec{0}) + q (-a \vec{u_z}) = \vec{0}
$$

Para el momento cuadripolar necesitamos tener cuidado con los índices: $k$ sobre las partículas, e $i$ y $j$ sobre las coordenadas cartesianas, es decir, $x_{ki}$ y $x_{kj}$ son las coordenadas $i$ y $j$ de la carga $k$:

$$
Q_{ij} = \sum_{k=1}^3 \left(3x'_{ki}x'_{kj} - r'^2 \delta_{ij}\right) q_k
$$

```{note}
Visto de otro modo, que quizás se hace más sencillo:

$$
\vec{r}_k = ({x}_k,{y}_k,{z}_k) = ({x}_{k1},{x}_{k2},{x}_{k3})
$$
```

Por lo tanto:

$$
Q_{11} &= q_1 (3x_{11}'^2-(x_{11}'^2+x_{12}'^2+x_{13}'^2)) + q_2 (3x_{21}'^2-(x_{21}'^2+x_{22}'^2+x_{23}'^2))  + q_3 (3x_{31}'^2-(x_{31}'^2+x_{32}'^2+x_{13}'^2))  \\
Q_{12} &= Q_{21} = q_1 (3x'_{11}x'_{12}) + q_2 (3x'_{21}x'_{22}) + q_3 (3x'_{31}x'_{32}) \\
...
$$

(*Usar unos instantes para verificar estas expresiones*)

```{note}
Estas expresiones, quizás se entiendan de una forma más sencilla así:

$$
Q_{11} &= q_1 (3x_1'^2-(x_1'^2+y_1'^2+z_1'^2)) + q_2 (3x_2'^2-(x_2'^2+y_2'^2+z_2'^2)) + q_3 (3x_3'^2-(x_3'^2+y_3'^2+z_3'^2)) \\
Q_{12} &= Q_{21} = q_1 (3x'_1y'_1) + q_2 (3x'_2y'_2) + q_3 (3x'_3y'_3) \\
...
$$

(*Verificar que son equivalentes*)
```

De esta forma,

$$
Q_{11} &= q(0-a^2) - 2q(0-0) + q(0-a^2) = -2qa^2 \\
Q_{22} &= q(0-a^2) - 2q(0-0) + q(0-a^2) = -2qa^2 \\
Q_{33} &= q(3a^2-a^2) - 2q(0-0) + q(3a^2-a^2) = 4qa^2 \\
Q_{12} &= Q_{21} = Q_{13} = Q_{31} = Q_{23} = Q_{32} = q(0) - 2q(0) + q(0) = 0
$$

Con esto, sabemos que el **potencial electrostático $\Phi$ a grandes distancias** de la configuración de carga, tiene la expresión aproximada:

$$
\Phi (\vec{r}) = \frac{1}{4\pi\varepsilon_0} \frac{1}{2} \sum_{i,j=1}^3 \frac{x_iQ_{ij}x_j}{r^5} + ...
$$

más un término de error que decae con la distancia como $1/r^4$ o más rápidamente.



## Apartado b)

Este apartado pide los momentos monopolar, dipolar y cuadripolar de una configuración esférica sin simetría:

$$
\sigma(\theta') = \sigma_0 \cos(\theta')
$$


```{figure} ../images/B1_S2_2_2.png
:alt: carga esferica sin simetría.png
:width: 300px
```

Al tratarse de una carga superficial, utilizamos las expresiones de las integrales en forma de integrales de superficie:

$$
Q_T &= \int_{S'} \sigma(\vec{r'}) dS' \\
\vec{p} &= \int_{S'} \vec{r'} \sigma(\vec{r'}) dS' \\
\overleftrightarrow{Q} &= \int_{S'} \left(3\vec{r'}\vec{r'} - r'^2 \overleftrightarrow{I}\right) \sigma(\vec{r'}) dS'
$$

La forma esférica nos sugiere que conviene parametrizar en coordenadas esféricas. Los elementos que aparecen en las integrales son:

$$
\sigma(\vec{r'}) &= \sigma_0 \cos\theta' \\
dS' &= a \sin\theta' d\theta' d\varphi' \\
\vec{r'} &= a(\sin\theta' \cos\varphi', \sin\theta' \sin\varphi', \cos\theta') \\
r'^2 &= a^2 
$$

Con esto, $Q_T = 0$ y:

$$
\vec{p} = \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} a(\sin\theta' \cos\varphi', \sin\theta' \sin\varphi', \cos\theta') \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi'
$$

Realizando las tres integrales de las componentes por separado, vemos que las integrales en las coordenadas $x$ e $y$ tienen la integral en $\varphi'$ de $\int_0^{2\pi}\sin\varphi' d\varphi'$ y $\int_0^{2\pi}\cos\varphi' d\varphi'$, ambas nulas. Sin embargo, la coordenada $z$

$$
p_z = \sigma_0 a^2 \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \cos^2\theta' \sin\theta' d\theta' d\varphi' = \frac{4}{3}\pi\sigma_0a^3
$$

Para el momento cuadripolar, observamos que necesitamos los siguientes elementos en las integrales:

$$
x_1'^2 &= x'^2 = a^2 \sin^2\theta' \cos^2\varphi' \\
x_2'^2 &= y'^2 = a^2 \sin^2\theta' \sin^2\varphi' \\
x_3'^2 &= z'^2 = a^2 \cos^2\theta' \\
r'^2 &= a^2 \\
x'_1x'_2 &= x'y' = a^2 (\sin\theta' \cos\varphi' \sin\theta' \sin\varphi') = a^2 (\sin^2 \theta' \cos\varphi' \sin\varphi') \\
x'_1x'_3 &= x'z' = a^2 (\sin\theta' \cos\varphi' \cos\theta') \\
x'_2x'_3 &= y'z' = a^2 (\sin\theta' \sin\varphi' \cos\theta') \\
$$

Con lo que:

$$
Q_{11} &= \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \left(3x'^2 - r'^2 \right) \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi' \\
Q_{22} &= \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \left(3y'^2 - r'^2 \right) \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi' \\
Q_{22} &= \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \left(3y'^2 - r'^2 \right) \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi' \\
Q_{12} &= Q_{21} = \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \left(3x'y' \right) \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi' \\
Q_{13} &= Q_{31} = \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \left(3x'z' \right) \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi' \\
Q_{23} &= Q_{32} = \int_{\theta'=0}^{\theta'=\pi}\int_{\varphi'=0}^{\varphi'=2\pi} \left(3y'z' \right) \sigma_0 \cos\theta' a \sin\theta' d\theta' d\varphi' \\
$$

Realizando todas las integrales, demostramos que son todas nulas. Por lo tanto, el **potencial electrostático $\Phi$ a grandes distancias** de la configuración de carga, tiene la expresión aproximada:

$$
\Phi (\vec{r}) = \frac{1}{4\pi\varepsilon_0} \frac{\vec{u_r}\cdot\vec{p}}{r^2} + ...
$$

más un término de error que decae con la distancia como $1/r^4$ o más rápidamente.
