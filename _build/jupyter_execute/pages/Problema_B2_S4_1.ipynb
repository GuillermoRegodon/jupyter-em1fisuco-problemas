# Problema.B2.S4.1

En este problema tenemos un hilo de carga caracterizado por una densidad lineal de carga constante $\lambda$. Lo único que hemos podido demostrar matemáticamente es: ¿cuál es la imagen y la posición de una carga puntual? Con esto podemos construir una densidad lineal de carga constante como una sucesión de cargas puntuales $q$ en la posición $z$ equiespaciadas con una distancia $dz$. Para tener la imagen clara, el problema original es el siguiente:


```{figure} ../images/B2_S4_1_1.png
:alt: lineal y esfera.png
:width: 320px
```

El problema equivalente en el que la densidad lineal de carga se descompone en trozos diferenciales es el siguiente:

```{figure} ../images/B2_S4_1_2.png
:alt: cargas diferenciales y esfera.png
:width: 320px
```

Claramente, $q = \lambda \space dz$. **Cada uno de estos trozos diferenciales tienen su propia imagen** $q'$ a una distancia $z'$ del centro de la esfera. 

$$
q' &= -\frac{a}{z}q \\ 
z' &= \frac{a^2}{z}
$$

Si utilizamos la definición de $q = \lambda dz$ y $q' = \lambda' dz'$ obtenemos

$$
\lambda' dz' = -\frac{a}{z} \lambda dz
$$

Para poder identificar la expresión de $\lambda'$ como una función de $z'$, en la ecuación **falta establecer la relación entre $dz$ y $dz'$**, lo que podemos hacer diferenciando en la expresión de $z'$.

$$
dz' = -\frac{a^2}{z^2}dz \quad \Rightarrow \quad \left|\frac{dz'}{dz}\right| = \frac{a^2}{z^2} 
$$

He ignorado el signo, porque **sobre el trozo diferencial $q'$ en $z'$ solo nos interesa su longitud $dz'$, no si está a la derecha o a la izquierda**. Con esto:

$$
\lambda'\frac{a^2}{z^2} = -\frac{a}{z} \lambda
$$

**Finalmente**, despejando y utilizando la relación entre $z$ y $z'$, obtenemos:

$$
\lambda' = -\lambda\frac{a}{z'}
$$

Por lo tanto, **la distribución imagen es una distribución lineal de carga variable $\lambda'(z')$ entre $z' = a/3$ y $z'=a/2$**, que son las posiciónes de las imagenes en $3a$ y $2a$, respectivamente. **Para obtener la carga total de la imagen, utilizamos la carga total de una distribución lineal de carga**.

$$
Q_T = \int_{a/3}^{a/2}\lambda'(z') dz'
$$

Y finalmente, para obtener **el potencial a partir de las dos distribuciones lineales de carga** utilizamos la expresión del potencial:

$$
\Phi(z) = \frac{1}{4 \pi \varepsilon_0} \int_C \frac{\lambda(\vec{r}')dl'}{|\vec{r} - \vec{r}'|}
$$

**Tenemos dos contribuciones, la de $\lambda$ y la de $\lambda'(z')$**. Vamos a detallar los elementos de cada una de las dos integrales, aunque primero vamos a observar que tanto la distribución lineal de carga como el punto $z$ están en el mismo eje. Además, **tenemos que tener cuidado con la notación, ya que las primas $'$ en la integral se refieren a las variables espaciales en la región de la fuente que genera el potencial**. Por eso, cuando haga la integral en el hilo $\lambda$, también voy a usar $z'$, como la posición en el hilo que genera el potencial. Vamos a calcular tan solo la integral para $z>3a$. Para los rangos $z \in [a,2a]$ y $z \in [2a,3a]$ las integrales son similares, pero hay que tener cuidado con los signos en el valor absoluto del denominador para que el denominador sea positivo (*Obviamente, para $z<a$ el potencial es nulo*). Para el hilo original:

$$
\int_C &= \int_{z' = 2a}^{z' = 3a} \\
\lambda(\vec{r}') &= \lambda \\
dl &= dz' \\
|\vec{r} - \vec{r}'| &= z-z' 
$$

**Observemos que $|\vec{r} - \vec{r}'| = z-z'$ se puede hacer por que el enunciado pide calcular el valor para $z >3a$**, es decir, que $z$ siempre es mayor que $z'<=3a$. Para el hilo imagen:

$$
\int_C &= \int_{z' = a/3}^{z' = a/2} \\
\lambda(\vec{r}') &= \lambda(z') = -\lambda\frac{a}{z'} \\
dl &= dz' \\
|\vec{r} - \vec{r}'| &= z-z' 
$$

Con lo que **el potencial se puede obtener resolviendo la siguiente expresión**:

$$
\Phi(z) = \frac{1}{4 \pi \varepsilon_0} \left( \int_{z' = 2a}^{z' = 3a} \frac{\lambda dz'}{z-z'} + \int_{z' = a/3}^{z' = a/2} \frac{-\lambda a dz'}{z'(z-z')} \right)
$$

La primera integral es inmediata, y la segunda se puede resolver por fracciones simples ($ \frac{z}{z'(z-z')} = \frac{1}{z'} + \frac{1}{z-z'} $).
