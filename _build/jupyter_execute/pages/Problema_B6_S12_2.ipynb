# Problema B6.S12.2

En este problema tenemos dos espiras concéntricas, una mucho más grande que la otra. No son coplanarias, sino que sus normales subtienden un ángulo relativo $\theta$. Para calcular el coeficiente de inducción mutua, necesitamos el flujo de campo magnético $\Phi_{21}$ que pasa por una espira a causa de la corriente $I$ que pasa por la otra espira.

$$
M = \frac{\Phi_{21}}{I}
$$

```{figure} ../images/B6_S12_2_1.png
:alt: induccion
:width: 260px
```

Sabemos que es el mismo independientemente de cuál sea la espira que genera campo magnético y cuál la que integra el flujo, por lo que elegimos la más fácil.

```{note}
Muchos problemas de magnetostática se reducen a obtener dos integrales:

* Una integral para obtener el campo magnético $\vec{B_1}$ que genera una configuración de corriente 1. En esta integral se parametriza la configuración de corriente 1.

* Una integral para obtener el flujo magnético $\Phi_{21}$ que genera la espira 1 sobre la espira 2. En esta integral se parametriza la superficie delimitada por la configuración de corriente 2.

Alguna, o las dos integrales, se pueden resolver por métodos simplificados (*podría ser Ampere, o producto $B_1S_2$... cada problema debe ser analizado para comprobarlo*).
```

Como la espira de radio $a$ es mucho más pequeña, el campo magnético que la atraviesa será más o menos constante, por lo que podremos aproximar el flujo por:

$$
\Phi_{21} = \int_S \vec{B_{21}}\cdot\vec{dS} \approx B_{21} \pi a^2 \cos\theta
$$

Estoy es válido si elegimos la espira 1 como la grande, de radio $b$, por la que pasa la corriente I, de forma que el campo magnético $\vec{B_{21}}$ en la región donde esta la espira 2, la pequeña de radio $a$, es prácticamente constante. **Si eligíeramos al revés, es decir, si la espira que genera el campo magnético fuera la pequeña, el campo que atraviesa la otra, la grande, no sería constante y sería difícil de integrar**.

```{figure} ../images/B6_S12_2_2.png
:alt: induccion
:width: 460px
```

Por lo tanto, en este problema, elegimos que la corriente $I$ pase por la espira grande de radio $b$ y calculamos el flujo sobre la espira pequeña de radio $a$.

$$
M = \frac{\Phi_{21}}{I} \approx \frac{B_{21} \pi a^2 \cos\theta}{I}
$$

Tan solo nos falta tener $B_{21}$, el campo generado por una espira en su centro. Este se calcula fácilmente a través de la integral de campo magnético. No tiene ninguna simetría para usar Ampere que, aunque se cumpla, no me permite obtener el resultado de forma sencilla. El resultado es:

$$
B_{21} = B_1(z{=}0) = \frac{\mu_0I}{2 b}
$$

Con lo que el coeficiente de inducción mutua es:

$$
M = \frac{\mu_0\pi a^2\cos\theta}{2 b}
$$

Para obtener el segundo apartado, el efecto de la corriente de una espira sobre la otra ocurre esencialmente a través de la ley de inducción de Lenz-Faraday. La espira pequeña se pone a girar a velocidad angular constante $\omega$, es decir, $\theta = \omega t$. **Ponemos una resistencia $R$ en la espira con corriente inducida por dos motivos**: Para limitar la corriente y así poder **despreciar los efectos de autoinducción**, y para poder **convertir la fuerza electromotriz $\mathcal{E}$ en una corriente**.

$$
I_2 = \frac{\mathcal{E}}{R}
$$

La fuerza electromotriz solo puede obtenerse a partir de la ley de inducción:

$$
\mathcal{E} = -\frac{d\Phi_{21}}{dt}
$$

Con esto, finalmente:

$$
I_2 = -\frac{1}{R} \frac{d\Phi_{21}}{dt} = -\frac{1}{R} \frac{d}{dt}\left( MI \right) = -\frac{1}{R} \frac{d}{dt}\left( \frac{\mu_0I\pi a^2\cos(\omega t)}{2 b} \right) = \frac{\mu_0I\pi a^2\omega\sin(\omega t)}{2 bR}
$$
