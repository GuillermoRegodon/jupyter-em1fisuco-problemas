# Problema B2.S3.1

Este problema es el más básico de imágenes, pero es interesante porque pide calcular muchas cosas distintas en los distintos subapartados. Se trata de una partícula de carga $q$ a una distancia $d$ de un plano de conductor eléctrico perfecto (*CEP*) conectada a un potencial nulo, $\Phi_{CEP} = 0$.


```{figure} ../images/B2_S3_1_1.png
:alt: plano CEP y carga.png
:width: 350px
```

La imagen de la carga puntal $q$ es fácil, pero antes de establecerla y empezar a hacer cálculos, vamos a considerar las condiciones que debe cumplir la imagen para que sea válida (*El argumento también será válido en configuraciones más complejas*).

* Queremos sustituir el CEP por una configuración de cargas que permita que el potencial en la superficie del CEP sea $\Phi_{CEP} = 0$.

```{note}
Recordemos que un conductor eléctrico perfecto (*CEP*) tiene las siguiente características, derivadas de la libertad de movimiento de las cargas dentro del CEP.
* El CEP impone un potencial constante en su superficie (*y en general en todo su volumen*).
* El campo eléctrico $\vec{E}_{CEP}$ en la superficie del CEP es perpendicular (*o nulo*) a dicha superficie. Está garantizado si el potencial el constante en la superficie.
* La densidad de carga superficial es la que permite cumplir estas dos condiciones. Puede ser cualquiera, dependiendo de las cargas fuera del conductor. Solamente se puede calcular a posteriori, después de conocer el campo eléctrico $\vec{E}_{CEP}$ en la superficie del CEP.
```

* La imagen debe garantizar que el potencial en la superficie del CEP es la apropiada, $\Phi_{CEP} = 0$, permitiendo sustituir una por la otra en los cálculos.

* El potencial $\Phi(\vec{r})$ y el campo eléctrico $\vec{E}(\vec{r})$ calculados con las cargas y sus imágenes solo es válido fuera del CEP. Dentro del CEP, el modelo no es válido, y el potencial es constante $\Phi_{CEP}(\vec{r}) = \Phi_{CEP} = 0$.

Pensar en términos de potencial es a menudo más sencillo en problemas de imágenes. Por lo tanto, necesitamos una configuración de cargas que permita que el potencial sea 0 en la superficie del CEP, teniendo una carga $q$ que ya impone un potencial distinto de 0. Por lo tanto, **contamos con la posición y el valor de las cargas imágenes**: Aunque se puede demostrar matemáticamente, por inspección vemos que la única opción es una carga negativa $-q$ a una distancia $d$ dentro del CEP, de tal modo que el potencial que impone es del mismo valor y de signo contrario. La combinación de ambos da 0 en la superficie, y podemos retirar el CEP (*como la solución es única, y hemos encontrado una solución aunque sea por inspección, la solución es la correcta*).

Es decir, la configuración equivalente, que verifica que el potencial en la superficie del CEP es $\Phi_{CEP} = 0$ es la siguiente:


```{figure} ../images/B2_S3_1_2.png
:alt: plano CEP y carga.png
:width: 350px
```

Elegimos el Sistema de Referencia de tal modo que el plano de CEP está en $x=0$, y la carga $q$ está en $\vec{r}'_q = d\vec{u_x}$, entonces la imagen $q'$:

$$
q' &= -q \\
\vec{r}'_{q'} &= -d\vec{u_x}
$$

Procedemos ahora a calcular los resultados que pide el problema:

### Fuerza sobre la carga:

El efecto del plano de CEP se sustituye por el efecto de la imagen, por lo que el campo eléctrico al que está sometida la carga $q$ es el que crea su imagen $-q$:

$$
\vec{E}_{-q}(\vec{r}) = \frac{1}{4\pi\varepsilon_0} \frac{-q (\vec{r}-\vec{r}'_{-q})}{|\vec{r}-\vec{r}'_{-q}|^3}
$$

(*Por supuesto, esta expresión solamente es válida para $x>0$*). Y la fuerza a la que está sometida la carga $q$ es:

$$
\vec{F}_{-q\rightarrow q} = q \vec{E}_{-q}(\vec{r}_q) 
$$

Sustituyendo:

$$
\vec{F}_{-q\rightarrow q} = -\frac{q^2}{16\pi\varepsilon_0 d^2} \vec{u_x}
$$


### La densidad de carga inducida en el pano de CEP

Sabemos que el CEP exige que el campo eléctrico en su superficie $\vec{E}_{CEP}$ sea perpendicular a la misma, mientras que dentro del propio CEP, el campo eléctrico es nulo. Esta discontinuidad en el campo eléctrico solamente se puede explicar a través de la densidad superficial de carga inducida, mediante la expresión de la condición:

$$
E_2 - E_1 = \frac{\sigma}{\varepsilon_0}
$$

(*¿Cómo se demuestra esta expresión a partir de la ley de Gauss?*) Si aplicamos esta expresión en la frontera, $E_2 = E_{CEP}$ y $E_1 = 0$. Por lo tanto, solo nos queda calcular el campo eléctrico en la frontera, verificar que es normal a la superficie y aplicar la expresión para despejar $\sigma$. En la superficie,

$$
\vec{r}=(0,y,z)
$$

Y el vector entre el punto $\vec{r}$ y la posición de la carga $\pm q$,

$$
\vec{r} - \vec{r}'_{\pm q} = (\mp d, y, z)
$$

Por lo que el campo eléctrico en la superficie del CEP,

$$
\vec{E}_{CEP}(0,y,z) &= \vec{E}_{q}(0,y,z) + \vec{E}_{-q}(0,y,z) \\
&= \frac{1}{4\pi\varepsilon_0} \frac{q (-d,y,z)}{(d^2 + y^2 + z^2)^{3/2}} + \frac{1}{4\pi\varepsilon_0} \frac{-q (d,y,z)}{(d^2 + y^2 + z^2)^{3/2}}
$$

Comprobamos que la suma de los dos términos, efectivamente tiene dirección $\vec{u_x}$, como debía ocurrir si la imagen está bien obtenida.

$$
\vec{E}_{CEP}(0,y,z) = \frac{1}{4\pi\varepsilon_0} \frac{-2qd}{(d^2 + y^2 + z^2)^{3/2}} \vec{u_x}
$$

Y finalmente la densidad superficial de carga es inmediata:

$$
\sigma_{CEP} =  -\frac{qd}{2\pi (d^2 + y^2 + z^2)^{3/2}}
$$

(*Observar que las unidades son las correctas*) Finalmente, pide la carga total inducida, para lo que podemos integrar $\sigma_{CEP}$ en todo el plano. Realizando el cambio de variable de cartesianas a polares, $(y,z)\Rightarrow (r,\theta)$, la integral es separable y el resultado es fácil de probar.

$$
y &= r \cos \theta \\
z &= r \sin \theta \\
dy \space dz &= r dr\space d\theta \\
r \in [0, \infty) &\quad \theta \in [0, 2\pi)
$$

De tal modo que:

$$
Q_T &= \int_{CEP} \sigma_{CEP} dS \\
&= \int_{y=-\infty}^{y=\infty} \int_{z=-\infty}^{z=\infty} \frac{-qd}{2\pi (d^2 + y^2 + z^2)^{3/2}} dy \space dz \\
&= \int_{r=0}^{r=\infty} \int_{\theta=0}^{\theta=2\pi} \frac{-qd}{2\pi (d^2 + r^2)^{3/2}} r dr\space d\theta =-q
$$

### El potencial y el campo eléctrico en el punto medio entre la carga y el plano

Entre la carga y el plano, estamos fuera del CEP, por lo que el plano se puede sustituir por la carga imagen. Dicho punto es $\vec{r} = \frac{d}{2}\vec{u_x}$. Para calcular el campo eléctrico, por ejemplo, realizaríamos los mismos pasos que en el apartado anterior, cambiando en la solución $(0,y,z)$ por $(\frac{d}{2},0,0)$:

$$
\vec{r} - \vec{r}'_{q} &= (\frac{d}{2},0,0) - (d,0,0) = (-\frac{d}{2},0,0) \\
\vec{r} - \vec{r}'_{-q} &= (\frac{d}{2},0,0) - (-d,0,0) = (\frac{3d}{2},0,0)
$$

Las componentes $E_y$ y $E_z$ del campo eléctrico van a ser cero, viendo las componentes del vector que une el punto y cada una de las cargas. Por tanto:

$$
E_x &= E_{x,-q}(\frac{d}{2},0,0) + E_{x,q}(\frac{d}{2},0,0) \\
&= \frac{1}{4\pi\varepsilon_0} \frac{q (-\frac{d}{2})}{(\frac{d}{2})^3} + \frac{1}{4\pi\varepsilon_0} \frac{-q (\frac{3d}{2})}{(\frac{3d}{2})^3}
$$

Un procedimiento similar permite resolver el potencial en dicho punto a partir del potencial generado por cada una de las cargas.


### Expresión aproximada para $r\rightarrow\infty$

Cuando la distancia en la que calculamos el campo eléctrico es mucho mayor que la distancia $d$ entre el plano de CEP y la carga $q$ (*siempre para $x>0$*), observamos que tenemos una configuración de dos cargas $\pm q$ de signo opuesto separadas una distancia $2d$, por lo que la configuración es equivalente a un dipolo de valor $\vec{p} = 2qd\vec{u_x}$.


### Variaciones

El problema del conductor eléctrico perfecto CEP plano puede tener variaciones:

* ¿Qué ocurre si ponemos una carga $q$ entre dos planos de CEP paralelos? ¿La imagen $q'$ de un plano de CEP genera a su vez una imagen en el otro plano de CEP? ¿Cuántas imágenes hay en total?

* ¿Que ocurre si tenemos dos semiplanos de CEP formando un ángulo de 90º? ¿Cuántas imágenes tiene una carga $q$?
