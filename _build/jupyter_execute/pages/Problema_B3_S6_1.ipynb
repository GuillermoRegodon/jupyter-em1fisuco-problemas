# Problema B3.S6.1

El problema pide resolver la capacidad y la energía electrostática para dos configuraciones esféricas diferentes, con dos distintas configuraciones de dieléctricos en el espacio intermedio entre los dos electrodos (metálicos, de conductor eléctrico perfecto o *CEP*). Las relaciones entre el campo eléctrico $\vec{E}$, el vector polarización $\vec{P}$ y el vector desplazamiento $\vec{D}$ son:

$$
\vec{D} &= \varepsilon_i \vec{E} \\
\vec{P} &= \vec{D} - \varepsilon_0 \vec{E} = (\varepsilon_i - \varepsilon_0) \vec{E}
$$

Y la condición de contorno que nos impone la conexión del condensador es $\Phi(c) = 0$ y $\Phi(a) = V$.

````{note}
Vamos a resolver el problema con un solo dieléctrico.


```{figure} ../images/B3_S6_1_1.png
:alt: condensador.png
:width: 200px
```

La simetría esférica de dicho problema, nos permite utilizar la ecuación de Gauss en forma integral para materiales dieléctricos. Supongamos que el conductor interno tiene un potencial $V$ y una carga libre $Q_l$ total repartida homogéneamente en dicho conductor esférico.

$$
\oint_S \vec{D}\cdot\vec{S} = \oint_S \varepsilon_1 \vec{E}\cdot\vec{S} = Q_l
$$

Y por tanto, para $r\in[a,c]$,

$$
\varepsilon_1 E4\pi r^2 = Q_l \quad \Rightarrow \quad E = \frac{Q_l}{4\pi\varepsilon_1 r^2}
$$

Para relacionar la constante $Q_l$ con el potencial, utilizamos la definición de potencial en forma integral. Utilizando un radio para el camino integral, $\vec{dl} = dr \vec{u_r}$ .

$$
V = \Phi(r{=}a) - \Phi(r{=}c) = -\int\vec{E}\cdot\vec{dl} = -\int_{r=c}^{r=a} E dr = \frac{Q_l}{4\pi\varepsilon_1}\left( \frac{1}{a} - \frac{1}{c} \right)
$$

La capacidad y la energía electrostática acumulada en el condensador son:

$$
C = \frac{Q_l}{V} = \frac{4\pi\varepsilon_1}{\frac{1}{a} - \frac{1}{c}} \\
U_e = \frac{1}{2} C V^2 = \frac{2\pi\varepsilon_1 V^2}{\frac{1}{a} - \frac{1}{c}} 
$$
````

Volvemos a las configuraciones *(a)* y *(b)* con dos dieléctricos: La configuración *(a)* tiene claramente simetría esférica, por lo que podemos utilizar el método de superficies de Gauss esféricas para resolverlo. Veamos qué podemos hacer con la configuración *(b)*, pero para ello vamos a considerar lo siguiente:

```{note}
Dado que se cumple $\vec{D} = \varepsilon_i \vec{E}$ y dentro de los dieléctricos no hay cargas libres, la ley de Gauss nos da lo siguiente (*¡Dentro de los dieléctricos!*):

$$
\nabla\cdot\vec{D} &= 0 \\
\nabla\cdot\vec{E} &= \frac{1}{\varepsilon_i} \nabla\cdot\vec{D} = 0
$$

Es decir, **no hay ni cargas libres ni cargas ligadas en el interior de los dieléctricos**. Efectivamente, lo podemos obtener de la expresión de la densidad volumétrica de carga equivalente de polarización:

$$
\rho_P = -\nabla\cdot\vec{P} = 0
$$

Por lo tanto, **cuando la constante dieléctrica de un medio es constante, la densidad volumétrica de carga equivalente de polarización es nula**. Sin embargo, la densidad superficial de carga equivalente de polarización en general no es nula.

$$
\sigma_P = \vec{P}\cdot\vec{n}
$$
```

Por lo tanto, la única densidad de carga que nos puede aparecer es en la frontera entre los dieléctricos y entre los dieléctricos y los conductores eléctricos perfectos. Confirmamos que efectivamente, la configuración *(a)* tiene simetría esférica también después de considerar la densidad superficial de carga equivalente de polarización. En la configuración *(b)* puede aparecer una densidad de carga equivalente que rompe la simetría esférica. Si observamos la ecuación de Gauss en forma integral para medios dieléctricos.

$$
\oint_S \vec{D}\cdot\vec{S} = Q_l
$$

Vemos que en esta ecuación, la constante dieléctrica no aparece, pero la falta de simetría esférica hace que no podamos asegurar que la carga libre se reparte homogéneamente en la superficie del CEP interno (*¡No lo hace!*), ni que el vector desplazamiento $\vec{D}$ sea constante en una esfera de Gauss (*¡No lo es!*).



### Configuración *(a)*

Supongamos que tenemos una determinada carga libre $Q_l$ en el conductor interior. La ecuación de Gauss en forma integral para medios dieléctricos, con la simetría esférica, eligiendo una superficie de Gauss esférica centrada en el origen con un radio $r$ nos da, para la componente radial (*y única componente*) del vector desplazamiento $\vec{D}$:

$$
\oint_S \vec{D}\cdot\vec{S} = Q_l \quad \Rightarrow \quad D4\pi r^2 = Q_l
$$

Si ahora particularizamos en cada uno de los dos medios, dado que la carga libre encerrada es la misma, tenemos que en el medio 1 y 2 respectivamente:

$$
D_1 4\pi r^2 = D_2 4\pi r^2 = Q_l
$$

Es decir, ambos tienen la misma expresión. La diferencia es que la relación para el vector desplazamiento $D_1$ es válida en el rango $r\in[a,b]$ y la ecuación para $D_2$ es válida en el rango $r\in[b,c]$. Vamos a escribir también las expresiones para el campo eléctrico $\vec{E}$ y para el vector polarización $\vec{P}$.

$$
\vec{D_1} &= \frac{Q_l}{4\pi r^2}\vec{u_r}; \quad \vec{E_1} = \frac{Q_l}{4\pi \varepsilon_1 r^2}\vec{u_r}; \quad \vec{P_1} = (1-\frac{\varepsilon_0}{\varepsilon_1})\frac{Q_l}{4\pi r^2}\vec{u_r} \quad r\in[a,b]\\
\vec{D_2} &= \frac{Q_l}{4\pi r^2}\vec{u_r}; \quad \vec{E_2} = \frac{Q_l}{4\pi \varepsilon_2 r^2}\vec{u_r}; \quad \vec{P_2} = (1-\frac{\varepsilon_0}{\varepsilon_2})\frac{Q_l}{4\pi r^2}\vec{u_r} \quad r\in[b,c]
$$

Como $\varepsilon_1 \ne \varepsilon_2$, vemos que hay una discontinuidad en el campo eléctrico $\vec{E}$ entre los dos dieléctricos, salto producido por la densidad superficial de carga de polarización entre ambos materiales dieléctricos, que es la suma de dos contribuciones, una por cada material. Por lo tanto, en $r=b$:

$$
\sigma_1 = \vec{P_1}\cdot\vec{n} = \vec{P_1}\cdot\vec{u_r} = (1-\frac{\varepsilon_0} {\varepsilon_1})\frac{Q_l}{4\pi b^2} \\
\sigma_2 = \vec{P_2}\cdot\vec{n} = \vec{P_1}\cdot(-\vec{u_r}) = (\frac{\varepsilon_0} {\varepsilon_2}-1)\frac{Q_l}{4\pi b^2}
$$

Y por tanto:

$$
\sigma_{r=b} = (\frac{\varepsilon_0} {\varepsilon_2}-\frac{\varepsilon_0} {\varepsilon_1})\frac{Q_l}{4\pi b^2}
$$

y efectivamente:

$$
E_2(r{=}b) - E_1(r{=}b) = \frac{\sigma_{r=b}}{\varepsilon_0}
$$

(*Entre el material y el CEP también aparecerá otra densidad superficial de carga de polarización, pero como la carga en el CEP es cualquiera que permita cumplir que el potencial sea constante, por lo que siempre va a poder cancelar cualquier densidad superficial de carga de polarización que pudiera aparecer.*)

Lo único que falta es relacionar la carga libre con el potencial $V$. Lo hacemos relacionando el potencial $\Phi(r)$ con el campo eléctrico $\vec{E}$ conocido. Utilizamos la definición de potencial en forma integral:

$$
V = \Phi(r{=}a) - \Phi(r{=}b) = -\int\vec{E}\cdot\vec{dl}
$$

Eligiendo un radio, $\vec{dl} = dr\vec{u_r}$

$$
V = -\int\vec{E}\cdot\vec{dl} = - \int_{r=a}^{r=b} E_1 dr - \int_{r=b}^{r=c} E_2 dr = \frac{Q_l}{4\pi \varepsilon_1 r}|_{r=a}^{r=b} + \frac{Q_l}{4\pi \varepsilon_2 r}|_{r=b}^{r=c} = \frac{Q_l}{4\pi \varepsilon_1} \left( \frac{1}{b} - \frac{1}{a} \right) + \frac{Q_l}{4\pi \varepsilon_2} \left( \frac{1}{c} - \frac{1}{b} \right)
$$

Por último, la capacidad y la energía electrostática acumulada en el condensador son:

$$
C = \frac{Q_l}{V} = \frac{1}{\frac{1}{4\pi \varepsilon_1} \left( \frac{1}{b} - \frac{1}{a} \right) + \frac{1}{4\pi \varepsilon_2} \left( \frac{1}{c} - \frac{1}{b} \right)} \\
U_e = \frac{1}{2}CV^2 = \frac{1}{2}\frac{1}{\frac{1}{4\pi \varepsilon_1} \left( \frac{1}{b} - \frac{1}{a} \right) + \frac{1}{4\pi \varepsilon_2} \left( \frac{1}{c} - \frac{1}{b} \right)} V^2
$$

Observemos que la capacidad $C$ del condensador formado es la configuración con dos dieléctricos es el resultado de la configuración en serie de un condensador esférico $C_1$ de radios $a$ y $b$ con permitividad eléctrica $\varepsilon_1$ y otro condensador esférico $C_2$ de radios $b$ y $c$ con permitividad eléctrica $\varepsilon_2$

$$
C = \frac{1}{\frac{1}{C_1} + \frac{1}{C_2}} \\
C_1 = \frac{4\pi\varepsilon_1}{\frac{1}{b} - \frac{1}{a}} \quad C_2 = \frac{4\pi\varepsilon_2}{\frac{1}{c} - \frac{1}{b}}
$$

Y si $\varepsilon_1=\varepsilon_2$, recuperamos el caso del condensador esférico con un solo dieléctrico.



### Configuración *(b)*

En esta configuración, puede aparecer una densidad superficial de carga de polarización entre los dieléctricos, que rompe la simetría esférica, y entre los dieléctricos y los CEP. Sin embargo, a pesar de esto, vamos a suponer que tanto el campo eléctrico $\vec{E}$, el vector desplazamiento $\vec{D}$ y el vector polarización $\vec{P}$ son radiales. Veremos que esto da una solución consistente y por tanto, dado que la solución en electrostática es única, es la correcta.

```{note}
En muchos contextos, tenemos varias ecuaciones posibles, y algunas de ellas nos ofrecen una visión más clara que otras. Tenemos la ecuación de Gauss en forma integral para medios materiales, que no permite ver simetría esférica, debido a que no hay motivo para que la carga $Q_l$ total del CEP interno esté homogéneamente distribuida, ni que el vector desplazamiento $\vec{D}$ sea constante en una esfera de Gauss.

$$
\oint_S \vec{D}\cdot\vec{S} = Q_l
$$

 Tenemos la ecuación de Gauss en forma integral.

$$
\oint_S \vec{E}\cdot\vec{S} = \frac{Q_T}{\varepsilon_0}
$$

Que tampoco lo permite, por el mismo motivo, ya que la carga total incluye la carga libre y la ligada. También tenemos las versiones diferenciales en los materiales, $\nabla\cdot\vec{D} = 0$ y $\nabla\cdot\vec{E} = \frac{1}{\varepsilon_i} \nabla\cdot\vec{D} = 0$, que hemos utilizado para argumentar que no hay densidad volumétrica de carga equivalente de polarización. Tenemos por último, la ecuación de Poisson:

$$
\nabla^2\Phi = 0
$$

Con las condiciones de contorno, $\Phi(c) = 0$ y $\Phi(a) = V$. Tanto la condición de contorno como la ecuación tienen simetría esférica. ¿El cambio de material entre los dieléctricos supone una ruptura de la simetría esférica? Veamos las condiciones de contorno sobre el potencial, entre los dos materiales:

* El potencial debe de ser continuo.

* El campo eléctrico radial (*paralelo a la superficie que separa los materiales*) y por lo tanto la derivada radial del potencial también.

¿Es suficiente para convencernos que que la ecuación de Poisson tiene simetría esférica? Pero en última instancia, lo único que podemos hacer es resolver el problema con las suposiciones indicadas y comprobar el resultado.
```

Hemos argumentado que la ecuación de Poisson $\nabla^2\Phi = 0$ y la condición de contorno $\Phi(c) = 0$ y $\Phi(a) = V$ sobre el potencial sí tienen simetría radial, por lo que vamos a resolverla, lo que nos dará el campo eléctrico $\vec{E}$, el vector desplazamiento $\vec{D}$ y el vector polarización $\vec{P}$ radiales. Poniendo la ecuación de Poisson en esféricas obtenemos:

$$
\Phi(r) = \frac{A}{r} + B
$$

Sustituyendo las condiciones de contorno:

$$
\Phi(r) = V \frac{\frac{1}{r} - \frac{1}{c}}{\frac{1}{a} - \frac{1}{c}}
$$

El campo eléctrico es inmediato a partir de la definición de potencial. Observamos que es igual en ambos medios dieléctricos:

$$
\vec{E}(r) = \vec{E_1}(r) = \vec{E_2}(r) = \frac{V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}}\vec{u_r}
$$

Las demás magnitudes vectoriales salen a partir del campo eléctrico $\vec{E}$.

$$
\vec{D_1}(r) &= \frac{\varepsilon_1 V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}}\vec{u_r}; \quad \vec{P_1}(r) &= \frac{(\varepsilon_1 - \varepsilon_0)  V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}}\vec{u_r}\\
\vec{D_2}(r) &= \frac{\varepsilon_2 V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}}\vec{u_r}; \quad \vec{P_2}(r) &= \frac{(\varepsilon_2 - \varepsilon_0) V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}}\vec{u_r}
$$

Si quisiéramos poner el resultado en función de la carga libre total del conductor interno $Q_l$, la debemos calcular a partir de la ley de Gauss. Usamos una superficie esférica $S$ que se puede dividir en dos mitades $S_1$ y $S_2$, cada una de las cuales está en el volumen ocupado por cada medio dieléctrico.

$$
Q_l = \oint_S\vec{D}\cdot\vec{dS} = \int_{S_1}\vec{D_1}\cdot\vec{dS_1} + \int_{S_2}\vec{D_2}\cdot\vec{dS_2} = D_1S_1 + D_2S_2 = \frac{\varepsilon_1 V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}} 2\pi r^2 + \frac{\varepsilon_2 V}{r^2}\frac{1}{\frac{1}{a} - \frac{1}{c}} 2\pi r^2 = \frac{(\varepsilon_1 + \varepsilon_2) V}{\frac{1}{a} - \frac{1}{c}} 2\pi
$$

Por último, la capacidad y la energía electrostática acumulada en el condensador son:

$$
C = \frac{Q_l}{V} = \frac{2\pi(\varepsilon_1 + \varepsilon_2)}{\frac{1}{a} - \frac{1}{c}} \\
U_e = \frac{1}{2}CV^2 = \frac{\pi(\varepsilon_1 + \varepsilon_2) V^2}{\frac{1}{a} - \frac{1}{c}}
$$

Observemos que la capacidad $C$ del condensador formado es la configuración con dos dieléctricos es el resultado de la configuración en paralelo de medio condensador esférico $C_1$ de radios $a$ y $c$ con permitividad eléctrica $\varepsilon_1$ y otro medio condensador esférico $C_2$ de los mismos radios $a$ y $c$ con permitividad eléctrica $\varepsilon_2$.

$$
C = C_1 + C_2 \\
C_1 = \frac{2\pi\varepsilon_1}{\frac{1}{a} - \frac{1}{c}}  \quad C_2 = \frac{2\pi\varepsilon_2}{\frac{1}{a} - \frac{1}{c}}
$$

Y si $\varepsilon_1=\varepsilon_2$, recuperamos el caso del condensador esférico con un solo dieléctrico.

```{note}
En esencia, hemos supuesto que la densidad superficial de carga de polarización $\sigma_P$ entre los dos dieléctricos es nula, $\sigma_P=0$, junto con el hecho que que la densidad volumétrica de carga de polarización $\rho_p$ también es nula, $\rho_P=0$, por ser la permitividad $\varepsilon_0$ constante, para poder utilizar la ecuación de Poisson con el término de la densidad volumétrica de carga nulo (*total, libre más ligada*). La solución resultante efectivamente es consistente con $\sigma_P=0$ entre los dieléctricos.

*Obviamente, la densidad superficial de carga de polarización entre los diléctricos y los CEP no es nula. ¿Cuánto vale?*
```
