# Problema B3.S7.2

En este problema, la polarización no viene dada, sino que se obtiene a partir de las condiciones del problema. El enunciado establece que el campo eléctrico en el infinito es constante. Para aprovechar la simetría, vamos a elegir orientarlo en la dirección $z$.

* Pensemos en las condiciones que nos impone el problema. **¿Qué va a pasar con la esfera de dieléctrico al colocarla en un campo eléctrico?** (*Pensar un momento*)

Claramente, la esfera, al ser colocada en un campo eléctrico, dependiendo de su susceptibilidad eléctrica, o lo que es lo mismo, de su permitividad eléctrica $\varepsilon$, va a sufrir una polarización. Inicialmente, la polarización va a ser en la misma dirección del campo eléctrico, pero como consecuencia, van a aparecer unas densidades de carga de polarización. Entonces necesitamos una polarización que genere unas densidades de carga de polarización que sean compatibles con las condiciones de contorno, en particular, con el campo eléctrico constante en el infinito.

* Podríamos pensar que el vector polarización $\vec{P}$ va a ser simplemente $\vec{P} = \vec{D} - \varepsilon_0\vec{E} = (\varepsilon-\varepsilon_0)\vec{E}$, pero es erróneo porque se generarán densidades de carga de polarización que serán fuentes de campo eléctrico. De hecho, comprobaremos que para permitividad eléctrica relativa muy próxima a 1 (*Polarización muy débil*) el resultado se puede aproximar por este valor (*Comprobar en el resultado final*).

Colocando la esfera de dieléctrico en el origen, tanto la densidad superficial de carga de polarización $\sigma_P$ como la densidad volumétrica de carga de polarización $\rho_P$ van a tener ambas simetría en torno al eje $z$. Respecto a la segunda, dado que la permitividad de la esfera es constante y además es dieléctrico y no tiene carga libre, podemos afirmar que la densidad volumétrica de carga de polarización es nula:

$$
\rho_P = -\nabla\cdot\vec{P} = -\nabla\cdot\left( \vec{D} - \varepsilon_0\vec{E} \right) = -\nabla\cdot\left( \vec{D}\left(1-\frac{\varepsilon_0}{\varepsilon}\right) \right) \stackrel{\varepsilon\space cte}{=} \left(\frac{\varepsilon_0}{\varepsilon}-1\right)\nabla\cdot\vec{D} = \left(\frac{\varepsilon_0}{\varepsilon}-1\right)\rho_L = 0
$$

Por lo tanto, solo tenemos densidad de carga superficial (*desconocida*) con simetría en torno al eje $z$. La solución es la que se obtiene a través de los polinomios de Legendre.

$$
\Phi = \sum_{l=0}^{\infty} \left( A_l r^l + \frac{B_l}{r^{l+1}} \right) P_l(\cos\theta)
$$

Con coeficientes distintos tanto dentro como fuera de la esfera. Consideremos primero el potencial fuera de la esfera.

* En muchos problemas que se resuelven a través de los polinomios de Legendre utilizamos que $\Phi(r{\rightarrow}\infty)=0$. Sin embargo, en este problema, esto no se puede utilizar: La condición es que $\vec{E_f}(r{\rightarrow}\infty)=E_0\vec{u_z}$. En este caso, vamos a ver que se deduce con el primer término de la serie de polinomios de Legendre.

$$
\Phi_1 = \left( A_1 r + \frac{B_1}{r^2} \right) \cos\theta \quad\Rightarrow\quad \vec{E} = -\left( A_1 - 2\frac{B_1}{r^3} \right) \cos\theta\vec{u_r} - \frac{1}{r}\left( A_1 r + \frac{B_1}{r^2} \right) (-\sin\theta)\vec{u_\theta}
$$

Cuando $r\rightarrow\infty$:

$$
\vec{E}(r{\rightarrow}\infty) = -A_1\left(\cos\theta\vec{u_r} - \sin\theta\vec{u_\theta}\right) = -A_1 \vec{u_z}
$$

Es decir, $A_1 = -E_0$, y todos los demás términos $A_l=0$ con $l\neq 1$.

* Por otro lado, dentro de la esfera solemos utilizar que el potencial no diverge y $B_l=0$. Esto sigue siendo válido en este caso.

Vamos a probar con los términos de orden 1 y veremos que es suficiente para resolver el problema.

$$
\Phi_d &= A_{d,1} r \cos\theta \\
\Phi_f &= (-E_0 r + \frac{B_{f,1}}{r^2})\cos\theta \\
\vec{E_d} &= A_{d,1} (\cos\theta\vec{u_r} - \sin\theta\vec{u_\theta}) \\
\vec{E_f} &= -E_0 \vec{u_z} - \frac{B_{f,1}}{r^3} \left( 2 \cos\theta \vec{u_r} + \sin\theta \vec{u_\theta} \right)
$$

Las tres condiciones de contorno son: Continuidad del potencial $\Phi$, Continuidad del vector desplazamiento $\vec{D}$ normal y Continuidad del campo eléctrico $\vec{E}$ tangencial.

$$
\Phi_d(r{=R}) &= \Phi_f(r{=R}) \\
\vec{D}_{d,r}(r{=R}) &= \vec{D}_{f,r}(r{=R}) \\
\vec{E}_{d,\theta}(r{=R}) &= \vec{E}_{d,\theta}(r{=R})
$$

La primera y la tercera imponen la misma condición:

$$
A_{d,1} = -E_0 + \frac{B_{f,1}}{R^3}
$$

La segunda, la del vector desplazamiento $\vec{D}$ normal:

$$
\varepsilon \left( A_{d,1} \right) = \varepsilon_0 \left( -E_0 - 2 \frac{B_{f,1}}{R^3} \right)
$$

Con el resultado para los coeficientes, utilizando la permitividad eléctrica relativa $\varepsilon_r = \frac{\varepsilon}{\varepsilon_0}$:

$$
A_{d,1} &= -\frac{3}{2 + \varepsilon_r} E_0 \\
B_{f,1} &= -\frac{1 - \varepsilon_r}{2 + \varepsilon_r} E_0 R^3
$$

Que cumple todas las condiciones de contorno en la frontera y en el infinito. Si intentamos considerar los términos de orden superior, de los polinomios de Legendre no se pueden cumplir las tres condiciones de contorno salvo que los dos coeficientes sean nulos, $A_{d,l} = B_{f,l} = 0$

