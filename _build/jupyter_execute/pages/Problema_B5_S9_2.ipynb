# Problema B5.S9.2

En este problema tenemos un cilindro de radio $a$ de material lineal por el que se hace pasar una corriente $I$ uniformemente distribuida a los largo del hilo en dirección del eje de simetría del cilindro, el eje $z$. La susceptibilidad magnética del material es $\chi_M$. La densidad de corriente $\vec{J}$ tiene la dirección $z$ y dado que la corriente está uniformemente distribuida, su valor es constante:

$$
\vec{J} = J\vec{u_z} = \frac{I}{\pi a^2}\vec{u_z}
$$


```{figure} ../images/B5_S9_2_1.png
:alt: J_cte
:width: 250px
```

```{note}
De forma similar a lo que nos encontrábamos en los ejercicios de Electrostática de materiales, en los problemas de Magnetostática de materiales, vamos a tener tres posibilidades, una de las cuales es particular de los problemas de magnetostática.

1. **Materiales lineales**: Se caracterizan por que la magnetización $\vec{M}$ es lineal con respecto al vector $\vec{H}$, típicamente tan solo una constante (*aunque podría ser una matriz, o tensor de orden 2*).

$$
\vec{M} = \chi_M \vec{H}
$$

2. **Materiales con magnetización constante**: Se caracterizan por que la magnetización es independiente del vector $\vec{H}$ externo. En este caso, las densidades de corriente de magnetización también son constantes, por lo que podemos considerarlas como fuentes de campo magnético $\vec{B}$ y sustituir el material por dichas corrientes.

$$
\vec{M} = cte \quad \Rightarrow \quad \vec{J_M} = \nabla\times\vec{M}; \space \vec{j_M} = \vec{M}\times\vec{n}
$$

Ambos casos, **1** y **2**, son equivalentes a problemas de Electrostática de materiales. El siguiente caso, **3**, aunque nos lo podríamos encontrar en Electrostática, allí tiene muy poca importancia, mientras que en Magnetostática tiene muchas aplicaciones prácticas.

3. **Materiales no lineales**: Se caracterizan por que la magnetización $\vec{M}$ responde al vector $\vec{H}$ externo de forma no lineal. Un ejemplo importante son los materiales con histéresis. Típicamente, le damos la vuelta a la dependencia para que $\vec{H}$ sea la variable independiente y $\vec{B}$ sea la variable dependiente:

$$
\vec{M} = f(\vec{H}) \quad \Rightarrow \quad \vec{B} = \mu_0(\vec{M} + f(\vec{H})) = g(\vec{H}) \quad \Rightarrow \quad \vec{H} = \vec{H}(\vec{B})
$$
```

Dado que el problema tiene una de las simetrías convenientes para aplicar Ampere, lo utilizamos. El circuito de Ampere que conviene tiene que encerrar una superficie atravesada por corriente y tiene que tener su frontera en la linea del campo magnético $\vec{B}$ y del vector $\vec{H}$, que conocemos por la regla de la mano derecha. La única posibilidad es un circulo perpendicular al eje $z$ centrado en dicho eje de radio arbitrario $s$.

```{figure} ../images/B5_S9_2_2.png
:alt: J_cte
:width: 250px
```

Distinguimos dos casos:

* $s{\le}a$: La corriente encerrada es la fracción de la corriente que atraviesa el circuito.

$$
\oint\vec{H}\cdot\vec{dl} = I_{enc} \quad \rightarrow \quad H(s)2\pi s = J\pi s^2
$$

O lo que es lo mismo:

$$
\vec{H}(s) = \frac{Js}{2}\vec{u_\phi} = \frac{Is}{2\pi a^2}\vec{u_\phi}
$$

* $s{>}a$: La corriente encerrada es la total, $I$:

$$
\oint\vec{H}\cdot\vec{dl} = I_{enc} \quad \rightarrow \quad H(s)2\pi s = I
$$

O lo que es lo mismo:

$$
\vec{H}(s) = \frac{I}{2\pi s}\vec{u_\phi} 
$$

Claramente el vector $\vec{H}$ es tangente y continuo en la frontera. Con esto ya podemos resolver el resto del problema.



### Apartado $a)$

El campo magnético fuera del hilo, donde no existe material ni magnetización, la relación entre el vector $\vec{H}$ y el campo magnético $\vec{B}$ es:

$$
\vec{B} = \mu_0\vec{H} = \frac{\mu_0 I}{2\pi s}\vec{u_\phi} \quad s>a
$$

Que coincide con el campo magnético generado por un hilo de corriente en el eje $z$. Dentro del material tenemos una magnetización $\vec{M}$ relacionada mediante la susceptibilidad $\chi_M$

$$
\vec{M} = \chi_M\vec{H} = \frac{\chi_MIs}{2\pi a^2}\vec{u_\phi} \quad s{\le}a
$$

De tal forma que el campo magnético $\vec{B}$ es:

$$
\vec{B} = \mu_0(\vec{H} + \vec{M}) = \frac{\mu_0(1+\chi_M)Is}{2\pi a^2}\vec{u_\phi} = \frac{\mu Is}{2\pi a^2}\vec{u_\phi} \quad s{\le}a
$$




### Apartado $b)$

Las densidades de corriente de magnetización se pueden obtener a partir de su definición:

$$
\vec{J_M} &= \nabla\times\vec{M} \\
\vec{j_M} &= \vec{M}\times\vec{n}
$$

En el rotacional, dado que $\vec{M} = M_\phi(s)\vec{u_\phi}$ que solo tiene dirección $\phi$ y ademas dicha componente solo depende de $s$, la coordenada radial de cilíndricas, el único término que sobrevive es (*Importante verificar la expresión: el producto $sM_\phi$ se hace antes de la derivada parcial*):

$$
\vec{J_M} = \nabla\times\vec{M} = \frac{1}{s} \left( \frac{\partial(sM_\phi)}{\partial s} \right) \vec{u_z} = \frac{\chi_MI}{\pi a^2}\vec{u_z}
$$

El vector unitario saliente $\vec{n}$ es justo el vector unitario radial de cilíndricas, $\vec{n}=\vec{u_s}$ (*Importante: $s$ es arbitrario, en la superficie $s$ se particulariza a $s=a$*):

$$
\vec{j_M} = \vec{M}\times\vec{n} = \left(\frac{\chi_MIs}{2\pi a^2}\vec{u_\phi}\times\vec{u_s}\right)_{s{=}a} = -\frac{\chi_MI}{2\pi a} \vec{u_z}
$$

* Observemos que el campo magnético $\vec{B}$ no es continuo en la frontera del material, y el salto corresponde justo con la densidad superficial de corriente de magnetización $\vec{j_M}$

$$
\left(B_f - B_d\right)_{s{=}a} = \mu_0 j_M
$$



### Apartado $c)$

Vamos a realizar ahora la comprobación sobre las densidades de corriente de magnetización: La corriente neta en el material es nula. Dado que tiene simetría a lo largo del eje $z$, si comprobamos que la corriente de magnetización neta que atraviesa una superficie $z=cte$, habremos comprobado que la corriente de magnetización total en el cilindro también es nula. Particularizando en un plano  cualquiera, la corriente total es la suma de la volumétrica más la superficial:

$$
I(z) = \int_S \vec{J_M}\cdot\vec{dS} + \oint_C j_M dl
$$

Lo corriente volumétrica total de magnetización es, dado que la densidad de corriente de magnetización es constante:

$$
\int_S \vec{J_M}\cdot\vec{dS} = J_MS = \chi_MI
$$

Mientras que la corriente superficial que atraviesa la frontera hacia arriba es:

$$
\oint_C j_M dl = j_M2\pi a = -\chi_MI
$$

Dando un valor total nulo.
