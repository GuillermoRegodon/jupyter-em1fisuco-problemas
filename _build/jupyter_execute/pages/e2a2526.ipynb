
Sabemos que la imagen de una carga puntual frente a una esfera de CEP conectada a tierra es**:

```{figure} ../images/B2_S3_3_2.png
:alt: q_puntual
:width: 300px
Imagen de una carga puntual frente a una esfera de CEP.
```

$$
q' &= -q\frac{a}{d}\\
d' &= \frac{a^2}{d}\\
\vec{E_{q'}} &= \frac{q'}{4 \pi \varepsilon_0 |\vec{r} - \vec{r_{q'}}|^3}(\vec{r} - \vec{r_{q'}})
$$

Es importante, respecto a esta configuración, recordar que el potencial en la superficie del CEP es $\Phi = 0$, pero en el enunciado e dice que el potencial es $\Phi = V$. Para conseguir dicho potencial, usamos que el **Principio de superposición** y añadimos una carga $Q_c$ en el centro de la esfera. **Es el único punto donde podemos poner una carga de forma que el potencial en la esfera siga siendo constante**:

```{figure} ../images/q_central.png
:alt: q_central
:width: 300px
Imagen de una carga puntual frente a una esfera de CEP con carga central.
```

$$
\vec{E_Q} = \frac{Q}{4 \pi \varepsilon_0 |\vec{r} - \vec{r_Q}|^3}(\vec{r} - \vec{r_Q})
$$

**Para obtener el valor de $Q$ usamos la condición del enunciado, $\Phi = V$**. Observar que la esfera está conectada a una fuente, por lo que la fuente le suministra la carga que necesite para cumplir las condiciones del problema.

$$
\vec{\Phi_Q} = \frac{Q}{4 \pi \varepsilon_0 |\vec{r} - \vec{r_Q}|} = V
$$

Es decir, dado que $|\vec{r} - \vec{r_Q}|=a$ el radio de la esfera,

$$
Q = 4\pi\varepsilon_0aV
$$

Falta poner el sistema de referencia. Si ponemos todas las cargas en el mismo eje $x$, y la esfera centrada en el origen:

$$
\vec{r_Q} &= 0\\
\vec{r_q} &= d \vec{u_x}\\
\vec{r_{q'}} &= d' \vec{u_x}
$$

Con este planteamiento, podemos resolver la fuerza sobre la carga $q$ sera la fuerza que generan las demás cargas de la configuración, las cargas $q'$ y $Q$.

$$
\vec{F_q} = q \left( \frac{q'}{4 \pi \varepsilon_0 (d-d')^2} + \frac{Q}{4 \pi \varepsilon_0 d^2} \right) \vec{u_x}
$$



```{note}
2. Un hilo de carga de densidad lineal de carga constante $\lambda$ se coloca perpendicularmente a un plano infinito de CEP conectado a tierra. Calcule los momentos multipolares de orden 0 y 1 de la configuración (Carga total y momento dipolar) (5 puntos).
```

El enunciado no indica la distancia a la que está el hilo (*por error de omisión*) por lo que elegimos nombrar $a$ como la distancia con respecto al plano CEP a la que está el punto más cercano y $b$ la distancia a la que está el punto más lejano, de tal forma que la longitud del hilo de densidad de carga constante $\lambda$ es $l=b-a$.

```{figure} ../images/e2a2526_2.png
:alt: e2a2526_2.png
:width: 300px
```

Una distribución de carga, en este caso el hilo de carga lineal constante $\lambda$, genera un campo eléctrico en el espacio tridimensional. Por otra parte, la condición de CEP conectado a tierra implica varias cosas: En particular, el potencial es nulo y, como consecuencia, el campo eléctrico es perpendicular a la superficie del CEP. El campo eléctrico, por la condición de frontera, implica una carga superficial en la superficie del CEP, necesaria para producir el campo eléctrico y el potencial que compensan los de la distribución original y conseguir las condiciones de CEP a tierra (*repetimos, potencial nulo y campo eléctrico perpendicular*). Por lo tanto, como resumen, **un CEP tiene una carga superficial que consigue las condiciones de CEP conectado a tierra**.

Sabemos que por el método de las imágenes, para algunas configuraciones de CEP, la densidad superficial de carga es equivalente a una imagen. En el caso de un CEP plano infinito, la imagen es una configuración reflejada con la carga opuesta. En nuestro caso, **la imagen es un hilo de densidad lineal de carga constante $-\lambda$** de las mismas dimensiones. **Es equivalente en el sentido de que el campo eléctrico y el potencial es el mismo** aunque destacamos en este punto que esta imagen no existe realmente.

Pasemos a calcular los momentos multipolares. Para la carga total $Q_t$ tenemos que sumar la carga del hilo más la carga en el CEP:

$$
Q_t = \int_{hilo} \lambda dl + \int_{CEP} \sigma dS
$$

De los varios ejemplos que hemos hecho, sabemos que podemos demostrar usando la ley de Gauss, que la carga total acumulada en la superficie del CEP es igual a la carga total de las imágenes en su interior, por lo que (*elegimos el SR de tal forma que los hilos están en el eje $x$ y $x=0$ corresponde con el plano de CEP*):

$$
Q_t = \int_{hilo} \lambda dl + \int_{imagen} \lambda' dl = \int_a^b \lambda dx + \int_{-b}^{-a} (-\lambda) dx = 0
$$

Para el momento dipolar, la expresión es la siguiente:

$$
\vec{p} = \int_{hilo} \vec{r} \lambda dl + \int_{CEP} \vec{r} \sigma dS
$$

Con el mismo SR:

$$
\vec{p} = \int_a^b x\vec{u_x} \lambda dx + \int_{y=-\infty}^{y=+\infty} \int_{z=-\infty}^{z=+\infty} (0, y, z) \sigma \space dy \space dz
$$

La primera integral es:

$$
\int_a^b x\vec{u_x} \lambda dx = \lambda \frac{x^2}{2}|_a^b = \frac{\lambda}{2}(b^2-a^2)
$$

La segunda integral requeriría conocer la densidad superficial de carga, pero sabiendo que es simétrica con respecto a $y=0$ y $z=0$, tenemos que las dos coordenadas son el producto de una integral de una función par por otra impar, en un dominio simétrico, por lo que la integral es 0, por ejemplo, la componente $y$

$$
\int_{y=-\infty}^{y=+\infty} \int_{z=-\infty}^{z=+\infty} y \sigma(y,z) \space dy \space dz = \int_{y=-\infty}^{y=+\infty} y \int_{z=-\infty}^{z=+\infty} \sigma(y,z) \space dy \space dz = \int_{y=-\infty}^{y=+\infty} y F(y) dy = 0
$$

Igual para la coordenada $z$. También se puede demostrar en polares 2D (*de $(y,z)$ a $(r,\theta)$*):

$$
\int_{CEP} \vec{r} \sigma dS = \int_{\theta=0}^{\theta=2\pi} \int_{r=0}^{r=\infty} (0, r\cos\theta, r\sin\theta) r \space dr \space d\theta = 0
$$

Por lo que:

$$
\vec{p} = \frac{\lambda}{2}(b^2-a^2)
$$


```{note}
3. Un cilindro infinito de material magnético con una permeabilidad magnética $\mu$ tiene un arrollamiento de hilos por los que circula una corriente $I$. En un segmento de longitud $L$ hay $N$ hilos. Calcule:
* A. Los campos $\vec{H}$, $\vec{B}$ y $\vec{M}$ (5 puntos).
* B. Calcule la reluctancia de un trozo de longitud $L$ de material ($R$, en unidades de $\frac{A}{Tm^2}$ (1 puntos).

```

**Este problema tiene simetría cilíndrica de rotación y de traslación en la dirección del eje**. Es decir, si rotamos la configuración un ángulo $\phi$ cualquiera o si desplazamos la configuración en la dirección $z$, la configuración queda igual. Además, teniendo un medio material distinto en el interior y en el exterior, lo conveniente es usar la ley de Ampere sobre $\vec{H}$, que es independiente del material.

$$
\int_C \vec{H} \cdot \vec{dl} = I_{enc}
$$

Debemos elegir un circuito de Ampere que nos permita resolver el problema. Vemos que por la regla de la mano derecha, el campo magnético debe tener la dirección $\vec{u_z}$.

```{note}
El circuito de Ampere debe elegirse de forma que **la superficie de dicho circuito de Ampere tenga lados paralelos al campo magnético**. Los lados restantes, si los hay, deberían ser perpendiculares al campo magnético. Para ello, debemos adivinar **la dirección del campo magnético usando la regla de la mano derecha**. Por otra parte, **el circuito de Ampere debe estar atravesado por la corriente que genera el campo**.
```

En este caso, un trozo del circuito de Ampere debe tener la dirección $\vec{u_z}$ por el interior del arrollamiento, para que sea colineal con el campo magnético. Ponemos otro por el exterior, para poder comparar el campo magnético interior y exterior y los dos lados restantes, perpendiculares a la dirección $\vec{u_z}$. De esta forma, el circuito de Ampere será un marco rectangular de altura $L$ con un lado dentro del arrollamiento y otro lado fuera, de forma que la corriente que atraviesa la superficie limitada por el circuito de Ampere es $I_{enc} = N I$.

Observamos que **la posición de los lados es indiferente** y que siempre le atraviesa la misma corriente. De esta manera deducimos que **el campo magnético tanto dentro como fuera del arrollamiento debe ser constante**. Y además, el campo exterior debe ser nulo, ya que si elegimos el circuito de Ampere cada vez mas extenso, hasta que llegue al infinito, el resultado de la integral no puede cambiar, y en particular esto implica que un lado tendrá el valor de $\vec{H}$ del infinito. Por lo tanto, la integral de Ampere resulta:

$$
H L = N I
$$

Con lo que podemos despejar $H$ y ponerle la dirección $\vec{u_z}$. Dentro del solenoide:

$$
\vec{H} = \frac{N I}{L}\vec{u_z}
$$

Mientras que fuera vale $\vec{H} = 0$. Para obtener $\vec{B}$ tenemos que multiplicar por $\mu$:

$$
\vec{B} = \mu\frac{N I}{L}\vec{u_z}
$$

Y para obtener $\vec{M}$, podemos usar la expresión que los relaciona:

$$
\vec{B} = \mu_0 (\vec{H} + \vec{M})
$$

Con lo que

$$
\vec{M} = \left(\frac{\mu}{\mu_0} -1 \right) \frac{N I}{L}\vec{u_z}
$$

La reluctancia $R$ de un trozo de longitud $L$ se puede obtener de su expresión:

$$
R = \frac{L}{\mu S}
$$

Donde $S$ es la sección del trozo de material, $S=\pi a^2$. Si no recordamos la expresión, vemos que por las unidades, $R$ relaciona una corriente (*en $A$, tiene que ser la fuerza magnetomotriz actuando sobre dicho trozo, o $fmm$*) con un flujo magnético (*$\Phi$ en $T m^2$*), es decir:

$$
R = \frac{fmm}{\Phi} = \frac{NI}{B S} = \frac{NI}{\mu\frac{N I}{L} \pi a^2} = \frac{L}{\mu \pi a^2}
$$


````{note}
4. En una red cristalina, un dipolo magnético $\vec{m_z}$ apunta en la dirección $z$, mientras que otro dipolo $\vec{m_x}$ en una posición relativa $\vec{r} = (d,0,d)$, apunta en la dirección $x$. Calcule la fuerza y el torque que sufre cada dipolo por el efecto del campo magnético del otro (4 puntos).
```{figure} ../images/e2a2526_3.png
:alt: e2a2526_3.png
:width: 250px
```
````

Este problema es aplicación directa de las expresiones del formulario, pero es importante tener cuidado con el sistema de coordenadas, con las derivadas en el gradiente, con qué es constante y que no... Para empezar, tenemos dos expresiones del campo magnético generado por un dipolo magnético:

$$
\vec{B}(\vec{r}) = \frac{\mu_0}{4\pi} \left( \frac{3(\vec{m}\cdot\vec{r})\vec{r}}{r^5} - \frac{\vec{m}}{r^3} \right)
$$

y

$$
\vec{B}(\vec{r}) = \frac{\mu_0 m}{4\pi r^3}\left( 2\cos\theta\vec{u_r} - \sin\theta\vec{u_\theta} \right)
$$

La primera expresión tiene vectores, por lo que es independiente del sistema de referencia, mientras que la segunda es válida en coordenadas esféricas con el dipolo apuntando en la dirección $z$, como es el caso del dibujo. Es decir, la segunda expresión se podría usar directamente para realizar los cálculos de fuerza y torque del dipolo $\vec{m_z}$ sobre el dipolo $\vec{m_x}$, pero requiere modificación del SR para usarlo al revés. Es obvio que la segunda expresión se deduce de la primera, ya que es una particularización.

Con respecto a las coordenadas, podemos usar esféricas o cartesianas. Las cartesianas solo se pueden usar con la primera expresión para $\vec{B}$:

$$
\vec{m} &= \vec{m_z} = m_z \vec{u_z} \\
\vec{r} &= (x,y,z) \\
r &= \sqrt{x^2 + y^2 + z^2}
$$

es decir...

$$
\vec{B_z}(\vec{r}) = \frac{\mu_0}{4\pi} \left( \frac{3 m_z z (x,y,z)}{(x^2 + y^2 + z^2)^{5/2}} - \frac{m_z \vec{u_z}}{(x^2 + y^2 + z^2)^{3/2}} \right)
$$

(*De hecho, con la primera expresión también se pueden usar esféricas, dando expresiones muy parecidas a usar directamente la segunda expresión*) Vamos a utilizar este campo magnético para las expresiones de la fuerza y del torque. Vamos a interpretarlas:

$$
\vec{F} &= \nabla(\vec{m}\cdot\vec{B}) \\
\vec{\tau} &= \vec{m}\times\vec{B}
$$

En estas expresiones, $\vec{m}$ es el momento dipolar del dipolo que sufre la fuerza, y $\vec{B}$ es el campo magnético en la posición de dicho dipolo. Es decir, si lo aplicamos a este caso, dado que hemos obtenido $\vec{B_z}$, podemos obtener la fuerza y el torque que dicho dipolo $\vec{m_z}$ ejerce sobre el otro dipolo $\vec{m_x}$.

$$
\vec{F}_{\vec{m_z}\rightarrow \vec{m_x}} &= \nabla(\vec{m_x}\cdot\vec{B_z}) \\
\vec{\tau}_{\vec{m_z}\rightarrow \vec{m_x}} &= \vec{m_x}\times\vec{B_z}
$$

Vamos a empezar sustituyendo el torque, que no tiene operador $\nabla$:

$$
\vec{m_x} &= m_x\vec{u_x} \\
\vec{r} &= (d,0,d) \\
r &= \sqrt(2)d
$$

Con lo que:

$$
\vec{\tau}_{\vec{m_z}\rightarrow \vec{m_x}} = \vec{m_x}\times\vec{B_z} = m_x\vec{u_x} \times \left( \frac{\mu_0}{4\pi} \left( \frac{3 m_z d (d,0,d)}{d2^{5/2}} - \frac{m_z \vec{u_z}}{d2^{3/2}} \right) \right)
$$

Como $\vec{u_x}\times\vec{u_x} = 0$, solo nos queda componente $\vec{u_y}$. Sustituyendo y simplificando:

$$
\vec{\tau}_{\vec{m_z}\rightarrow \vec{m_x}} = -\frac{\mu_0m_zm_x}{16\pi\sqrt{2}d^3}\vec{u_y}
$$

A continuación, obtenemos la fuerza. **¡Veamos que involucra un gradiente!** Esto significa que necesitamos conocer el producto escalar $\vec{m_x}\cdot\vec{B_z}$ en los puntos en torno al punto deseado, hacer el gradiente, y sustituir **después** de hacer el gradiente. El producto escalar:

$$
\vec{m_x}\cdot\vec{B_z} = m_x\vec{u_x} \cdot \left( \frac{\mu_0}{4\pi} \left( \frac{3 m_z z (x,y,z)}{(x^2 + y^2 + z^2)^{5/2}} - \frac{m_z \vec{u_z}}{(x^2 + y^2 + z^2)^{3/2}} \right) \right) = \frac{\mu_0 3 m_z m_x zx}{4\pi(x^2 + y^2 + z^2)^{5/2}}
$$

A esta expresión le podemos obtener el gradiente en cartesianas. Por último, sustituimos en el punto después de hacer el gradiente, que lo que queda:

$$
\vec{F}_{\vec{m_z}\rightarrow \vec{m_x}} = \nabla(\vec{m_x}\cdot\vec{B_z})|_{(x,y,z)=(d,0,d)} = -\frac{9\mu_0m_zm_x}{8\sqrt{2}\pi d^4} (\vec{u_x}+\vec{u_z})
$$

La fuerza y el torque que ejerce el segundo dipolo $\vec{m_x}$ sobre el primero $\vec{m_z}$ se puede calcular cuidadosamente de la misma forma, pero como era de esperar, por el principio de acción-reacción, da los mismos valores con el signo cambiado:

$$
\vec{F}_{\vec{m_x}\rightarrow \vec{m_z}} &= -\vec{F}_{\vec{m_z}\rightarrow \vec{m_x}} \\
\vec{\tau}_{\vec{m_x}\rightarrow \vec{m_z}} &= -\vec{\tau}_{\vec{m_z}\rightarrow \vec{m_x}}
$$

El problema también se puede resolver el esféricas, ya que tenemos el campo magnético en esféricas:

$$
\vec{\tau}_{\vec{m_z}\rightarrow \vec{m_x}} = \vec{m_x} \times \vec{B_z} = \left( m_x(\sin\theta \cos\phi \vec{u_r} + \cos\theta \cos\phi \vec{u_\phi} + \sin\theta\vec{u_\theta} \right) \times \left( \frac{\mu_0 m}{4\pi r^3}\left( 2\cos\theta\vec{u_r} - \sin\theta\vec{u_\theta} \right) \right)
$$

En este caso, tenemos que sustituir la posición del segundo dipolo,

$$
r &= \sqrt{2}d \\
\theta &= \frac{\pi}{4} \\
\phi &= 0
$$

Dando el mismo resultado. Para la fuerza, tenemos que realizar el producto escalar:

$$
\vec{m_x} \cdot \vec{B_z} = \left( m_x(\sin\theta \cos\phi \vec{u_r} + \cos\theta \cos\phi \vec{u_\phi} + \sin\theta\vec{u_\theta} \right) \cdot \left( \frac{\mu_0 m}{4\pi r^3}\left( 2\cos\theta\vec{u_r} - \sin\theta\vec{u_\theta} \right) \right)
$$

A este producto escalar, obtenerle el gradiente en esféricas, y **después**, sustituir la posición del segundo dipolo, dando, por supuesto el mismo resultado. Dado que la expresión en esféricas solo es válida para $\vec{m_z}$ centrado en el origen, no se puede usar tal cual para calcular los valores de la fuerza y del torque que generan el segundo dipolo $\vec{m_x}$ sobre el primero $\vec{m_z}$. Habría que cambiar el sistema de referencia para centrarlo en $\vec{m_x}$. Por supuesto, el principio de acción-reacción sigue siendo valido en esféricas y se puede usar para no tener que calcularlo.
