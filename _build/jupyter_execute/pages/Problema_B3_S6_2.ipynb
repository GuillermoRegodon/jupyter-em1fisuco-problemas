# Problema B3.S6.2

El problema pide resolver la capacidad y la energía electrostática en una configuración esférica con un dieléctrico en el espacio intermedio entre los dos electrodos (metálicos, de conductor eléctrico perfecto o *CEP*). Las relaciones entre el campo eléctrico $\vec{E}$, el vector polarización $\vec{P}$ y el vector desplazamiento $\vec{D}$ son:

$$
\vec{D} &= \varepsilon \vec{E} \\
\vec{P} &= \vec{D} - \varepsilon_0 \vec{E} = (\varepsilon - \varepsilon_0) \vec{E}
$$

Y la condición de contorno que nos impone la conexión del condensador es $\Phi(b) = 0$ y $\Phi(a) = V$. Este problema es de un condensador esférica con permitividad eléctrica relativa variable. Sin embargo, la dependencia radial de la permitividad relativa mantiene la simetría esférica, por lo que podemos utilizar el método de Gauss para resolverlo. Utilizamos la ecuación de Gauss en forma integral para medios materiales dieléctricos con simetría esférica, eligiendo una superficie de Gauss esférica centrada en el origen con un radio $r$, lo que nos da, para la componente radial (*y única componente*) del vector desplazamiento $\vec{D}$:

$$
\oint_S \vec{D}\cdot\vec{S} = Q_l \quad \Rightarrow \quad D4\pi r^2 = Q_l
$$

Siendo $Q_l$ la carga libre total en el conductor interno. Vamos a escribir también las expresiones para el campo eléctrico $\vec{E}$ y para el vector polarización $\vec{P}$.

$$
\vec{D} &= \frac{Q_l}{4\pi r^2}\vec{u_r}; \quad \vec{E} = \frac{Q_l}{4\pi \varepsilon r^2}\vec{u_r} = \frac{Q_l(1-kr)}{4\pi \varepsilon_0 r^2}\vec{u_r}; \quad \vec{P} = \frac{kQ_l}{4\pi r}\vec{u_r} \quad r\in[a,b]\\
$$

Para poder relacionar la carga libre $Q_l$ con el potencial $V$ utilizamos la definición de potencial en forma integral:

$$
V = \Phi(r{=}a) - \Phi(r{=}b) = -\int\vec{E}\cdot\vec{dl}
$$

Eligiendo un radio, $\vec{dl} = dr\vec{u_r}$

$$
V = -\int\vec{E}\cdot\vec{dl} = - \int_{r=b}^{r=a} E dr = - \int_{r=b}^{r=a} \frac{Q_l(1-kr)}{4\pi \varepsilon_0 r^2} dr = \frac{Q_l}{4\pi \varepsilon_0 r}|_{r=b}^{r=a} + \frac{kQ_l}{4\pi \varepsilon_0}\log(r)|_{r=b}^{r=a} = \frac{Q_l}{4\pi \varepsilon_0} \left( \frac{1}{a} - \frac{1}{b} \right) + \frac{kQ_l}{4\pi \varepsilon_0} \log\frac{a}{b}
$$

Con lo que la capacidad y la energía electrostática son inmediatas. Por ejemplo, para la capacidad:

$$
C = \frac{Q_l}{V} = \frac{1}{\frac{1}{4\pi \varepsilon_0} \left( \frac{1}{a} - \frac{1}{b} \right) + \frac{k}{4\pi \varepsilon_0} \log\frac{a}{b}}
$$


!Observamos que la polarización no tiene una divergencia nula! Esto significa que la densidad volumétrica de carga de polarización tampoco es nula. Usando la divergencia en esféricas:

$$
\rho_P = -\nabla\cdot\vec{P} = -\frac{1}{r^2}\frac{\partial}{\partial r}(r^2P_r) = -\frac{kQ_l}{4\pi r^2}
$$

Las densidades superficiales de carga de polarización se pueden obtener, teniendo cuidado con la dirección del vector unitario $\vec{n}$. En la frontera del dieléctrico en $r=a$:

$$
\vec{n} = -\vec{u_r} \quad \Rightarrow \quad \sigma_a = \vec{n}\cdot\vec{P} = -\vec{u_r}\cdot\vec{P}(r{=}a) = -\frac{kQ_l}{4\pi a}
$$

Mientras que en la frontera del dieléctrico en $r=b$:

$$
\vec{n} = \vec{u_r} \quad \Rightarrow \quad \sigma_b = \vec{n}\cdot\vec{P} = \vec{u_r}\cdot\vec{P}(r{=}b) = \frac{kQ_l}{4\pi b}
$$

Podemos verificar que la suma total de la carga de polarización del dieléctrico es nula:

$$
Q_T = \int_{S_a} \sigma_a dS_a + \int_{S_b} \sigma_b dS_b + \int_V \rho_P dV = 0
$$

La integral en la esfera de radio a, $S_a$:

$$
\int_{S_a} \sigma_a dS_a = \int_{\theta{=}0}^{\theta{=}\pi}\int_{\phi{=}0}^{\phi{=}2\pi}\left(-\frac{kQ_l}{4\pi a}\right)a^2\sin\theta d\theta d\phi = -kQ_la
$$

Similarmente:

$$
\int_{S_b} \sigma_b dS_b = kQ_lb
$$

Y finalmente:

$$
\int_V \rho_P dV = \int_{r{=}b}^{r{=}a}\int_{\theta{=}0}^{\theta{=}\pi}\int_{\phi{=}0}^{\phi{=}2\pi} \left(\frac{kQ_l}{4\pi r^2}\right)r^2\sin\theta dr d\theta d\phi = kQ_l(a-b)
$$

Con lo que comprobamos el resultado de teoría, $Q_T = 0$.
