# Problema B4.S8.6

Este problema nos pide calcular la fuerza que ejerce una configuración de corriente sobre otra. Para ello, **vamos a parametrizar matemáticamente la fuente de campo magnético y los puntos en los que tenemos que evaluar dicho campo campo magnético**. Colocamos el hilo de corriente, que es la fuente, en el eje $z$, de tal forma que el hilo se parametriza mediante la coordenada $z'$, a la que le ponemos la prima $'$ por que es la fuente de campo magnético. La corriente es $I$ y el diferencial de linea $\vec{dl'} = dz' \vec{u_z}$. 

$$
\vec{r'} &= (0,0,z') \\
\vec{dl'} &= dz' \vec{u_z} \quad z' \in[-\infty, \infty]
$$

```{figure} ../images/B4_S8_6_1.png
:alt: Espiras
:width: 300px
```

Esta configuración tiene simetría cilíndrica que nos permite usar la ley de Ampere para resolver el campo magnético fácilmente en cualquier punto $\vec{r} = \rho\vec{u_\rho}$ del espacio (*La ley de Ampere siempre es válida, aunque no siempre simplifica la resolución*).

$$
\vec{B_1} = \frac{\mu_0 I }{2\pi \rho}\vec{u_\phi}
$$

**Dado que el enunciado utilizada la variable $s$ como la distancia entre el hilo y el punto más cercano de la espira, utilizamos $\rho$ como variable radial**. Con esto, la fuerza sobre otra configuración de corriente sigue la expresión:

$$
\vec{F_{12}} = \int_C I_2 \vec{dl_2}\times\vec{B_1}
$$

Aún no hemos parametrizado la configuración de corriente sobre la que aplicamos el campo magnético. Eso es porque está definida por ramas y tenemos que dividirla en cuatro partes. Dejamos el 1 para el hilo infinito, y usamos 2, 3, 4 y 5 para los cuatro trozos de la espira. En todos los tramos, la corriente es la misma, $I$.

* $\vec{F_{12}}$: Este tramos se parametriza con la coordenada radial. De esta forma, el diferencial de linea $\vec{dl} = d\rho\vec{u_\rho}$.

$$
\vec{F_{12}} = \int_{\rho=s+a}^{\rho=s} I (d\rho \vec{u_\rho})\times (\frac{\mu_0 I }{2\pi\rho}\vec{u_\phi}) = \frac{\mu_0 I^2}{2\pi} \int_{\rho=s+a}^{\rho=s} \frac{d\rho }{\rho}\vec{u_z} = - \frac{\mu_0 I^2}{2\pi} \log{\frac{s+a}{s}}\vec{u_z}
$$

(*Observemos que es más fácil pensar en el diferencial de línea en la dirección estándar de las coordenadas y poner los límites de integración para definir los puntos donde empieza y acaba la corriente $I$. Alternativamente, podríamos haber usado los límites normales y haberle dado la vuelta a la corriente, $-I$*).

* $\vec{F_{13}}$: Este tramo es igual que 2, con el signo cambiado a través de los límites de integración.

$$
\vec{F_{13}} = \int_{\rho=s}^{\rho=s+a} I (d\rho \vec{u_\rho})\times (\frac{\mu_0 I }{2\pi\rho}\vec{u_\phi}) = \frac{\mu_0 I^2}{2\pi} \log{\frac{s+a}{s}}\vec{u_z}
$$

* $\vec{F_{14}}$: Este tramo se parametriza con la coordenada $z$, es decir, $\vec{dl} = z\vec{u_z}. Igual que 2, jugamos con los límites de integración para indicar la dirección de la corriente.

$$
\vec{F_{14}} = \int_{z=a}^{z=0} I (dz \vec{u_z})\times (\frac{\mu_0 I }{2\pi s}\vec{u_\phi}) = \frac{\mu_0 I^2 a}{2\pi s}\vec{u_\rho}
$$

(*Observemos que todos los puntos del tramo están a la misma distancia radial $\rho=s$*)

* $\vec{F_{15}}$: Este tramo es igual que 4, con el signo cambiado a través de los límites de integración y teniendo en cuenta que el tramo de corriente está un poco más lejos, en $\rho=s+a$.

$$
\vec{F_{15}} = \int_{z=0}^{z=a} I (dz \vec{u_z})\times (\frac{\mu_0 I }{2\pi(s+a)}\vec{u_\phi}) = -\frac{\mu_0 I^2 a}{2\pi (s+a)}\vec{u_\rho}
$$

Con esto, la fuerza total sobre la espira es:

$$
\vec{F} = \vec{F_{12}} + \vec{F_{13}} + \vec{F_{14}} + \vec{F_{15}} = \frac{\mu_0 I^2 a^2}{2\pi s(s+a)}\vec{u_\rho}
$$
