# Problema B4.S8.2

Tenemos una serie de espiras suficientemente juntas como para considerar que la corriente superficial resultante es prácticamente homogénea, *una corriente superficial $\vec{j}$*.


### Apartado $a)$

Podemos resolver este problema de varias maneras. Matemáticamente, las operaciones son muy parecidas, y no ahorramos mucho esfuerzo, pero conceptualmente es interesante observar la relación. 

1. **Por una parte, podemos plantear la integral del campo magnético,** directamente:

$$
\vec{B}(\vec{r}) = \frac{\mu_0}{4\pi}\int_{S'}\frac{(\vec{j}(\vec{r'})dS')\times(\vec{r}-\vec{r'})}{|\vec{r}-\vec{r'}|^3}
$$

```{figure} ../images/B4_S8_2_1.png
:alt: cilindro_rot
:width: 200px
```

En este caso, el punto de observación es un punto del eje $z$, es decir, $\vec{r} = z\vec{u_z}$. Caracterizamos el cilindro:

$$
\vec{r'} &= a\vec{u_s} + z'\vec{u_z} \\
\vec{r} - \vec{r'} &= (z-z')\vec{u_z} - a\vec{u_s} \\
|\vec{r} - \vec{r'}| &= \sqrt{(z-z')^2 + a^2} \\
dS' &= a\space d\phi'\space dz' \quad z' \in [-\frac{L}{2},\frac{L}{2}] \quad \phi' \in [0,2\pi]
$$

Por último, la densidad superficial de corriente es paralela a la superficie, dándole vueltas al cilindro $\vec{j} = j \vec{u_\phi}$. Si en $L$ tenemos $N$ vueltas, y la corriente en cada hilo es $I$, la densidad superficial de corriente es $j = \frac{NI}{L}$. Observamos que las unidades de $[\vec{j}] = \frac{A}{m}$ nos dan una pista para obtener esta expresión: *Significa que en un trozo vertical de longitud $L$ va a cruzar una corriente total $jL$, que en nuestro caso, es $jL = NI$*.

```{note}
El punto de observación en cilíndricas siempre es $\vec{r'} = s'\vec{u_s} + z'\vec{u_z}$. En la mayoría de las integrales, $\vec{u_s}$ no es constante, por lo que en la integral no se puede manejar directamente. Tipicamente, lo que hacemos es pasarlo a cartesianas, de tal forma que el vector unitario sale de la integral (*Los unitarios en cartesianas sí son constantes*) y la integral se puede resolver.

Sin embargo, en magnetostática, es común que los productos vectoriales sean más fáciles de manejar **en polares, cilíndricas o esféricas**, por lo que **suele ser conveniente no realizar el cambio a cartesianas hasta después de los productos vectoriales**.
```

Con esto, la expresión de la integral es:

$$
\vec{B}(\vec{r}) = \frac{\mu_0}{4\pi}\int_{z'{=}-\frac{L}{2}}^{z'{=}\frac{L}{2}} \int_{\phi'{=}0}^{\phi'{=}2\pi}
\frac{(\frac{NI}{L}\vec{u_\phi} a\space d\phi'\space dz')\times((z-z')\vec{u_z} - a\vec{u_s})}{((z-z')^2 + a^2)^{3/2}}
$$

Por simetría, sabemos que el campo magnético solo puede tener dirección $\vec{u_z}$ en el eje $z$, por lo que el término $\vec{u_phi}\times\vec{u_z}$ podemos eliminarlo porque sabemos que dará 0 (*se puede comprobar matemáticamente*).

$$
\vec{B}(\vec{r}) = \frac{\mu_0}{4\pi}\int_{z'{=}-\frac{L}{2}}^{z'{=}\frac{L}{2}} \int_{\phi'{=}0}^{\phi'{=}2\pi}
\frac{(\frac{NI}{L} a\space d\phi'\space dz' a\vec{u_z})}{((z-z')^2 + a^2)^{3/2}}
$$

Si realizamos la integral en $\phi'$ que es directa:

$$
\vec{B}(\vec{r}) = \frac{\mu_0 NI}{2L}\int_{z'{=}-\frac{L}{2}}^{z'{=}\frac{L}{2}} 
\frac{(a^2\space \space dz' \vec{u_z})}{((z-z')^2 + a^2)^{3/2}}
$$




2. **Por otra parte, podemos plantear la integral del campo magnético a partir del campo magnético de una espira en su eje de simetría**. Veamos:

En este caso, dado que el cilindro **se puede dividir fácilmente en espiras diferenciales que podemos integrar**, vamos a calcular el campo magnético total a partir del campo magnético de una espira:

$$
\vec{B} = \int d \vec{B}
$$

El campo magnético de **una espira de corriente $I_1$ y de radio $a$ en el eje $\vec{u_z}$ de simetría** es conocido:

$$
\vec{B_1} = \frac{\mu_0 I_1}{2} \frac{a^2}{(a^2 + z^2)^{3/2}} \vec{u_z}
$$

Ahora vamos a utilizar este resultado para **obtener el campo magnético generado por un trozo de cilindro** cuya posición sea $z'$. Vamos a particularizar el resultado para la espira diferencial en $z'$.

```{figure} ../images/B4_S8_2_2.png
:alt: cilindro_rot
:width: 200px
```

* Por una parte, **la posición ha cambiado. La espira diferencial está en $z'$ mientras que la original esta en 0, por lo que tenemos que sustituir $z$** para tener en cuenta la posición de la espira diferencial:

$$
z \rightarrow z-z'
$$

* Por otra parte, **tenemos que obtener la corriente que atraviesa la sección lateral de anchura $dz'$**. Se puede calcular como la densidad superficial de corriente $\vec{j}$ multiplicada por la anchura del segmento perpendicular $dz'$ (*comprobar que las unidades son correctas*).

$$
I_1 = j\space dz' = \frac{NI}{L}dz'
$$

Ya tenemos todos los elementos para obtener $\vec{dB}$.

$$
d\vec{B} = \frac{\mu_0 NI \space dz'}{2L} \frac{a^2}{(a^2 + (z-z')^2)^{3/2}} \vec{u_z}
$$

Por último, las espiras diferenciales tienen posiciones desde $z' = -L/2$ hasta $z' = +L/2$, que serán los límites de integración. Con esto, **el campo magnético total, obtenido a partir del campo magnético diferencial de cada espira diferencial** es:

$$
\vec{B} = \int d \vec{B} = \int_{z'=-L/2}^{z'=L/2} \frac{\mu_0 NI \space dz'}{2L} \frac{a^2}{(a^2 + (z-z')^2)^{3/2}} \vec{u_z}
$$

Exactamente la misma expresión que antes. El resultado de la integral es:

$$
\vec{B} = \frac{\mu_0 NI \vec{u_z}}{2L} \left( \frac{z+L/2}{\sqrt{a^2 + (z+L/2)^2}} - \frac{z-L/2}{\sqrt{a^2 + (z-L/2)^2}} \right)
$$


### Apartado $b)$

En el límite cuando el solenoide es infinito, con la densidad superficial de corriente constante $j = \frac{NI}{L} = cte$, tenemos una indeterminación de tipo $\infty-\infty$ que se resuelve de forma algebraica, manipulando la expresión para que la cancelación ocurra antes del límite. Hacemos denominador común y obtenemos:

$$
\lim_{L\rightarrow\infty} \vec{B} = \mu_0 j \vec{u_z}
$$
