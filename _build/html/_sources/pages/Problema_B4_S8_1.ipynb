# Problema B4.S8.1

El campo magnético tiene la siguiente expresión para una configuración de corriente dada:

$$
\vec{B}(\vec{r}) &= \frac{\mu_0}{4\pi}\int_{C'}\frac{(I\vec{dl'})\times(\vec{r}-\vec{r'})}{|\vec{r}-\vec{r'}|^3} \\
&= \frac{\mu_0}{4\pi}\int_{S'}\frac{(\vec{j}(\vec{r'})dS')\times(\vec{r}-\vec{r'})}{|\vec{r}-\vec{r'}|^3} \\
&= \frac{\mu_0}{4\pi}\int_{V'}\frac{(\vec{J}(\vec{r'})dV')\times(\vec{r}-\vec{r'})}{|\vec{r}-\vec{r'}|^3}
$$

Para cada una de los tres tipos de corrientes lineal, superficial y volumétrica. En este ejercicio, la configuración de corriente es lineal, por lo que usamos la primera expresión.

Tenemos que parametrizar la región donde está la fuente y la región donde está el punto de observación.

### Apartado $a)$

Elegimos el sistema de referencia con el punto $P$ en el origen, en previsión de poder usar coordenada polares convenientemente. En este caso el punto de observación $\vec{r} = 0$. Respecto a la fuente, la expresión de la integral la podemos dividir en cuatro tramos, numerados de la siguiente forma:

$$
\vec{B}(\vec{r}) = \frac{\mu_0}{4\pi}\int_{C'}\frac{(I\vec{dl'})\times(\vec{r}-\vec{r'})}{|\vec{r}-\vec{r'}|^3} = \int_1 ... + \int_2 ... + \int_3 ... + \int_4 ...
$$

```{figure} ../images/B4_S8_1_1.png
:alt: hilo de corriente.png
:width: 200px
```

Parametrizamos cada tramo:

1. **Los puntos forman un hilo en el eje $y$**:

$$
\vec{r'} &= y'\vec{u_y} \\
\vec{r} - \vec{r'} &= -y'\vec{u_y} \\
|\vec{r} - \vec{r'}| &= y' \\
dl' &= dy' \quad y' \in [a,b] \\
\vec{dl'} &= dy' \vec{u_y}
$$

Como en este tramo la corriente va hacia arriba, la corriente $I$ de la fórmula de $\vec{B}$ es la $I$ del enunciado del problema (*Es típico encontrarse con la misma letra, y generalmente no hay lugar a confusión*).

Vemos que en la integral tenemos el producto vectorial $(I\vec{dl})\times(\vec{r}-\vec{r'}) = (Idy' \vec{u_y})\times(-y'\vec{u_y}) = 0$, por lo que este campo magnético es nulo. **Esto podríamos haberlo deducido a priori, por que el punto está en la misma linea que el hilo recto de corriente**.

2. **Los puntos forma un hilo en el eje $x$**. Por el mismo motivo $\vec{B_2}=0$.

3. **Los puntos forman un arco circular alrededor del origen**:

$$
\vec{r'} &= a\vec{u_s} \\
\vec{r} - \vec{r'} &= -a\vec{u_s} \\
|\vec{r} - \vec{r'}| &= a \\
dl' &= a \space d\phi' \quad \phi' \in [0,\pi/2] \\
\vec{dl'} &= a\space d\phi' \vec{u_\phi}
$$

```{note}
El punto de observación en polares siempre es $\vec{r'} = r'\vec{u_r}$. En la mayoría de las integrales, $\vec{u_r}$ no es constante, por lo que en la integral no se puede manejar directamente. Tipicamente, lo que hacemos es pasarlo a cartesianas, de tal forma que el vector unitario sale de la integral (*Los unitarios en cartesianas sí son constantes*) y la integral se puede resolver.

Sin embargo, en magnetostática, es común que los productos vectoriales sean más fáciles de manejar **en polares, cilíndricas o esféricas**, por lo que **suele ser conveniente no realizar el cambio a cartesianas hasta después de los productos vectoriales**.
```

En este caso, la integral del campo magnético $\vec{B_3}$:

$$
\vec{B_3} = \frac{\mu_0}{4\pi}\int_{\phi'{=}0}^{\phi'{=}\pi/2}\frac{(Ia\space d\phi' \vec{u_\phi})\times(-a\vec{u_s})}{a^3} = \frac{\mu_0}{4\pi}\int_{\phi'{=}0}^{\phi'{=}\pi/2}\frac{(I\space d\phi' \vec{u_z})}{a} = \frac{\mu_0 I}{8a} \vec{u_z}
$$

Exactamente la cuarta parte del campo magnético de una espira de radio $a$ en su centro.

4. **Los puntos forman un arco circular alrededor del origen**. En este caso, cambiamos el radio $a$ por $b$, y para tener en cuenta que la corriente va en sentido contrario, cambiamos $I$ por $-I$.

$$
\vec{B_4} = -\frac{\mu_0 I}{8b} \vec{u_z}
$$

Por lo que:

$$
\vec{B} = \frac{\mu_0 I}{8}\left(\frac{1}{a} - \frac{1}{b}\right)
$$



### Apartado $b)$

Elegimos el sistema de referencia colocando $P$ en el origen por el mismo motivo, en previsión de poder usar coordenada polares convenientemente. En este caso el punto de observación $\vec{r} = 0$. Respecto a la fuente, la expresión de la integral la podemos dividir en tres tramos, numerados de la siguiente forma:


```{figure} ../images/B4_S8_1_2.png
:alt: hilo de corriente.png
:width: 300px
```

Parametrizamos cada tramo:

1. **Los puntos forman un hilo horizontal desplazado $a$ en la dirección $+y$**:

$$
\vec{r'} &= x'\vec{u_x} + a \vec{u_y} = (x', a, 0) \\
\vec{r} - \vec{r'} &= (-x', -a, 0) \\
|\vec{r} - \vec{r'}| &= \sqrt{x'^2 + a^2} \\
dl' &= dx' \quad x' \in [0,\infty) \\
\vec{dl'} &= dx' \vec{u_x}
$$

Como en este tramo la corriente va hacia la derecha, la corriente $I$ de la fórmula de $\vec{B}$ es la $I$ del enunciado del problema.

$$
\vec{B_1}(\vec{r}) = \frac{\mu_0}{4\pi}\int_{x'{=}0}^{x'{=}\infty}\frac{(I\space dx' \vec{u_x})\times(-x'\vec{u_x} - a \vec{u_y})}{(x'^2 + a^2)^{3/2}} = \frac{\mu_0}{4\pi}\int_{x'{=}0}^{x'{=}\infty}\frac{(-I\space dx' a \vec{u_z})}{(x'^2 + a^2)^{3/2}} = -\frac{\mu_0I\vec{u_z}}{4\pi a}
$$

Es decir, medio hilo infinito a distancia $a$.

2. **Los puntos forman un hilo horizontal desplazado $a$ en la dirección $-y$**. También es otro medio hilo infinito a la misma distancia.

$$
\vec{B_1} = \vec{B_2}
$$

3. **Los puntos forman un arco circular alrededor del origen**. Observamos que se trata de media espira. Observando la dirección de giro, deducimos:

$$
\vec{B_3} = -\frac{\mu_0 I}{4a} \vec{u_z}
$$
