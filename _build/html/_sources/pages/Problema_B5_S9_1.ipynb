# Problema B5.S9.1

En este problema, tenemos un material con una magnetización permanente. La configuración geométrica es la de un cilindro de longitud $L$ y radio $a$ con una magnetización permanente en la dirección del eje de simetría. Solocamos el sistema de referencia para que el eje de simetría coincida con el eje $z$, de tal forma que :

$$
\vec{M} = M\vec{u_z}
$$

```{figure} ../images/B5_S9_1_1.png
:alt: Magnet_cte
:width: 250px
```

```{note}
De forma similar a lo que nos encontrábamos en los ejercicios de Electrostática de materiales, en los problemas de Magnetostática de materiales, vamos a tener tres posibilidades, una de las cuales es particular de los problemas de magnetostática.

1. **Materiales lineales**: Se caracterizan por que la magnetización $\vec{M}$ es lineal con respecto al vector $\vec{H}$, típicamente tan solo una constante (*aunque podría ser una matriz, o tensor de orden 2*).

$$
\vec{M} = \chi_M \vec{H}
$$

2. **Materiales con magnetización constante**: Se caracterizan por que la magnetización es independiente del vector $\vec{H}$ externo. En este caso, las densidades de corriente de magnetización también son constantes, por lo que podemos considerarlas como fuentes de campo magnético $\vec{B}$ y sustituir el material por dichas corrientes.

$$
\vec{M} = cte \quad \Rightarrow \quad \vec{J} = \nabla\times\vec{M}; \space \vec{j} = \vec{M}\times\vec{n}
$$

Ambos casos, **1** y **2**, son equivalentes a problemas de Electrostática de materiales. El siguiente caso, **3**, aunque nos lo podríamos encontrar en Electrostática, allí tiene muy poca importancia, mientras que en Magnetostática tiene muchas aplicaciones prácticas.

3. **Materiales no lineales**: Se caracterizan por que la magnetización $\vec{M}$ responde al vector $\vec{H}$ externo de forma no lineal. Un ejemplo importante son los materiales con histéresis. Típicamente, le damos la vuelta a la dependencia para que $\vec{H}$ sea la variable independiente y $\vec{B}$ sea la variable dependiente:

$$
\vec{M} = f(\vec{H}) \quad \Rightarrow \quad \vec{B} = \mu_0(\vec{M} + f(\vec{H})) = g(\vec{H}) \quad \Rightarrow \quad \vec{H} = \vec{H}(\vec{B})
$$
```

Nos piden el campo magnético en un punto del eje de simetría $z$. El material, en estos casos, se puede sustituir por unas densidades de corriente de magnetización. La densidad (*volumétrica*) de corriente de magnetización:

$$
\vec{J} = \nabla\times\vec{M} = 0
$$

Debido a que la magnetización $\vec{M}$ es constante. Para la superficial, debemos distinguir tres superficies con normal $\vec{n}$ distintas, numeradas en la figura.

1. La tapa superior tiene un vector normal $\vec{n} = \vec{u_z}$. Realizando el producto vectorial, la densidad superficial de corriente es nula:

$$
\vec{j_1} = \vec{M}\times\vec{n} = (M\vec{u_z})\times\vec{u_z} = 0
$$

2. La tapa inferior tiene un vector normal $\vec{n} = -\vec{u_z}$, por lo que genera una densidad superficial de corriente $\vec{j_2}=0$ que también es nula.

3. La cara lateral tiene un vector normal $\vec{n} = \vec{u_s}$ en coordenadas cilíndricas. La densidad superficial de corriente es:

$$
\vec{j_3} = \vec{M}\times\vec{n} = (M\vec{u_z})\times\vec{u_s} = M\vec{u_\phi}
$$

```{note}
En magnetostática, es común que los productos vectoriales sean más fáciles de manejar **en polares, cilíndricas o esféricas**, por lo que **suele ser conveniente no realizar el cambio a cartesianas hasta después de los productos vectoriales**.
```

Por lo tanto, el problema es el de un cilindro de longitud $L$ y radio $a$ con densidad superficial de corriente $\vec{j_3} = M\vec{u_\phi}$. Este problema es el mismo que se resolvió en el boletín anterior, el B4.S8.2. Adaptando la notación:

$$
\frac{NI}{L} \rightarrow M
$$

Con lo que:

$$
\vec{B} = \frac{\mu_0 M \vec{u_z}}{2} \left( \frac{z+L/2}{\sqrt{a^2 + (z+L/2)^2}} - \frac{z-L/2}{\sqrt{a^2 + (z-L/2)^2}} \right)
$$

Por último, nos pide realizar el mismo cálculo utilizando las densidades de polos equivalentes de magnetización. Para ello utilizamos las expresiones de las densidades superficiales de carga con la magnetización:

$$
\sigma_M &= \vec{n}\cdot\vec{M} \\
\rho_M &= -\nabla\cdot\vec{M}
$$

```{note}
En los formularios, podemos encontrarnos $\vec{J_M} = \nabla'\times\vec{M}$ y $\rho_M = -\nabla'\cdot\vec{M}$, sobre las variables primadas. Esto es así por que típicamente, las densidades de corriente de magnetización y las densidades de polos equivalentes magnéticos, al igual que las densidades de carga de polarización, son las fuentes del campo magnético y del campo eléctrico. No siempre tiene que ser así, podríamos encontrar un problema en el que el material magnético fuera el objeto sobre en que se ejerce una fuerza magnética. Sin embargo, es lo común.

En cualquier caso, **las derivadas se realizan con respecto a las coordenadas que definan la posición geométrica del material**, sea primado o sin primar.
```

```{note}
Es importante observar, antes de seguir, que las densidades de polos equivalentes de magnetización, son **una metodología alternativa** con respecto a las densidades de corriente de magnetización. O se usa una, o se usa la otra, ¡pero no las dos!
```

Por el mismo motivo, $\rho_M = $, mientras que realizando los productos:

$$
\sigma_{M,1} &= M \\
\sigma_{M,2} &= -M
$$

Es decir, que la solución puede ser obtenida calculando el campo eléctrico que generan dos discos de carga superficial constante $M$ y $-M$ en las posiciones $z'+\frac{L}{2}$ y $z'-\frac{L}{2}$. Este problema se resolvió en electrostática, en el segundo apartado del problema propuesto B3.S7.P1. Aquel problema solo se propuso en puntos exteriores al cilindro. Vamos a resolverlo aquí para todos los puntos del eje $z$, utilizando el problema de referencia de un disco de radio $a$ con densidad superficial de carga constante $\sigma$ centrado en el origen y contenido en el plano $xy$. La solución es (*verificar, deberíamos poder resolverlo con cierta soltura*):

$$
E_z(z) = \frac{\sigma}{2\varepsilon_0} \left( \frac{z}{|z|} - \frac{z}{\sqrt{a^2 + z^2}}\right)
$$

Observemos que $|z|$ surge de $\sqrt{z^2}$ que no simplificamos para que dé el signo correcto tanto para $z>0$ como para $z<0$. Para obtener el campo magnético del disco de densidad superficial de polos equivalentes de magnetización $M$ en $z'=L/2$ tenemos que realizar los siguientes cambios:

$$
z &\rightarrow z-\frac{L}{2} \\
\sigma &\rightarrow M \\
\frac{1}{\varepsilon_0} &\rightarrow \mu_0
$$

Con lo que:

$$
B_{z,\frac{L}{2},M}(z) = \frac{\mu_0 M}{2} \left( \frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right)
$$

Mientras que para el disco de densidad de polos equivalentes de magnetización $-M$ en $z'=-L/2$: 

$$
B_{z,-\frac{L}{2},-M}(z) = -\frac{\mu_0 M}{2} \left( \frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right)
$$

Para obtener el campo magnético total, no nos podemos olvidar (*comprobar de las demostraciones de las expresiones*) de añadir $\mu_0\vec{M}$ el la región del material. Con esto, distinguimos 3 regiones:

* $z>\frac{L}{2}$: Aquí la magnetización no existe, porque estamos fuera del cilindro por arriba. Ambos cocientes $\frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} = \frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} = 1$ con lo que la suma de los dos términos:

$$
B_z = B_{z,\frac{L}{2},M}(z) + B_{z,-\frac{L}{2},-M}(z) &= \frac{\mu_0 M}{2} \left( \frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right) -\frac{\mu_0 M}{2} \left( \frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right) = \\
&= \frac{\mu_0 M}{2} \left( 1 - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right) -\frac{\mu_0 M}{2} \left( 1 - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right) = \\
&= \frac{\mu_0 M}{2} \left( -\frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}} + \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right)
$$

* $z<-\frac{L}{2}$: Aquí la magnetización tampoco existe, porque estamos fuera del cilindro por abajo. Ambos cocientes $\frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} = \frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} = -1$ con lo que la suma de los dos términos es la misma:

$$
B_z = B_{z,\frac{L}{2},M}(z) + B_{z,-\frac{L}{2},-M}(z) &= \frac{\mu_0 M}{2} \left( \frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right) -\frac{\mu_0 M}{2} \left( \frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right) = \\
&= \frac{\mu_0 M}{2} \left( -1 - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right) -\frac{\mu_0 M}{2} \left( -1 - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right) = \\
&= \frac{\mu_0 M}{2} \left( -\frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}} + \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right)
$$

* $z\in[-\frac{L}{2},\frac{L}{2}]$: Aquí la magnetización es $M_z = M$, por lo que tenemos que sumar el término $\mu_0 M$. Además, los cocientes tienen signo contrario, $\frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} = -1$ y $\frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} = 1$. Por lo tanto, en esta región:

$$
B_z = B_{z,\frac{L}{2},M}(z) + B_{z,-\frac{L}{2},-M}(z) + \mu_0 M &= \frac{\mu_0 M}{2} \left( \frac{z-\frac{L}{2}}{|z-\frac{L}{2}|} - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right) -\frac{\mu_0 M}{2} \left( \frac{z+\frac{L}{2}}{|z+\frac{L}{2}|} - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right) + \mu_0 M = \\
&= \frac{\mu_0 M}{2} \left( -1 - \frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}}\right) -\frac{\mu_0 M}{2} \left( 1 - \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right) + \mu_0 M= \\
&= \frac{\mu_0 M}{2} \left( -\frac{z-\frac{L}{2}}{\sqrt{a^2 + (z-\frac{L}{2})^2}} + \frac{z+\frac{L}{2}}{\sqrt{a^2 + (z+\frac{L}{2})^2}}\right)
$$

**¡Otra vez, dando el mismo resultado!** En general esta verificación no será necesaria, ya demostramos en teoría que las densidades de corriente de magnetización y las densidades de polos equivalentes de magnetización ofrecen el mismo resultado para el campo magnético $\vec{B}$, y aquí hemos podido comprobarlo numéricamente en un ejemplo concreto.
