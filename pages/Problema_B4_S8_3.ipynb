# Problema B4.S8.3

Cada trozo de disco gira con una velocidad angular $\omega$ constante. Colocamos el sistema de referencia en el centro del disco de tal forma que dicho disco está contenido en el plano $xy$. De esta forma, $\omega$ tiene carácter vectorial en la dirección $z$

$$
\vec{\omega} = \omega \vec{u_z}
$$

La posición de cada trozo de disco es, en coordenadas cilíndricas, $\vec{r'} = s'\vec{u_s}$. La velocidad de cada trozo de disco, o diferencial de superficie $dS'$ es en la dirección del vector unitario $u_\phi$, de forma que se cumple:

$$
\vec{v} = \vec{\omega}\times\vec{r'} = \omega s'\vec{u_\phi}
$$

```{note}
El punto de observación en cilíndricas siempre es $\vec{r'} = s'\vec{u_s} + z'\vec{u_z}$. Claramente, los puntos del disco cumplen que $z' = 0$. Recordamos que en magnetostática, es común que los productos vectoriales sean más fáciles de manejar **en polares, cilíndricas o esféricas**, por lo que **suele ser conveniente no realizar el cambio a cartesianas hasta después de los productos vectoriales**.
```

* **El primer apartado pide calcular el campo magnético en el eje $z$**. Para poder calcular el campo magnético $\vec{B}$ necesitamos la densidad de corriente generada por el movimiento de estas cargas superficiales. Dado que la carga total es $Q$ homogéneamente distribuida, concluimos que la densidad superficial de carga es:

$$
\sigma = \frac{Q}{\pi b^2}
$$

Claramente, dado que estamos en una superficie, vamos a tener una densidad superficial de corriente, y vamos a utilizar dicha la expresión del campo magnético $\vec{B}$ en función de dicha densidad superficial de corriente:

$$
\vec{B}(\vec{r}) = \frac{\mu_0}{4\pi}\int_{S'}\frac{(\vec{j}(\vec{r'})dS')\times(\vec{r}-\vec{r'})}{|\vec{r}-\vec{r'}|^3}
$$

Nos falta obtener dicha densidad superficial de corriente:

$$
\vec{j} = \sigma \vec{v} = \frac{Q}{\pi b^2} \omega s'\vec{u_\phi}
$$

(*Comprobar que las unidades son las correctas*)

```{note}
Para recordar o deducir la expresión de $\vec{j}$, primero observamos que tiene **las unidades correctas, de $[A/m]$**. Notamos que **la velocidad debe estar contenida en la superficie** para que la expresión tenga sentido, de forma que las cargas se desplazan dentro de la superficie.

Es similar a la expresión conocida $\vec{J} = \rho \vec{v}$, con unidades de $[A/m^2]$
```

Con esto podemos plantear la integral. El punto de observación está en el eje $z$, por lo que $\vec{r} = z\vec{u_z}$. Recopilando:

$$
\vec{r} &= z\vec{u_z} \\
\vec{r'} &= s'\vec{u_s} \\
\vec{r} - \vec{r'} &= z\vec{u_z} - s'\vec{u_s} \\
|\vec{r} - \vec{r'}| &= \sqrt{z^2 + s'^2} \\
\vec{j} &= \frac{Q}{\pi b^2} \omega s'\vec{u_\phi} \\
dS' &= s'ds'\space d\phi' \quad s' \in [0, b] \quad \phi' \in [0,2\pi]
$$

Con lo que:

$$
\vec{B}(z) = \frac{\mu_0}{4\pi}\int_{s'{=}0}^{s'{=}b}\int_{\phi'{=}0}^{\phi'{=}2\pi} \frac{Q}{\pi b^2} \frac{\omega s'^2ds'\space d\phi'(z\vec{u_s} + s'\vec{u_z})}{(z^2 + s'^2)^{3/2}}
$$

Veamos que por simetría cilíndrica, dado que $\vec{B}$ es una magnitud vectorial y la estamos observando en el eje de simetría, la única componente que no es nula es la componente $z$ (*Desde el punto de vista de la integral, si ponemos $\vec{u_s}$ en cartesianas, $\vec{u_s} = \cos\phi'\vec{u_x} + \sin\phi'\vec{u_y}$, nos quedan dos integrales definidas en $\phi'\in[0,2\pi]$ que dan 0*).

Finalmente, si queremos obtener la expresión analítica de $B_z$.

$$
B_z = \frac{\mu_0 Q\omega}{4\pi^2 b^2}\int_{s'{=}0}^{s'{=}b}\int_{\phi'{=}0}^{\phi'{=}2\pi}  \frac{ s'^3ds'\space d\phi'}{(z^2 + s'^2)^{3/2}} = \frac{\mu_0 Q\omega z}{2\pi b^2}\int_{\frac{s'}{z}{=}0}^{\frac{s'}{z}{=}\frac{b}{z}} \frac{ (\frac{s'}{z})^3(\frac{ds'}{z})}{(1 + (\frac{s'}{z})^2)^{3/2}} = \frac{\mu_0 Q\omega z}{2\pi b^2} \frac{(\frac{z'}{z})^2 + 2}{\sqrt{1 + (\frac{z'}{z})^2}}|_{\frac{s'}{z}{=}0}^{\frac{s'}{z}{=}\frac{b}{z}} = \frac{\mu_0 Q\omega z}{2\pi b^2} \left( \frac{(\frac{b}{z})^2 + 2}{\sqrt{1 + (\frac{b}{z})^2}} - 2\right)
$$

* **El segundo apartado pide el momento dipolar magnético del disco**. Para ello, vamos a beneficiarnos de que conocemos el momento dipolar magnético de una corriente circular, y construimos el momento dipolar magnético total. Para una corriente circular de radio arbitrario $s'$:

$$
m_1 = I_1 \pi s'^2
$$

Dividimos el disco en trozos circulares de grosor $ds'$, de tal forma que la corriente $I_1$ que circula es $I_1 = j\space ds'$. Entonces:

$$
dm = j\space ds' \space \pi s'^2 = \frac{Q \omega s'^3}{b^2} \omega ds'
$$

Con lo que el momento dipolar magnético total es:

$$
m = \int dm = \int_{s'{=}0}^{s'{=}b} \frac{Q \omega s'^3}{b^2} ds' = \frac{Q \omega b^2}{4}
$$

(*Para calcular el momento dipolar magnético podríamos haber usado $s$ en lugar de $s'$, ya que no hay fuente y punto de observación, pero dado que las expresiones anteriores, y en particular la de $j$ que hemos usado, estaban con $s'$, hemos usado ésta por coherencia notacional*)
