# Ejemplo 1 (1.6 Griffiths)

Calcule **la integral de línea** del campo vectorial $ \vec{v} = y^2 \,\hat{x} + 2x(y+1)\,\hat{y} $ desde el punto $a = (1,1,0)$ hasta el punto $b = (2,2,0)$, a lo largo de los caminos (1) y (2) de la figura 1. ¿Cuánto vale $\oint \vec{v} \cdot d\vec{l}$ **para el lazo que va de $a$ a $b$ por el camino (1) y vuelve a $a$ por el camino (2)**?


```{figure} ../images/Ejemplo_B1_1.png
:alt: Ejemplo_B1_1
:width: 330px
Figura 1.
```

---

**Solución:**  
Como siempre,  
$$
d\vec{l} = dx \,\hat{x} + dy \,\hat{y} + dz \,\hat{z}.
$$  

**El camino (1) consta de dos partes**. A lo largo del segmento *horizontal*, $dy = dz = 0$, por lo que (i) $d\vec{l} = dx \,\hat{x}, \quad y = 1$, 

$$
\vec{v} \cdot d\vec{l} = y^2 dx = dx,
$$  

entonces  

$$
\int \vec{v} \cdot d\vec{l} = \int_1^2 dx = 1.
$$

En el tramo *vertical*, $dx = dz = 0$, así que (ii) $d\vec{l} = dy \,\hat{y}, \quad x = 2$,  

$$
\vec{v} \cdot d\vec{l} = 2x(y+1) dy = 4(y+1) dy,
$$  

por tanto  

$$
\int \vec{v} \cdot d\vec{l} = 4 \int_1^2 (y+1) dy = 10.
$$

**Por el camino (1), entonces,**  

$$
\int_a^b \vec{v} \cdot d\vec{l} = 1 + 10 = 11.
$$

---

Mientras tanto, **en el camino (2)**, $x = y$, $dx = dy$ y $dz = 0$, de modo que  

$$
d\vec{l} = dx \,\hat{x} + dx \,\hat{y}, \quad
\vec{v} \cdot d\vec{l} = x^2 dx + 2x(x+1) dx = (3x^2 + 2x) dx,
$$  

por tanto  

$$
\int_a^b \vec{v} \cdot d\vec{l} = \int_1^2 (3x^2 + 2x) dx = (x^3 + x^2)\Big|_1^2 = 10.
$$

```{note}
La estrategia aquí es expresar todo en función de una sola variable; también podría haber eliminado $x$ en favor de $y$; En ocasiones se puede usar un parámetro, si la curva está expresada de forma paramétrica).
```

---

Para el lazo que va por (1) y vuelve por (2), entonces,  

$$
\oint \vec{v} \cdot d\vec{l} = 11 - 10 = 1.
$$
