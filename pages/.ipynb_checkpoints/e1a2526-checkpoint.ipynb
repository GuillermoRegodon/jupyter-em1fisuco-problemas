````{note}
1.  El potencial en la superficie de una esfera de radio $a$ est√° dado por $V = K\cos^3\theta$ donde $K$ es una
constante y $\theta$ es el √°ngulo polar de las coordenadas esf√©ricas. Calcule:

* A. El potencial dentro y fuera de la esfera (3 puntos).
* B. La densidad superficial de carga en la esfera (2 puntos).
````

Este problema es una aplicaci√≥n m√°s o menos directa de la metodolog√≠a de separaci√≥n de variables.

**¬øPor qu√©?** Tenemos que hay que resolver la ecuaci√≥n de Poisson en dos regiones, dentro y fuera de la esfera. En ninguna de las dos regiones hay material ni carga, sino solamente vac√≠o, con la condici√≥n de frontera en la esfera de radio $a$ en la que conocemos el valor del potencial y en la que hay una densidad de carga superficial.

La soluci√≥n es conocida:

$$
\Phi(r,\theta) = \sum_{l=0}^{\infty} \left( A_l r^l + \frac{B_l}{r^{l+1}} \right) P_l(\cos\theta)
$$

Sabiendo que el polinomio de Legendre de orden $l$ tienen un t√©rmino $\cos^l\theta$ y todos los t√©rminos de la misma paridad inferiores, s√≥lo necesitamos los t√©rminos de orden $l=1$ y $l=3$:

$$
\Phi(r,\theta) &=  \left( A_1 r + \frac{B_1}{r^{2}} \right) P_1(\cos\theta) + \left( A_3 r^3 + \frac{B_3}{r^{4}} \right) P_3(\cos\theta) = \\
&= \left( A_1 r + \frac{B_1}{r^{2}} \right) \cos\theta + \left( A_3 r^3 + \frac{B_3}{r^{4}} \right) \frac{1}{2}(5\cos^3\theta - 3\cos\theta)
$$

Por los argumentos comunes de convergencia en el origen y en el infinito, el potencial dentro y fuera:

$$
\Phi_d(r,\theta) &=  A_1 r \cos\theta + A_3 r^3 \frac{1}{2}(5\cos^3\theta - 3\cos\theta) \\
\Phi_f(r,\theta) &=  \frac{B_1}{r^{2}} \cos\theta + \frac{B_3}{r^{4}} \frac{1}{2}(5\cos^3\theta - 3\cos\theta)
$$

Como los coeficientes $A_l$ solo son v√°lidos dentro y los $B_l$ solo fuera, no es necesario distinguir los coeficientes de otra manera. Utilizamos la condici√≥n de frontera en $r=a$.

$$
\Phi_d(r{=}a,\theta) &=  A_1 a \cos\theta + A_3 a^3 \frac{1}{2}(5\cos^3\theta - 3\cos\theta) = K\cos^3\theta\\
\Phi_f(r{=}a,\theta) &=  \frac{B_1}{a^{2}} \cos\theta + \frac{B_3}{a^{4}} \frac{1}{2}(5\cos^3\theta - 3\cos\theta) = K\cos^3\theta
$$

Igualando t√©rminos de $\cos^l\theta$.

$$
\frac{5}{2} A_3 a^3 &= K \\
A_1 a - \frac{3}{2} A_3 a^3 &= 0 \\
\frac{5}{2} \frac{B_3}{a^{4}} &= K \\
\frac{B_1}{a^{2}} - \frac{3}{2} \frac{B_3}{a^{4}} &= 0
$$

Lo que nos permite obtener todos los coeficientes y las expresiones finales del potencial:

$$
\Phi_d(r,\theta) &= \frac{3Kr}{5a} \cos^3\theta + \frac{2Kr^3}{5a^3} \frac{1}{2}(5\cos^3\theta - 3\cos\theta) = K \frac{r^3}{a^3}¬†\cos^3\theta + K \frac{3}{5} \left( \frac{r}{a} - \frac{r^3}{a^3} \right) \cos\theta \\
\Phi_f(r,\theta) &= \frac{3Ka^2}{5r^2} \cos^3\theta + \frac{2Ka^4}{5r^{4}} \frac{1}{2}(5\cos^3\theta - 3\cos\theta) = K \frac{a^4}{r^4} \cos^3\theta + K \frac{3}{5} \left( \frac{a^2}{r^2} - \frac{a^4}{r^4} \right) \cos\theta 
$$

Para obtener la densidad superficial de carga en la esfera, que es la que genera este potencial, solo tenemos una opci√≥n, que es comparando en campo el√©ctrico dentro y fuera:

$$
\sigma_S = \varepsilon_0 (E_{f\bot} - E_{d\bot})
$$

La superficie es una esfera, por lo que la direcci√≥n perpendicular es la radial $\vec{u_r}$. Comparando con la expresi√≥n del gradiente en esf√©ricas, podemos comprobar que la componente radial del campo el√©ctrico es:

$$
E_\bot = -\frac{\partial \Phi}{\partial r}
$$

Por lo que realizando las derivadas, en $r=a$, y restando, obtenemos la densidad superficial de carga pedida.

$$
E_{d\bot} &= - K 3\frac{r^2}{a^3} \cos^3\theta - K \frac{3}{5} \left( \frac{1}{a} - 3\frac{r^2}{a^3} \right) \cos\theta \\
E_{f\bot} &= K 4\frac{a^4}{r^5} \cos^3\theta + K \frac{3}{5} \left( 2\frac{a^2}{r^3} - 4\frac{a^4}{r^5} \right) \cos\theta 
$$

$$
\sigma_S = \varepsilon_0 \frac{K}{a} \left( 7 \cos^3\theta - \frac{12}{5} \cos\theta \right)
$$








````{note}
2. Se coloca una esfera met√°lica de radio $a$muy peque√±o a una distancia $b\gg a$ de un plano infinito de metal. Ambos metales se consideran conductores el√©ctricos perfectos. La esfera met√°lica tiene una carga total $Q$.
* A. Calcule el potencial al que se encuentra la esfera met√°lica (4 puntos).
* B. Calcule el valor de la capacidad que se forma entre la esfera y el conductor (1 punto).
````

En este problema debemos hacer algunas aproximaciones para obtener unas expresiones manejables. La esfera muy peque√±a con una carga total $Q$ va a tener una imagen con respecto al plano con la misma forma y carga total $-Q$.

```{figure} ../images/e1a2526_1.png
:alt: e1a2526_1.png
:width: 250px
```

Si la esfera fuera grande, claramente la densidad superficial de carga no ser√≠a uniforme, y en el l√≠mite contrario, la esfera se convierte en una carga puntual $Q$. En primera aproximaci√≥n, vamos a considerar la esfera de carga como uniformemente distribuida. El potencial ser√≠a el de una esfera hueca de carga homog√©nea centrada en su posici√≥n, mientras que el de la imagen es similar:

$$
\Phi(\vec{r}) = \Phi_Q(\vec{r}) + \Phi_{-Q}(\vec{r}) = \frac{Q}{4\pi\varepsilon_0|\vec{r}-\vec{r_Q}|} + \frac{-Q}{4\pi\varepsilon_0|\vec{r}-\vec{r_{-Q}}|}
$$

El potencial en la superficie de la esfera original se divide en dos t√©rminos. En el primero de los dos, la distancia entre la esfera y su centro es obviamente $|\vec{r}-\vec{r_Q}|=a$.

$$
\Phi_{Q, esf} \approx \frac{Q}{4\pi\varepsilon_0 a}
$$

El segundo de los t√©rminos, el de la carga imagen, es aproximadamente el que genera a una distancia $2b$, dos veces la distancia entre la esfera y el plano.

$$
\Phi_{-Q, esf} \approx \frac{-Q}{4\pi\varepsilon_0 (2b)}
$$

Para obtener este resultado hemos hecho varias aproximaciones:

* La carga de la esfera est√° distribuida de forma pr√°cticamente homog√©nea. El potencial que genera es tambi√©n pr√°cticamente radial.

* El potencial que genera la imagen (*o equivalentemente, el plano de metal con densidad superficial de carga*) es pr√°cticamente constante en la regi√≥n de la esfera de carga.

La capacidad de la configuraci√≥n entre la esfera y el plano se calcula dividiendo la diferencia de potencial entre los electrodos $V = \Phi_{esf} = \Phi_{Q,esf} + \Phi_{-Q,esf}$ y la carga $Q$:

$$
C = \frac{Q}{V} \approx \frac{4\pi\varepsilon_0}{Q(\frac{1}{a} - \frac{1}{2b})}
$$







````{note}
3. Una bobina de secci√≥n rectangular tiene un radio interior $a$, un radio exterior $b$ y una altura $h$. La bobina tiene un arrollamiento de ùëÅ espiras rectangulares de dimensiones $b-a$ y $h$ por las que circula una corriente $I$. El interior de la bobina tiene un material con permeabilidad magn√©tica $\mu$, mientras que el exterior es vac√≠o.
* A. Calcule el campo magn√©tico en el interior de la bobina de secci√≥n rectangular. Tenga en cuenta que depende de la coordenada radial $s$ de las coordenadas cil√≠ndricas (2 puntos).
* B. Calcule las densidades de corriente equivalentes de magnetizaci√≥n que aparecen en el material (2 puntos).
* C. Si el arrollamiento se acumula en un trozo del material, qu√© condici√≥n debe cumplir el material para que la soluci√≥n del apartado A siga siendo v√°lida (1 puntos).

```{figure} ../images/e1a2526_2.png
:alt: e1a2526_2.png
:width: 550px
```
````

Esta es una de las configuraciones de corriente con simetr√≠a cil√≠ndrica que podemos resolver por Ampere. Para poder distinguir las direcciones unitarias en coordenadas cil√≠ndricas, ponemos el sistema de referencia.

```{figure} ../images/e1a2526_3.png
:alt: e1a2526_3.png
:width: 300px
```

Vemos que siguiente la regla de la mano derecha, la direcci√≥n del campo magn√©tico es en el sentido de $-\vec{u_\phi}$. Planteamos el circuito de Ampere con forma de circunferencia centrada en el eje de tal forma que la circunferencia est√° integramente en el interior del material:

$$
\oint_C \vec{H}\cdot\vec{dl} = I_{enc}
$$

Comenzando con el t√©rmino de la izquierda, vemos que la direcci√≥n de $\vec{H}$ es en el sentido de $-\vec{u_\phi}$, es decir, $\vec{H} = -H \vec{u_\phi}$. El t√©rmino $\vec{dl}$ podemos recorrerlo en el sentido $\vec{u_\phi}$ y manejar el signo negativo como parte de la corriente, o directamente recorrerlo en el sentido $-\vec{u_\phi}$. Los dos dan lo mismo. Elegimos la primera opci√≥n.

$$
\oint_C \vec{H}\cdot\vec{dl} = \oint_C (-H \vec{u_\phi})\cdot(dl\vec{u_\phi}) = -H 2\pi s
$$

Hemos resuelto la integral gracias a que $\vec{H}$ y $\vec{dl}$ son colineales y de sentido contrario, el m√≥dulo de $\vec{H}$ en constante por simetr√≠a cil√≠ndrica, y que $dl = s \space d\phi$.

El t√©rmino de la derecha es la corriente total encerrada. Habiendo elegido esta direcci√≥n la recorrer el circuito de Ampere, la direcci√≥n positiva de la superficie encerrada es hacia arriba, $\vec{u_z}$. Vemos que todos los hilos de corriente la cruzan en el sentido contrario, por lo que:

$$
I_{enc} = -N I
$$

Con esto,

$$
\vec{H} = -H \vec{u_\phi} = -\frac{N I}{2\pi s}\vec{u_\phi}
$$

Y por supuesto, el campo magn√©tico $B$:

$$
\vec{B} = -\frac{\mu N I}{2\pi s}\vec{u_\phi}
$$

Para obtener las densidades de corriente equivalentes de magnetizaci√≥n necesitamos la magentizaci√≥n $\vec{M}$. Del formulario:

$$
\vec{B} = \mu_0(\vec{M} + \vec{H}) = \mu\vec{H}
$$

De hay que:

$$
\vec{M} = (\frac{\mu}{\mu_0} - 1)\vec{H} = -(\frac{\mu}{\mu_0} - 1)\frac{N I}{2\pi s}\vec{u_\phi} = -M\vec{u_\phi}
$$

Podemos obtener la densidad de corriente volum√©trica equivalente de magnetizaci√≥n a trav√©s del rotacional en cil√≠ndricas, pero como el material es lineal y $\mu$ es constante (*el material es homog√©neo*), va a dar $\vec{J_M} = 0$. La densidad de corriente superficial de magnetizaci√≥n aparece en las cuatro fronteras del material:

* La cara superior: $\vec{n} = \vec{u_z}$, con lo que $\vec{j_M} = \vec{M}\times\vec{n} = -M \vec{u_s} = -(\frac{\mu}{\mu_0} - 1)\frac{N I}{2\pi s} \vec{u_s}$

* La cara inferior: $\vec{n} = -\vec{u_z}$, con lo que $\vec{j_M} = \vec{M}\times\vec{n} = M \vec{u_s} = (\frac{\mu}{\mu_0} - 1)\frac{N I}{2\pi s} \vec{u_s}$

* La cara exterior: $\vec{n} = \vec{u_s}$ y $s=b$, con lo que $\vec{j_M} = \vec{M}\times\vec{n} = M \vec{u_z} = (\frac{\mu}{\mu_0} - 1)\frac{N I}{2\pi b} \vec{u_z}$

* La cara interior: $\vec{n} = -\vec{u_s}$ y $s=a$, con lo que $\vec{j_M} = \vec{M}\times\vec{n} = -M \vec{u_z} = -(\frac{\mu}{\mu_0} - 1)\frac{N I}{2\pi a} \vec{u_z} $

Por √∫ltimo, en el apartado C el arrollamiento se acumula en una regi√≥n del toroide. Para que la soluci√≥n sea la misma, el campo magn√©tico debe quedarse dentro del material. Para ello, la permeabilidad magn√©tica del material debe ser muy alta:

$$
\mu\gg\mu_0
$$

Esta es la condici√≥n que le ped√≠amos a los circuitos ferromagn√©ticos para que el flujo de campo magn√©tico fuera constante y todo el campo magn√©tico pasara por dentro del material. Observamos que para esta condici√≥n, la densidad de corriente superficial de magnetizaci√≥n es muy elevada, mucho mayor que la densidad de corriente libre superficial generada por los hilos con corriente $I$. 






````{note}
4. Calcule el coeficiente de inducci√≥n mutua entre dos dipolos magn√©ticos con la misma orientaci√≥n en cualquier posici√≥n relativa (5 puntos).
````

Los dipolos magn√©ticos se caracterizan por la corriente que los circula y por la superficie que encierra dicha corriente.

$$
m = I S
$$

En el problema queremos tener en cuenta cualquier posici√≥n relativa. Uno de los dos dipolos podemos ponerlo como queramos, que es lo mismo que poner el sistema de referencia donde nos convenga. Por ejemplo, para poder utilizar las expresiones del formulario del campo magn√©tico de un dipolo magn√©tico, ponemos dicho dipolo en el origen, apuntando en la direcci√≥n y sentido $\vec{u_z}$. El segundo dipolo tendr√° la misma direcci√≥n, pero estar√° colocado en cualquier posici√≥n del espacio que llamamos gen√©ricamente $\vec{r} = r\vec{u_r}$

El campo magn√©tico generado por el primer dipolo, en el origen, es:

$$
\vec{B} = \frac{\mu_0}{4\pi}\left( \frac{3(\vec{m}\cdot\vec{r})\vec{r}}{r^5} - \frac{\vec{m}}{r^3} \right)
$$

Para calcular el flujo de campo magn√©tico que atraviesa el segundo dipolo, en $\vec{r}$, calculamos la siguiente integral:

$$
\Phi_{21} = \int\vec{B}\cdot\vec{dS}
$$

Dado que el dipolo tiene una superficie peque√±a, podemos aproximar el campo magn√©tico como aproximadamente constante en toda la regi√≥n donde est√° el segundo dipolo en torno a $\vec{r}$. De esta forma:

$$
\Phi_{21} \approx \vec{B}\cdot\vec{S} = \frac{\mu_0}{4\pi}\left( \frac{3(\vec{m}\cdot\vec{r})\vec{r}}{r^5} - \frac{\vec{m}}{r^3} \right) \cdot \vec{S}
$$

Ahora utilizamos $\vec{m} = m\vec{u_z}$. Claramente, la superficie va en la misma direcci√≥n, $\vec{S} = S\vec{u_z}$. Con esto:

$$
\Phi_{21} \approx \frac{\mu_0 S}{4\pi}\left( \frac{3(m\vec{u_z}\cdot\vec{r})(\vec{r}\cdot\vec{u_z})}{r^5} - \frac{m\vec{u_z}\cdot\vec{u_z}}{r^3} \right) = \frac{\mu_0 S m}{4\pi}\left( \frac{3\cos^2\theta - 1}{r^3} \right)
$$

Para calcular el coeficiente de inducci√≥n mutua, $M_{21}$, utilizamos la expresi√≥n del formulario:

$$
M_{21} = \frac{\Phi_{21}}{I_1}
$$

Utilizando de nuevo que $m = SI_1$, por √∫ltimo, obtenemos:

$$
M_{21} \approx \frac{\mu_0 S^2}{4\pi}\left( \frac{3\cos^2\theta - 1}{r^3} \right)
$$
